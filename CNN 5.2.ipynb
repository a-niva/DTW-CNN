```python
import glob
import torch.nn.functional as F
from contextlib import nullcontext
import concurrent.futures  # Pour acc√©der √† concurrent.futures.as_completed
import pyarrow.parquet as pq
from pyarrow import parquet
import numpy as np
import pandas as pd
import os
import time
from datetime import datetime
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import faiss
import matplotlib.pyplot as plt
from tqdm import tqdm
import pickle
import h5py
import psutil
import logging
import json
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
from sklearn.manifold import TSNE
from typing import Dict, List, Tuple, Optional, Union, Any
import warnings
import random
# Imports pour l'optimisation Bay√©sienne
from skopt import gp_minimize
from skopt.space import Real, Integer, Categorical
from skopt.callbacks import CheckpointSaver
warnings.filterwarnings('ignore')




# Optimisations CUDA et cuDNN
if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = True  # Acc√©l√®re les convolutions
    torch.backends.cuda.matmul.allow_tf32 = True  # Utilise TF32 sur Ampere+ GPUs
    torch.backends.cudnn.allow_tf32 = True
    
use_mixed_precision = True  # Mettre √† False si instable
if use_mixed_precision and torch.cuda.is_available():
    scaler = torch.cuda.amp.GradScaler()

class Config:
    # Base paths and storage
    BASE_PATH = r"C:\Users\Alex\Python\Finance\@@@ Downloader @@@\parquet\datasets_technicals"
    EMBEDDINGS_PATH = os.path.join(BASE_PATH, "embeddings")
    INDEX_PATH = os.path.join(BASE_PATH, "vector_index")
    MODELS_PATH = os.path.join(BASE_PATH, "models")
    LOGS_PATH = os.path.join(BASE_PATH, "logs")
    OPTIMIZATION_RESULTS_PATH = os.path.join(BASE_PATH, "optimization_results")
    
    # Data parameters
    INTERVALS = ['15m']
    METRICS = [
        'Open', 'High', 'Low', 'Close', 'Volume',
        'RSI14', 'MACD', 'SMA50', 'EMA20', 'log_return'
    ]
    WINDOW_SIZES = {
        '1d': 60,
        '1h': 120,
        '30m': 180,
        '15m': 240,
        '5m': 300,
        'default': 60
    }
    
    # Model parameters
    EMBEDDING_DIM = 128
    USE_MIXED_PRECISION = True
    
    # GPU and optimization settings
    USE_GPU = torch.cuda.is_available()
    DEVICE = torch.device('cuda' if USE_GPU else 'cpu')
    NUM_WORKERS = max(1, os.cpu_count() - 1)
    
    # Training parameters
    BATCH_SIZE = 1024
    MAX_BATCH_SIZE = 16384 * 4
    MIN_BATCH_SIZE = 1024
    CHUNK_SIZE = 1000
    NUM_EPOCHS = 200
    LEARNING_RATE = 0.0005
    WEIGHT_DECAY = 1e-5
    VALIDATION_SPLIT = 0.2
    EARLY_STOPPING_PATIENCE = 15
    
    # Analysis parameters
    FORECAST_HORIZON = 20
    HISTORICAL_BARS = 30
    N_SIMILAR = 40
    MIN_SIMILARITY = 0.01
    
    # Cache settings
    CACHE_SIZE = 200
    CACHE_CAPACITY_MB = 2000
    
    # DTW parameters
    DTW_WINDOW = 0.1
    DTW_USE_FAST = True
    DTW_USE_PRUNING = True
    DTW_PERCENT = 0.25
    
    # Optimization parameters
    OPT_N_TRIALS = 10
    OPT_SAMPLE_SIZE = 50
    OPT_METRIC_WEIGHTS = {
        'direction_weight': 0.3,
        'magnitude_weight': 0.3,
        'ir_weight': 0.2,
        'calibration_weight': 0.1,
        'pattern_var_weight': 0.1
    }
    
    # Misc
    LOG_LEVEL = 'INFO'
    
    @staticmethod
    def get_window_size(interval):
        return Config.WINDOW_SIZES.get(interval, Config.WINDOW_SIZES['default'])
    
    @staticmethod
    def get_batch_size():
        return Config.BATCH_SIZE
    
    @staticmethod
    def get_max_window():
        return max(Config.WINDOW_SIZES.values())
    
    @staticmethod
    def get_optimization_config():
        return {
            'n_trials': Config.OPT_N_TRIALS,
            'optimization_sample_size': Config.OPT_SAMPLE_SIZE,
            'dtw_percent': Config.DTW_PERCENT,
            'metric_weights': Config.OPT_METRIC_WEIGHTS,
            'use_gpu': Config.USE_GPU,
            'device': Config.DEVICE,
            'use_mixed_precision': Config.USE_MIXED_PRECISION,
            'batch_size': Config.BATCH_SIZE
        }
    
    @staticmethod
    def get_dtw_config():
        return {
            'window': Config.DTW_WINDOW,
            'use_fast': Config.DTW_USE_FAST,
            'use_pruning': Config.DTW_USE_PRUNING
        }
    
    @staticmethod
    def initialize():
        os.makedirs(Config.EMBEDDINGS_PATH, exist_ok=True)
        os.makedirs(Config.INDEX_PATH, exist_ok=True)
        os.makedirs(Config.MODELS_PATH, exist_ok=True)
        os.makedirs(Config.LOGS_PATH, exist_ok=True)
        os.makedirs(Config.OPTIMIZATION_RESULTS_PATH, exist_ok=True)
        if Config.USE_GPU and not torch.cuda.is_available():
            warnings.warn("GPU requested but not available, falling back to CPU")
            Config.USE_GPU = False
            Config.DEVICE = torch.device('cpu')
    
    @staticmethod
    def get_runtime_config():
        return {
            'intervals': Config.INTERVALS,
            'metrics': Config.METRICS,
            'window_sizes': Config.WINDOW_SIZES,
            'embedding_dim': Config.EMBEDDING_DIM,
            'batch_size': Config.BATCH_SIZE,
            'device': Config.DEVICE,
            'use_gpu': Config.USE_GPU,
            'cache_size': Config.CACHE_SIZE,
            'dtw_config': Config.get_dtw_config(),
            'optimization': Config.get_optimization_config()
        }



logger = logging.getLogger(__name__)
logger.setLevel(Config.LOG_LEVEL)
handler = logging.StreamHandler()
handler.setFormatter(logging.Formatter('%(asctime)s %(levelname)s %(message)s'))
logger.addHandler(handler)


class StreamingDataManager:
    """Gestionnaire de donn√©es optimis√© avec chargement par chunks et mise en cache efficace."""
    
    def __init__(self, base_path, chunk_size=Config.CHUNK_SIZE, cache_capacity_mb=Config.CACHE_CAPACITY_MB, use_gpu=Config.USE_GPU):
        self.base_path = base_path
        self.chunk_size = chunk_size
        self.cache_capacity_mb = cache_capacity_mb
        self.device = Config.DEVICE
        
        # True LRU cache with access tracking
        from collections import OrderedDict
        self.data_cache = OrderedDict()
        self.cache_size = 0
        self.cache_hits = 0
        self.cache_misses = 0
        
        # Cache for DataFrame sizes to avoid recalculation
        self._df_size_cache = {}
        
        # Index available tickers
        self.available_tickers = self._build_ticker_index()
        
        # Metrics used
        self.metrics = Config.METRICS
        
        logger.info(f"DataManager initialized: {self.device}, Cache: {cache_capacity_mb}MB")
    
    def _calculate_df_size_mb(self, df):
        """Calculates DataFrame memory size with caching for efficiency."""
        # Generate signature key based on shape and dtypes
        shape_key = (df.shape[0], df.shape[1])
        dtypes_key = tuple(str(dtype) for dtype in df.dtypes)
        df_key = hash((shape_key, dtypes_key))
        
        # Check if we've already calculated size for similar DataFrame
        if df_key in self._df_size_cache:
            return self._df_size_cache[df_key]
        
        # Calculate actual size
        size_mb = df.memory_usage(deep=True).sum() / (1024 * 1024)
        
        # Cache the result
        self._df_size_cache[df_key] = size_mb
        
        # Limit size cache to 1000 entries
        if len(self._df_size_cache) > 1000:
            # Remove oldest entries
            keys_to_remove = list(self._df_size_cache.keys())[:100]
            for key in keys_to_remove:
                self._df_size_cache.pop(key)
        
        return size_mb
    
    def _add_to_cache(self, key, data):
        """Adds data to cache with LRU eviction policy and efficient size calculation."""
        # Calculate memory size once using our cached method
        data_size_mb = self._calculate_df_size_mb(data)
        
        # Skip if single item is larger than cache capacity
        if data_size_mb > self.cache_capacity_mb:
            logger.warning(f"Data for {key} exceeds cache capacity ({data_size_mb:.2f}MB > {self.cache_capacity_mb}MB)")
            return
        
        # Ensure the cache has enough space
        while (self.cache_size + data_size_mb) > self.cache_capacity_mb and self.data_cache:
            # Remove the least recently used item (first item in OrderedDict)
            oldest_key, oldest_data = self.data_cache.popitem(last=False)
            oldest_size = self._calculate_df_size_mb(oldest_data)
            self.cache_size -= oldest_size
            
            logger.debug(f"Evicted {oldest_key} from cache ({oldest_size:.2f}MB), new size: {self.cache_size:.2f}MB")
        
        # Add to cache (at the end of OrderedDict for most recent)
        self.data_cache[key] = data
        self.cache_size += data_size_mb
        
        logger.debug(f"Added {key} to cache ({data_size_mb:.2f}MB), current size: {self.cache_size:.2f}MB")
    
    def get_cache_stats(self):
        """Returns detailed cache statistics."""
        return {
            'size_mb': self.cache_size,
            'capacity_mb': self.cache_capacity_mb,
            'utilization': self.cache_size / max(1, self.cache_capacity_mb),
            'entries': len(self.data_cache),
            'hits': self.cache_hits,
            'misses': self.cache_misses,
            'hit_ratio': self.cache_hits / max(1, (self.cache_hits + self.cache_misses))
        }
    
    def load_ticker_chunk(self, ticker, interval, start_idx=0, chunk_size=None):
        """Loads a chunk of data for a ticker with improved LRU cache."""
        chunk_size = chunk_size or self.chunk_size
        
        # Cache key
        cache_key = f"{ticker}_{interval}_{start_idx}_{chunk_size}"
        
        # Check cache with proper LRU update
        if cache_key in self.data_cache:
            # Move accessed item to end of OrderedDict (most recent)
            self.data_cache.move_to_end(cache_key)
            self.cache_hits += 1
            return self.data_cache[cache_key]
        
        self.cache_misses += 1
            
        # File path
        file_path = os.path.join(Config.BASE_PATH, interval, f"{ticker}.parquet")
        
        if not os.path.exists(file_path):
            return None
            
        try:
            # Use parquet row group filtering when possible
            file_size = os.path.getsize(file_path)
            use_filters = file_size > 10 * 1024 * 1024  # Only use filtering for files > 10MB
            
            if use_filters:
                # Get metadata to determine total rows
                parquet_file = pq.ParquetFile(file_path)
                total_rows = parquet_file.metadata.num_rows
                
                # Calculate valid range
                if start_idx >= total_rows:
                    return None
                    
                end_idx = min(start_idx + chunk_size, total_rows)
                
                # Read only the specific rows needed
                ticker_data = pd.read_parquet(
                    file_path,
                    engine='pyarrow',
                    filters=[('index', '>=', start_idx), ('index', '<', end_idx)]
                )
            else:
                # For smaller files, read all and slice in memory
                ticker_data = pd.read_parquet(file_path)
                
                # Filter to get the requested chunk
                if start_idx >= len(ticker_data):
                    return None
                    
                end_idx = min(start_idx + chunk_size, len(ticker_data))
                ticker_data = ticker_data.iloc[start_idx:end_idx]
            
            # Check essential metrics
            essential_metrics = ['Open', 'High', 'Low', 'Close', 'Volume']
            missing_metrics = [col for col in essential_metrics if col not in ticker_data.columns]
            
            if missing_metrics:
                return None
                
            # Calculate log_return if necessary
            if 'log_return' not in ticker_data.columns:
                ticker_data['log_return'] = np.log(ticker_data['Close']).diff()
                
            # Add to cache with capacity management
            self._add_to_cache(cache_key, ticker_data)
            
            return ticker_data
            
        except Exception as e:
            logger.error(f"Error loading chunk for {ticker}_{interval}: {str(e)}")
            return None
    
    def _load_ticker_data(self, ticker, interval):
        """Loads the complete ticker data with improved file handling."""
        file_path = os.path.join(Config.BASE_PATH, interval, f"{ticker}.parquet")
        
        # V√©rification robuste des fichiers
        if not os.path.exists(file_path):
            # V√©rifier si le ticker existe avec une extension (par ex. "TICKER.EXT.parquet")
            interval_path = os.path.join(self.base_path, interval)
            if os.path.exists(interval_path):
                matching_files = [f for f in os.listdir(interval_path) 
                                 if f.startswith(f"{ticker}.") and f.endswith('.parquet')]
                
                if matching_files:
                    file_path = os.path.join(interval_path, matching_files[0])
                    logger.debug(f"Utilisation d'un fichier alternatif pour {ticker}: {matching_files[0]}")
                else:
                    logger.warning(f"File not found: {file_path}")
                    return None
            else:
                logger.warning(f"Interval directory not found: {interval_path}")
                return None
        
        try:
            # Read the file and log basic info
            ticker_data = pd.read_parquet(file_path)
            logger.debug(f"Loaded {ticker} data: {ticker_data.shape}, columns: {ticker_data.columns.tolist()}")
            
            # Ensure required columns exist
            essential_metrics = ['Open', 'High', 'Low', 'Close', 'Volume']
            missing_metrics = [col for col in essential_metrics if col not in ticker_data.columns]
            if missing_metrics:
                logger.warning(f"Missing essential metrics for {ticker}: {missing_metrics}")
                return None
                
            # Calculate log_return if necessary
            if 'log_return' not in ticker_data.columns:
                ticker_data['log_return'] = np.log(ticker_data['Close']).diff()
                
            return ticker_data
        except Exception as e:
            logger.error(f"Error loading data for {ticker}_{interval}: {str(e)}")
            return None
    
    def get_latest_reversal_date(self, ticker, interval):
        """Determines the most recent price reversal date for a ticker."""
        # Load recent data
        data = self.load_ticker_chunk(ticker, interval, chunk_size=100)
        if data is None or len(data) < 5:
            return None
        return data.index[-1]

    
    def _build_ticker_index(self):
        """Construit un index efficace des tickers disponibles."""
        index = {}
        
        for interval in Config.INTERVALS:
            interval_path = os.path.join(self.base_path, interval)
            if not os.path.exists(interval_path):
                continue
                
            # Utiliser listdir pour obtenir les fichiers parquet (plus rapide que glob)
            parquet_files = [f for f in os.listdir(interval_path) if f.endswith('.parquet')]
            
            # CORRIG√â: Pr√©server le nom complet du ticker (sans l'extension .parquet)
            tickers = [os.path.splitext(f)[0] for f in parquet_files]  # Remove extension properly
            
            # Stocker la liste avec m√©tadonn√©es minimales
            index[interval] = {ticker: {'available': True} for ticker in tickers}
            
        return index



    
    def create_tensor_window(self, ticker, interval, window_size=None):
        """Creates a tensor for time window analysis with GPU memory monitoring."""
        if window_size is None:
            window_size = Config.get_window_size(interval)
        # Log GPU memory at start
        if 'gpu_tracker' in globals():
            globals()['gpu_tracker'].log_level = logging.DEBUG
            globals()['gpu_tracker'].log(f"create_tensor_window_start: {ticker}")
        
        # Load data
        raw_data = self.load_ticker_chunk(ticker, interval, chunk_size=window_size + 20)
        if raw_data is None or len(raw_data) < window_size:
            return None

        # Preprocess and normalize
        processed_data = enhanced_financial_preprocessing(raw_data, normalize=True)
        if processed_data is None or len(processed_data) < window_size:
            return None
        
        # Check for sufficient valid data (at least 30% per column)
        min_valid_ratio = 0.3
        valid_cols = []
        for col in processed_data.columns:
            valid_ratio = processed_data[col].notna().mean()
            if valid_ratio >= min_valid_ratio:
                valid_cols.append(col)
        
        # Ensure at least prices and volumes are available
        essential_cols = ['Close_norm', 'Open_norm']
        if not all(col in valid_cols for col in essential_cols) or len(valid_cols) < 3:
            return None
        
        # Filter to keep only columns with sufficient data
        processed_data = processed_data[valid_cols]
        
        # Create tensor with NaN
        n_features = processed_data.shape[1]
        n_windows = max(1, len(processed_data) - window_size + 1)
        
        # Log memory before tensor allocation
        if 'gpu_tracker' in globals():
            globals()['gpu_tracker'].log_level = logging.DEBUG
            globals()['gpu_tracker'].log(f"pre_tensor_alloc: {ticker}", 
                                        f"shape: ({n_windows}, {n_features}, {window_size})")
        
        # Optimize memory allocation strategy for GPU efficiency
        try:
            # More efficient to create on CPU first for small tensors
            if n_windows * n_features * window_size < 100000:
                # For small tensors, create on CPU then transfer
                windows_tensor = torch.full(
                    (n_windows, n_features, window_size),
                    float('nan'),
                    dtype=torch.float32
                )
                # Only transfer to GPU at the end
                if str(self.device) != 'cpu':
                    windows_tensor = windows_tensor.to(self.device)
            else:
                # For large tensors, create directly on device
                windows_tensor = torch.full(
                    (n_windows, n_features, window_size),
                    float('nan'),
                    dtype=torch.float32,
                    device=self.device
                )
        except RuntimeError as e:
            # Fall back to CPU if GPU memory is insufficient
            logger.warning(f"GPU memory error for {ticker}, falling back to CPU: {str(e)}")
            windows_tensor = torch.full(
                (n_windows, n_features, window_size),
                float('nan'),
                dtype=torch.float32
            )
        
        # Log memory after tensor allocation
        if 'gpu_tracker' in globals():
            globals()['gpu_tracker'].log_level = logging.DEBUG
            globals()['gpu_tracker'].log(f"post_tensor_alloc: {ticker}")
        
        # Fill tensor in batches
        batch_size = Config.get_batch_size()
        values = processed_data.values  # With NaNs preserved
        
        for i in range(0, n_windows, batch_size):
            batch_end = min(i + batch_size, n_windows)
            batch_indices = range(i, batch_end)
            
            # Create batch preserving NaNs
            batch_data = np.array([values[j:j+window_size].T for j in batch_indices], dtype=np.float32)
            
            # Convert to tensor - use CPU for conversion then transfer
            if str(self.device) != 'cpu':
                batch_tensor = torch.tensor(batch_data, dtype=torch.float32)
                # Use non-blocking transfer for efficiency
                windows_tensor[i:batch_end] = batch_tensor.to(self.device, non_blocking=True)
            else:
                batch_tensor = torch.tensor(batch_data, dtype=torch.float32)
                windows_tensor[i:batch_end] = batch_tensor
        
        # Log final GPU memory
        if 'gpu_tracker' in globals():
            globals()['gpu_tracker'].log_level = logging.DEBUG
            globals()['gpu_tracker'].log(f"create_tensor_window_end: {ticker}", 
                                    f"final shape: {windows_tensor.shape}")
        
        return windows_tensor
        
    def get_available_tickers(self, interval, min_data_points=100):
        """Retourne les tickers disponibles avec filtrage minimal."""
        if interval not in self.available_tickers:
            return []
            
        # Version simplifi√©e: retourner tous les tickers
        return list(self.available_tickers[interval].keys())
        
class OptimizedSimilarityAnalyzer:
    """Analyseur de similarit√© avec algorithmes optimis√©s pour Windows."""
    
    def __init__(self, data_manager, use_gpu=Config.USE_GPU):
        self.data_manager = data_manager
        self.device = Config.DEVICE
        
        # Configurable weights
        self.metric_weights = {
            'Close_norm': 1.0,
            'Open_norm': 0.7,
            'High_norm': 0.7, 
            'Low_norm': 0.7,
            'Volume_norm': 0.5,
            'RSI14_norm': 0.4,
            'MACD_norm': 0.4,
            'SMA50_norm': 0.5,
            'EMA20_norm': 0.5,
            'log_return_norm': 0.8
        }
        
        # Time decay parameters
        self.decay_type = 'exponential'
        self.decay_factor = 0.9
        
        # Window and tensor storage
        self.window_cache = {}
        
        # Parall√©lisation - calcule une fois au d√©marrage et r√©utilise
        # On utilise une valeur fixe qui fonctionne bien pour le GPU disponible
        self.n_jobs = Config.NUM_WORKERS
        self._analysis_cache = {}  # Cache global pour les r√©sultats d'analyse
        
        #print(f"Analyseur initialis√©: {self.device}")
    
    def _get_optimal_workers(self, cpu_intensive=True, io_intensive=False):
        """Determines optimal worker count based on workload type and system resources."""
        # Get physical and logical CPU cores
        physical_cores = psutil.cpu_count(logical=False) or 1
        logical_cores = psutil.cpu_count(logical=True) or 2
        
        # Get current CPU load
        cpu_load = psutil.cpu_percent(interval=0.1) / 100.0
        
        # Get available memory
        memory = psutil.virtual_memory()
        available_memory_gb = memory.available / (1024 * 1024 * 1024)
        
        # Determine worker count based on task type
        if cpu_intensive:
            # For CPU intensive tasks, use fewer workers and consider current load
            available_cores = max(1, int(physical_cores * (1 - cpu_load * 0.5)))
            
            # Limit based on memory (~2GB per worker)
            memory_limited_workers = max(1, int(available_memory_gb / 2))
            
            workers = min(available_cores, memory_limited_workers)
        elif io_intensive:
            # For IO intensive tasks, use more workers
            workers = max(2, logical_cores)
        else:
            # For mixed tasks
            workers = max(1, physical_cores)
        
        # Never use less than 1 or more than logical_cores
        return max(1, min(workers, logical_cores))

    def find_similar_patterns_simple(self, ticker, interval, n_similar=None):
        """Version simplifi√©e sans DTW pour l'optimisation"""
        if n_similar is None:
            n_similar = Config.N_SIMILAR
            
        try:
            # Obtenir le tenseur de r√©f√©rence
            reference_tensor = self.data_manager.create_tensor_window(
                ticker, interval, window_size=Config.get_window_size(interval)
            )
            
            if reference_tensor is None or reference_tensor.shape[0] == 0:
                return None
                
            # Appliquer les poids
            weighted_ref = self._apply_weights(reference_tensor[-1:])
            
            # Obtenir des tickers (limit√© pour l'optimisation)
            all_tickers = self.data_manager.get_available_tickers(interval)
            
            # Limiter pour l'optimisation (plus rapide)
            max_tickers_for_optimization = 300
            if len(all_tickers) > max_tickers_for_optimization:
                import random
                all_tickers = random.sample(all_tickers, max_tickers_for_optimization)
            
            # Assurer que le ticker principal est inclus
            if ticker not in all_tickers:
                all_tickers.append(ticker)
            
            # Traiter par petits lots
            all_similarities = []
            batch_size = 15  # Petits lots pour l'optimisation
            
            for i in range(0, len(all_tickers), batch_size):
                batch_tickers = all_tickers[i:i+batch_size]
                
                # Traiter le lot
                batch_similarities = self._process_ticker_batch_simple(batch_tickers, weighted_ref, interval)
                all_similarities.extend(batch_similarities)
            
            # Trier par similarit√© d√©croissante
            all_similarities.sort(key=lambda x: x['similarity'], reverse=True)
            
            # Garder les n_similar plus similaires
            top_similarities = all_similarities[:n_similar]
            
            return {
                'ticker': ticker,
                'interval': interval,
                'similar_patterns': top_similarities,
                'tickers_analyzed': len(all_tickers)
            }
            
        except Exception as e:
            logger.error(f"Error in find_similar_patterns_simple for {ticker}: {str(e)}")
            return None

    def _process_ticker_batch_simple(self, batch_tickers, weighted_ref, interval):
        """Version simplifi√©e du traitement par lot"""
        ticker_similarities = []
        
        try:
            for compare_ticker in batch_tickers:
                try:
                    compare_tensor = self.data_manager.create_tensor_window(
                        compare_ticker, interval, window_size=Config.get_window_size(interval)
                    )
                    
                    if compare_tensor is not None and compare_tensor.shape[0] > 0:
                        weighted_compare = self._apply_weights(compare_tensor)
                        
                        # Calculer similarit√© pour quelques fen√™tres seulement
                        max_windows = min(5, weighted_compare.shape[0])  # Tr√®s limit√© pour l'optimisation
                        
                        for window_idx in range(max_windows):
                            similarity = self.calculate_similarity(
                                weighted_ref[0], weighted_compare[window_idx]
                            )
                            if similarity > 0.1:  # Filtrer les tr√®s faibles similarit√©s
                                ticker_similarities.append({
                                    'ticker': compare_ticker,
                                    'window_idx': window_idx,
                                    'similarity': similarity
                                })
                except Exception as e:
                    continue  # Ignorer les erreurs de tickers individuels
            
            return ticker_similarities
            
        except Exception as e:
            logger.error(f"Error processing batch: {str(e)}")
            return []
    
    # S'assurer que calculate_expected_return existe et fonctionne
    def calculate_expected_return(self, similarity_results, forecast_horizon=None):
        """Calculates expected return with adaptive outlier management."""
        if forecast_horizon is None:
            forecast_horizon = Config.FORECAST_HORIZON
        if similarity_results is None or 'similar_patterns' not in similarity_results:
            return None
        
        # Extract information
        ticker = similarity_results['ticker']
        interval = similarity_results['interval']
        patterns = similarity_results['similar_patterns']
        
        # Accumulators
        weighted_returns = []
        
        # Analyser chaque pattern similaire (limit√© pour l'optimisation)
        max_patterns = min(len(patterns), 10)  # Limiter pour l'optimisation
        
        for pattern in patterns[:max_patterns]:
            pattern_ticker = pattern['ticker']
            window_idx = pattern['window_idx']
            similarity = pattern['similarity']
            
            try:
                # Charger les donn√©es futures
                pattern_data = self.data_manager.load_ticker_chunk(
                    pattern_ticker, interval, 
                    start_idx=window_idx, 
                    chunk_size=forecast_horizon+5  # Un peu plus pour la s√©curit√©
                )
                
                if pattern_data is None or len(pattern_data) < 2:
                    continue
                
                # Calculer le rendement futur
                start_price = pattern_data['Close'].iloc[0]
                if start_price <= 0:
                    continue
                
                # Dernier prix disponible
                last_idx = min(forecast_horizon, len(pattern_data)-1)
                end_price = pattern_data['Close'].iloc[last_idx]
                
                # Rendement relatif
                pattern_return = (end_price / start_price) - 1
                
                # Pond√©rer par similarit√© au carr√©
                weight = similarity ** 2
                weighted_returns.append((pattern_return, weight))
                
            except Exception as e:
                continue
        
        # V√©rifier qu'il y a suffisamment de donn√©es
        if len(weighted_returns) < 2:
            return None
        
        # Calcul simple sans d√©tection d'outliers (pour l'optimisation)
        total_weight = sum(weight for _, weight in weighted_returns)
        if total_weight <= 0:
            return None
            
        expected_return = sum(ret * weight for ret, weight in weighted_returns) / total_weight
        
        return expected_return
    
    def _process_ticker_batch(self, batch_tickers, weighted_ref, interval):
        """Traite un lot de tickers pour trouver des similarit√©s."""
        try:
            ticker_similarities = []
            
            # Cr√©er tenseurs pour tout le lot
            batch_tensors = []
            for compare_ticker in batch_tickers:
                compare_tensor = self.data_manager.create_tensor_window(
                    compare_ticker, interval, window_size=Config.get_window_size(interval)
                )
                if compare_tensor is not None and compare_tensor.shape[0] > 0:
                    batch_tensors.append((compare_ticker, self._apply_weights(compare_tensor)))
            
            # RETOUR √Ä L'ALGORITHME ORIGINAL - Calculer similarit√©s pour chaque fen√™tre
            for compare_ticker, weighted_compare in batch_tensors:
                for window_idx in range(weighted_compare.shape[0]):
                    similarity = self.calculate_similarity(
                        weighted_ref[0], weighted_compare[window_idx]
                    )
                    ticker_similarities.append({
                        'ticker': compare_ticker,
                        'window_idx': window_idx,
                        'similarity': similarity
                    })
            
            return ticker_similarities
        except Exception as e:
            logger.error(f"Error processing batch: {str(e)}")
            return []
        
    def find_similar_patterns(self, ticker, interval, n_similar=Config.N_SIMILAR, 
                             convergence_threshold=0.001, 
                             initial_tickers=500, 
                             increment_size=50, 
                             max_tickers=1200):
        """Finds similar patterns using ALL available tickers."""
        # Obtenir le tenseur de r√©f√©rence
        reference_tensor = self.data_manager.create_tensor_window(
            ticker, interval, window_size=Config.get_window_size(interval)
        )
        
        if reference_tensor is None or reference_tensor.shape[0] == 0:
            logger.warning(f"No valid reference tensor for {ticker}")
            return None
            
        # Appliquer les poids
        weighted_ref = self._apply_weights(reference_tensor[-1:])
        
        # Obtenir TOUS les tickers disponibles
        all_tickers = self.data_manager.get_available_tickers(interval)
        
        # Assurer que le ticker principal est inclus
        if ticker not in all_tickers:
            all_tickers.append(ticker)
        
        # Trier pour rendre d√©terministe
        all_tickers.sort()
        
        # Traiter TOUS les tickers (plus de convergence incr√©mentale)
        all_similarities = []
        
        # Traitement par lot pour m√©moire GPU
        batch_size = Config.get_batch_size()
        ticker_batches = [all_tickers[i:i+batch_size] for i in range(0, len(all_tickers), batch_size)]
        
        logger.info(f"Processing ALL {len(all_tickers)} tickers for {ticker} on {interval}")
        start_time = time.time()
        
        for batch_idx, batch_tickers in enumerate(ticker_batches):
            batch_start = time.time()
            
            # Traiter le lot avec la fonction standard
            batch_similarities = self._process_ticker_batch(batch_tickers, weighted_ref, interval)
            all_similarities.extend(batch_similarities)
            
            # Log pour suivi
            logger.info(f"Processed batch {batch_idx+1}/{len(ticker_batches)} with {len(batch_tickers)} tickers "
                       f"in {time.time() - batch_start:.2f}s")
            
            # Lib√©rer m√©moire GPU
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
        
        # Trier par similarit√© d√©croissante
        all_similarities.sort(key=lambda x: x['similarity'], reverse=True)
        
        # Garder les n_similar plus similaires
        top_similarities = all_similarities[:n_similar]
        
        # R√©sultat organis√©
        result = {
            'ticker': ticker,
            'interval': interval,
            'similar_patterns': top_similarities,
            'tickers_analyzed': len(all_tickers),
            'execution_time_s': time.time() - start_time
        }
        
        logger.info(f"Analyzed ALL {len(all_tickers)} tickers for {ticker} in {result['execution_time_s']:.2f}s")
        
        return result
    
    def _apply_weights(self, tensor):
        """Applique les poids et d√©croissance temporelle."""
        if tensor is None:
            return None
            
        # Appliquer poids des m√©triques
        weighted_tensor = tensor.clone()
        for i, name in enumerate(self.metric_weights.keys()):
            if i < tensor.shape[1]:  # V√©rifier que l'index est valide
                weighted_tensor[:, i, :] *= self.metric_weights.get(name, 1.0)
        
        # Appliquer d√©croissance temporelle
        time_weights = self._calculate_time_weights(tensor.shape[2])
        time_tensor = torch.tensor(time_weights, device=self.device)
        
        # Multiplication (broadcast)
        for i in range(tensor.shape[1]):
            weighted_tensor[:, i, :] *= time_tensor
        
        return weighted_tensor
    
    def _calculate_time_weights(self, length):
        """Calcule les poids de d√©croissance temporelle."""
        # Cr√©er indices invers√©s
        indices = np.arange(length)[::-1]
        norm_indices = indices / max(1, length - 1)
        
        # Diff√©rents types de d√©croissance
        if self.decay_type == 'linear':
            weights = 1.0 - (1.0 - self.decay_factor) * norm_indices
        elif self.decay_type == 'exponential':
            weights = self.decay_factor ** norm_indices
        elif self.decay_type == 'quadratic':
            weights = 1.0 - (1.0 - self.decay_factor) * norm_indices**2
        else:  # Pas de d√©croissance
            weights = np.ones(length)
        
        return weights
    
    def calculate_similarity(self, tensor1, tensor2):
        if tensor1 is None or tensor2 is None:
            return 0.0
            
        # Ensure tensors are on the same device
        if tensor1.device != tensor2.device:
            tensor2 = tensor2.to(tensor1.device)
        
        # REMPLACER TOUTE LA SUITE PAR:
        
        # Aplatir et traiter les NaN en une seule fois pour √©viter des transferts multiples
        tensor1_flat = tensor1.reshape(-1)
        tensor2_flat = tensor2.reshape(-1)
        
        # Cr√©er les masques sur GPU
        valid_mask1 = ~torch.isnan(tensor1_flat)
        valid_mask2 = ~torch.isnan(tensor2_flat)
        valid_mask = valid_mask1 & valid_mask2
        
        # Si aucune valeur valide, retourner similarit√© faible
        if valid_mask.sum() == 0:
            return 0.1
        
        # Calculer le ratio de couverture
        coverage = valid_mask.float().mean().item()
        
        # Au lieu d'extraire, utiliser where pour remplacer les NaN par z√©ros
        # Cela √©vite des transferts et reste sur GPU
        t1_valid = torch.where(valid_mask, tensor1_flat, torch.tensor(0.0, device=tensor1.device))
        t2_valid = torch.where(valid_mask, tensor2_flat, torch.tensor(0.0, device=tensor2.device))
        
        # Normalisation robuste sur GPU
        epsilon = 1e-8
        norm1 = torch.norm(t1_valid) + epsilon
        norm2 = torch.norm(t2_valid) + epsilon
        
        # Produit scalaire
        dot_product = torch.dot(t1_valid, t2_valid)
        similarity = dot_product / (norm1 * norm2)
        
        # P√©nalit√© de couverture am√©lior√©e
        penalty = 1.0 / (1.0 + np.exp(-10 * (coverage - 0.5)))
        penalty = max(0.2, penalty)
        
        # Similarit√© ajust√©e
        adjusted_similarity = similarity.item() * penalty
        
        # Borner entre 0 et 1
        return max(0.0, min(1.0, adjusted_similarity))
        

    
        
    def calculate_expected_return(self, similarity_results, forecast_horizon=Config.FORECAST_HORIZON):
        """Calculates expected return with adaptive outlier management."""
        if similarity_results is None or 'similar_patterns' not in similarity_results:
            return None
        
        # Extract information
        ticker = similarity_results['ticker']
        interval = similarity_results['interval']
        patterns = similarity_results['similar_patterns']
        
        # Accumulators
        all_returns = []
        weighted_returns = []
        
        # Analyze each similar pattern
        for pattern in patterns:
            pattern_ticker = pattern['ticker']
            window_idx = pattern['window_idx']
            similarity = pattern['similarity']
            
            try:
                # Load future data
                pattern_data = self.data_manager.load_ticker_chunk(
                    pattern_ticker, interval, 
                    start_idx=window_idx, 
                    chunk_size=forecast_horizon+1
                )
                
                if pattern_data is None or len(pattern_data) < 2:
                    continue
                
                # Calculate future return
                start_price = pattern_data['Close'].iloc[0]
                if start_price <= 0:
                    continue
                
                # Last available price
                last_idx = min(forecast_horizon, len(pattern_data)-1)
                end_price = pattern_data['Close'].iloc[last_idx]
                
                # Relative return
                pattern_return = (end_price / start_price) - 1
                
                # Store return and similarity
                all_returns.append(pattern_return)
                
                # Weight by squared similarity
                weight = similarity ** 2
                weighted_returns.append((pattern_return, weight))
                
            except Exception as e:
                logger.warning(f"Error calculating return for {pattern_ticker} at index {window_idx}: {str(e)}")
                continue
        
        # Check for sufficient data
        if not weighted_returns:
            return None
        
        # Adaptive outlier detection instead of fixed trimming
        # Use interquartile range (IQR) method
        all_returns = np.array([r for r, _ in weighted_returns])
        
        # Calculate IQR
        q1 = np.percentile(all_returns, 25)
        q3 = np.percentile(all_returns, 75)
        iqr = q3 - q1
        
        # Define outlier thresholds - 1.5*IQR is standard practice
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Filter out outliers
        filtered_returns = []
        for ret, weight in weighted_returns:
            if lower_bound <= ret <= upper_bound:
                filtered_returns.append((ret, weight))
        
        # If filtering removed too many points, use original data with smaller bounds
        if len(filtered_returns) < max(3, len(weighted_returns) * 0.3):
            # Use more conservative bounds
            lower_bound = q1 - 2.5 * iqr
            upper_bound = q3 + 2.5 * iqr
            
            filtered_returns = []
            for ret, weight in weighted_returns:
                if lower_bound <= ret <= upper_bound:
                    filtered_returns.append((ret, weight))
        
        # If still too few points, use all data
        if len(filtered_returns) < 3:
            filtered_returns = weighted_returns
        
        # Calculate weighted average
        total_weight = sum(weight for _, weight in filtered_returns)
        if total_weight <= 0:
            return None
            
        expected_return = sum(ret * weight for ret, weight in filtered_returns) / total_weight
        
        return expected_return
    
    def flasc_cluster_trajectories(self, similarity_results, ticker_data, n_forecast=None, analysis_date=None, min_cluster_size=5, min_flare_size=3, eps=0.05, historical_bars=None, use_full_history=True):
        """FLASC clustering with complete historical data matching."""
        if n_forecast is None:
            n_forecast = Config.FORECAST_HORIZON
        if historical_bars is None:
            historical_bars = Config.HISTORICAL_BARS
        if similarity_results is None or 'similar_patterns' not in similarity_results:
            return None
        
        # Extract key information
        ticker = similarity_results['ticker']
        analysis_date = analysis_date or similarity_results.get('analysis_date')
        if analysis_date is None:
            return None
        patterns = similarity_results['similar_patterns']
        interval = similarity_results['interval']
        
        # Find the analysis point index
        if isinstance(ticker_data.index, pd.DatetimeIndex):
            try:
                idx = ticker_data.index.get_loc(analysis_date)
            except KeyError:
                idx = ticker_data.index.get_indexer([analysis_date], method='nearest')[0]
        else:
            idx = ticker_data.reset_index().index[ticker_data.index == analysis_date][0]
        
        # Reference price for normalization
        reference_price = ticker_data['Close'].iloc[idx]
        
        # Use ALL available history - the full length of ticker_data up to analysis point
        n_history = idx + 1  # Inclut tout l'historique jusqu'au point d'analyse
        
        # Prepare normalized trajectories
        trajectories = []
        pattern_info = []
        
        # Cr√©er la s√©quence historique compl√®te du ticker principal (pour r√©f√©rence)
        main_ticker_sequence = ticker_data['Close'].iloc[:idx+1].values / reference_price
        
        # Pour chaque pattern, essayer d'obtenir le maximum d'historique correspondant
        for pattern in patterns:
            try:
                pattern_ticker = pattern['ticker']
                window_idx = pattern['window_idx']
                similarity = pattern['similarity']
                
                # Load FULL ticker data
                pattern_data = self.data_manager._load_ticker_data(pattern_ticker, interval)
                if pattern_data is None or window_idx >= len(pattern_data):
                    continue
                
                # Get future data availability
                future_available = min(n_forecast, len(pattern_data) - window_idx - 1)
                if future_available <= 0:
                    continue
                    
                # Normalize price at analysis point (window_idx)
                pattern_ref_price = pattern_data['Close'].iloc[window_idx]
                if pattern_ref_price <= 0:
                    continue
                    
                # Create full trajectory with ALL available history + future
                # Total length: maximum historical length (n_history) + forecast horizon
                full_trajectory = np.full(n_history + n_forecast, np.nan)
                
                # Prepare ALL historical data (use as much as available)
                hist_start = max(0, window_idx - (n_history - 1))  # Go back as far as possible
                hist_data = pattern_data['Close'].iloc[hist_start:window_idx+1].values / pattern_ref_price
                
                # Position in full trajectory - align with end point (window_idx corresponds to n_history-1)
                hist_offset = n_history - len(hist_data)
                if hist_offset >= 0:
                    # Not enough history - place what we have
                    full_trajectory[hist_offset:n_history] = hist_data
                else:
                    # More history than needed - take the most recent portion
                    full_trajectory[:n_history] = hist_data[-n_history:]
                    
                # Add future data
                future_data = pattern_data['Close'].iloc[window_idx+1:window_idx+1+future_available].values / pattern_ref_price
                full_trajectory[n_history:n_history+len(future_data)] = future_data
                
                # Verify trajectory contains sufficient future data
                min_future_points = min(5, n_forecast // 2)
                if np.sum(~np.isnan(full_trajectory[n_history:])) < min_future_points:
                    continue
                    
                # Add valid trajectory
                trajectories.append(full_trajectory)
                pattern_info.append({
                    'ticker': pattern_ticker, 
                    'window_idx': window_idx,
                    'similarity': similarity,
                    'is_main_ticker': pattern_ticker == ticker
                })
                    
            except Exception as e:
                logger.error(f"Error processing pattern: {str(e)}")
                continue
        
        # The rest of the function remains similar...
        if len(trajectories) < min_cluster_size:
            return None
        
        # Convert to array
        trajectory_array = np.array(trajectories)
        
        # Trajectory distance function that handles NaNs
        def trajectory_distance(t1, t2):
            valid_indices = ~np.isnan(t1) & ~np.isnan(t2)
            if not np.any(valid_indices):
                return float('inf')
            return np.sqrt(np.mean((t1[valid_indices] - t2[valid_indices])**2))
        
        # Build distance matrix
        n_trajectories = len(trajectories)
        dist_matrix = np.zeros((n_trajectories, n_trajectories))
        
        for i in range(n_trajectories):
            for j in range(i+1, n_trajectories):
                dist = trajectory_distance(trajectory_array[i], trajectory_array[j])
                dist_matrix[i, j] = dist
                dist_matrix[j, i] = dist
        
        # Calculate adaptive radius for each point based on local density
        from sklearn.neighbors import NearestNeighbors
        k = min(15, n_trajectories // 4)
        k = max(3, k)
        
        nbrs = NearestNeighbors(n_neighbors=k+1, metric='precomputed').fit(dist_matrix)
        distances, indices = nbrs.kneighbors(dist_matrix)
        
        adaptive_radius = np.zeros(n_trajectories)
        for i in range(n_trajectories):
            adaptive_radius[i] = np.mean(distances[i, 1:]) * 1.2
        
        # Create graph with adaptive radii
        import networkx as nx
        G = nx.Graph()
        
        # Add nodes
        for i in range(n_trajectories):
            G.add_node(i, trajectory=trajectories[i], info=pattern_info[i])
        
        # Add edges with weights based on distance
        for i in range(n_trajectories):
            for j in range(i+1, n_trajectories):
                dist = dist_matrix[i, j]
                if dist <= adaptive_radius[i] or dist <= adaptive_radius[j]:
                    weight = 1.0 / (dist + 1e-6)
                    G.add_edge(i, j, weight=weight, distance=dist)
        
        # Find communities
        try:
            from networkx.algorithms import community
            communities = community.louvain_communities(G)
        except:
            from networkx.algorithms.community import greedy_modularity_communities
            communities = list(greedy_modularity_communities(G))
        
        # Identify flares within clusters
        flares = []
        
        for comm_idx, comm in enumerate(communities):
            if len(comm) < min_cluster_size:
                continue
            
            # Create subgraph for this community
            subgraph = G.subgraph(comm)
            
            # Identify potential flare endpoints (peripheral nodes)
            degree_dict = dict(subgraph.degree())
            peripheral_nodes = [n for n, d in degree_dict.items() if d <= 2]
            
            if len(peripheral_nodes) >= 2:
                try:
                    # Find betweenness centrality
                    centrality = nx.betweenness_centrality(subgraph)
                    
                    # Find potential branch points
                    branch_candidates = sorted(
                        [(n, c) for n, c in centrality.items()], 
                        key=lambda x: x[1], 
                        reverse=True
                    )[:min(5, len(comm) // 3)]
                    
                    # Check paths between peripheral nodes
                    for start_idx, start in enumerate(peripheral_nodes):
                        for end in peripheral_nodes[start_idx+1:]:
                            try:
                                all_paths = list(nx.all_shortest_paths(subgraph, start, end))
                                
                                if not all_paths:
                                    continue
                                    
                                for path in all_paths:
                                    if len(path) < min_flare_size*2:
                                        continue
                                    
                                    for branch, _ in branch_candidates:
                                        if branch in path and branch != start and branch != end:
                                            branch_idx = path.index(branch)
                                            flare1 = path[:branch_idx+1]
                                            flare2 = path[branch_idx:]
                                            
                                            if len(flare1) >= min_flare_size:
                                                flares.append({
                                                    'nodes': flare1,
                                                    'branch_point': branch,
                                                    'community': comm_idx
                                                })
                                            
                                            if len(flare2) >= min_flare_size:
                                                flares.append({
                                                    'nodes': flare2,
                                                    'branch_point': branch,
                                                    'community': comm_idx
                                                })
                            except nx.NetworkXNoPath:
                                continue
                except Exception as e:
                    logger.error(f"Error processing flares for community {comm_idx}: {str(e)}")
                    continue
        
        # Organize results
        clusters = {}
        
        # Add main communities
        for i, comm in enumerate(communities):
            if len(comm) < min_cluster_size:
                continue
                
            clusters[i] = {
                'trajectories': [trajectory_array[n] for n in comm],
                'pattern_info': [pattern_info[n] for n in comm],
                'nodes': list(comm),
                'is_flare': False,
                'community_id': i,
                'size': len(comm)
            }
        
        # Add flares as separate clusters if significant
        flare_counter = 0
        for flare in flares:
            if len(flare['nodes']) >= min_flare_size:
                flare_id = len(clusters) + flare_counter
                flare_counter += 1
                
                clusters[flare_id] = {
                    'trajectories': [trajectory_array[n] for n in flare['nodes']],
                    'pattern_info': [pattern_info[n] for n in flare['nodes']],
                    'nodes': flare['nodes'],
                    'is_flare': True,
                    'branch_point': flare['branch_point'],
                    'parent_community': flare['community'],
                    'size': len(flare['nodes'])
                }
        
        # Compute statistics for each cluster
        for cluster_id, cluster in clusters.items():
            # Calculate mean similarity
            cluster['mean_similarity'] = np.mean([p['similarity'] for p in cluster['pattern_info']])
            
            # Calculate mean trajectory
            trajectories_array = np.array(cluster['trajectories'])
            mean_trajectory = np.nanmean(trajectories_array, axis=0)
            std_trajectory = np.nanstd(trajectories_array, axis=0)
            
            cluster['mean_trajectory'] = mean_trajectory
            cluster['std_trajectory'] = std_trajectory
            
            # Calculate terminal values
            terminal_values = []
            for traj in trajectories_array:
                last_valid_idx = np.where(~np.isnan(traj))[0]
                if len(last_valid_idx) > 0:
                    terminal_values.append(traj[last_valid_idx[-1]])
            
            if terminal_values:
                cluster['mean_terminal'] = np.mean(terminal_values)
                cluster['std_terminal'] = np.std(terminal_values)
                cluster['terminal_change'] = ((cluster['mean_terminal'] - 1.0) * 100)
            else:
                cluster['mean_terminal'] = np.nan
                cluster['std_terminal'] = np.nan
                cluster['terminal_change'] = np.nan
        
        # Sort clusters by size
        sorted_clusters = {k: v for k, v in sorted(
            clusters.items(), 
            key=lambda item: (0 if item[1]['is_flare'] else 1, -item[1]['size'])
        )}
        
        # Create color mapping
        color_mapping = {}
        import matplotlib.pyplot as plt
        main_cmap = plt.cm.get_cmap('tab10', max(len(communities), 1))
        flare_cmap = plt.cm.get_cmap('Set2', max(flare_counter, 1))
        
        main_idx = 0
        flare_idx = 0
        
        for cluster_id, cluster in sorted_clusters.items():
            is_flare = cluster.get('is_flare', False)
            
            if is_flare:
                color_mapping[cluster_id] = flare_cmap(flare_idx % max(flare_counter, 1))
                flare_idx += 1
            else:
                color_mapping[cluster_id] = main_cmap(main_idx % max(len(communities), 1))
                main_idx += 1
        
        result = {
            'ticker': ticker,
            'analysis_date': analysis_date,
            'clusters': sorted_clusters,
            'reference_price': reference_price,
            'n_trajectories': n_trajectories,
            'n_communities': len(communities),
            'n_clusters': len(clusters),
            'n_flares': flare_counter,
            'color_mapping': color_mapping,
            'n_history': n_history,  # This is now the FULL historical length
            'use_full_history': True,
            'main_ticker_sequence': main_ticker_sequence  # Store the main ticker history for verification
        }
        
        return result

        
    def _dtw_distance(self, series1, series2, weights=None):
        """Optimized DTW distance calculation with adaptive radius and improved NaN handling."""
        if series1 is None or series2 is None:
            return float('inf')
        
        # Handle empty or single-point series
        if series1.size == 0 or series2.size == 0:
            return float('inf')
        
        # Copy data to avoid modifying originals
        series1_copy = series1.copy()
        series2_copy = series2.copy()
        
        # Clean data with robust interpolation for NaNs
        for i in range(series1_copy.shape[0]):
            mask = np.isnan(series1_copy[i])
            if np.any(mask) and not np.all(mask):
                valid_indices = np.where(~mask)[0]
                nan_indices = np.where(mask)[0]
                
                # Only interpolate if we have enough valid points
                if len(valid_indices) >= 2:
                    # Use Pchip interpolation that preserves monotonicity (better for financial data)
                    try:
                        from scipy.interpolate import PchipInterpolator
                        interpolator = PchipInterpolator(valid_indices, series1_copy[i, valid_indices])
                        series1_copy[i, nan_indices] = interpolator(nan_indices)
                    except:
                        # Fallback to simple linear interpolation
                        for nan_idx in nan_indices:
                            # Find nearest valid indices
                            left_idx = valid_indices[valid_indices < nan_idx]
                            right_idx = valid_indices[valid_indices > nan_idx]
                            
                            if len(left_idx) > 0 and len(right_idx) > 0:
                                # Interpolate
                                left = left_idx[-1]
                                right = right_idx[0]
                                ratio = (nan_idx - left) / (right - left)
                                series1_copy[i, nan_idx] = series1_copy[i, left] + ratio * (series1_copy[i, right] - series1_copy[i, left])
                            elif len(left_idx) > 0:
                                # Forward fill
                                series1_copy[i, nan_idx] = series1_copy[i, left_idx[-1]]
                            elif len(right_idx) > 0:
                                # Backward fill
                                series1_copy[i, nan_idx] = series1_copy[i, right_idx[0]]
                
                # Any remaining NaNs set to 0
                series1_copy[i, np.isnan(series1_copy[i])] = 0
        
        # Same process for series2
        for i in range(series2_copy.shape[0]):
            mask = np.isnan(series2_copy[i])
            if np.any(mask) and not np.all(mask):
                valid_indices = np.where(~mask)[0]
                nan_indices = np.where(mask)[0]
                
                if len(valid_indices) >= 2:
                    try:
                        from scipy.interpolate import PchipInterpolator
                        interpolator = PchipInterpolator(valid_indices, series2_copy[i, valid_indices])
                        series2_copy[i, nan_indices] = interpolator(nan_indices)
                    except:
                        for nan_idx in nan_indices:
                            left_idx = valid_indices[valid_indices < nan_idx]
                            right_idx = valid_indices[valid_indices > nan_idx]
                            
                            if len(left_idx) > 0 and len(right_idx) > 0:
                                left = left_idx[-1]
                                right = right_idx[0]
                                ratio = (nan_idx - left) / (right - left)
                                series2_copy[i, nan_idx] = series2_copy[i, left] + ratio * (series2_copy[i, right] - series2_copy[i, left])
                            elif len(left_idx) > 0:
                                series2_copy[i, nan_idx] = series2_copy[i, left_idx[-1]]
                            elif len(right_idx) > 0:
                                series2_copy[i, nan_idx] = series2_copy[i, right_idx[0]]
                
                series2_copy[i, np.isnan(series2_copy[i])] = 0
        
        # Apply weights if provided
        if weights is not None:
            series1_copy = series1_copy * weights[:, np.newaxis]
            series2_copy = series2_copy * weights[:, np.newaxis]
        
        # Calculate time weights based on dynamics
        time_len = max(series1_copy.shape[1], series2_copy.shape[1])
        
        # Detect trend changes in both series for adaptive weighting
        trends1 = np.zeros(time_len)
        trends2 = np.zeros(time_len)
        
        for i in range(series1_copy.shape[0]):
            if series1_copy.shape[1] > 1:
                # Detect sign changes (potential trend reversals)
                diffs = np.diff(series1_copy[i])
                trends1[:len(diffs)] += np.abs(np.sign(diffs[1:]) - np.sign(diffs[:-1]))
        
        for i in range(series2_copy.shape[0]):
            if series2_copy.shape[1] > 1:
                diffs = np.diff(series2_copy[i])
                trends2[:len(diffs)] += np.abs(np.sign(diffs[1:]) - np.sign(diffs[:-1]))
        
        # Combine trend information from both series
        combined_trends = np.zeros(time_len)
        combined_trends[:len(trends1)] += trends1
        combined_trends[:len(trends2)] += trends2
        
        # Normalize trend changes
        if np.max(combined_trends) > 0:
            combined_trends = combined_trends / np.max(combined_trends)
        
        # Base weights - recent periods more important (end of series)
        time_weights = np.linspace(0.7, 1.0, time_len)
        
        # Add more weight to periods with trend changes
        time_weights = time_weights + 0.3 * combined_trends
        
        # Re-normalize
        time_weights = time_weights / np.mean(time_weights)
        
        # Apply time weights
        series1_copy = series1_copy * time_weights[np.newaxis, :time_len][:, :series1_copy.shape[1]]
        series2_copy = series2_copy * time_weights[np.newaxis, :time_len][:, :series2_copy.shape[1]]
        
        # Adaptive radius calculation based on series length and complexity
        max_series_length = max(series1_copy.shape[1], series2_copy.shape[1])
        
        # Dynamic radius strategy based on empirical testing
        if max_series_length <= 30:
            radius = max(2, int(0.3 * max_series_length))
        elif max_series_length <= 100:
            radius = max(3, int(0.2 * max_series_length))
        else:
            radius = max(5, int(0.15 * max_series_length))
        
        # Cap radius for very long series to balance accuracy vs speed
        if max_series_length > 200:
            radius = min(radius, 30)
        
        # Calculate DTW distance
        from fastdtw import fastdtw
        from scipy.spatial.distance import euclidean
        
        # Log memory usage before DTW
        if torch.cuda.is_available():
            before_mem = torch.cuda.memory_allocated() / (1024 * 1024)
            logger.debug(f"GPU memory before DTW: {before_mem:.2f} MB")
        
        # Calculate similarity with warping constraint
        distance, path = fastdtw(series1_copy, series2_copy, dist=euclidean, radius=radius)
        
        # Log memory after DTW
        if torch.cuda.is_available():
            after_mem = torch.cuda.memory_allocated() / (1024 * 1024)
            logger.debug(f"GPU memory after DTW: {after_mem:.2f} MB, diff: {after_mem-before_mem:.2f} MB")
        
        # Normalize by path length for fair comparison between series of different lengths
        normalized_distance = distance / len(path)
        
        # Apply similarity function (convert distance to similarity score)
        similarity = np.exp(-normalized_distance / 20.0)
        
        # Apply volatility adjustment - penalize matching high volatility to low volatility
        vol1 = np.nanstd(series1_copy)
        vol2 = np.nanstd(series2_copy)
        vol_ratio = max(vol1, vol2) / (min(vol1, vol2) + 1e-6)
        
        # Penalize only severe volatility mismatches (>2x difference)
        if vol_ratio > 2.0:
            vol_penalty = np.exp(-(vol_ratio - 2.0) / 3.0)
            similarity *= vol_penalty
        
        return float(normalized_distance)

    
    def find_optimal_ticker_count(self, interval, metric='expected_return'):
        """
        Toujours retourner le nombre total de tickers disponibles.
        Cette fonction est maintenue pour compatibilit√©.
        """
        # Obtenir tous les tickers disponibles
        all_tickers = self.data_manager.get_available_tickers(interval)
        total_count = len(all_tickers)
        logger.info(f"Using ALL {total_count} tickers for {interval}")
        
        # Retourner simplement le nombre total
        return total_count

        
    def find_similar_patterns(self, ticker, interval, n_similar=None, convergence_threshold=0.001, 
                            initial_tickers=500, increment_size=50, max_tickers=1200):
        """Finds similar patterns using ALL available tickers."""
        if n_similar is None:
            n_similar = Config.N_SIMILAR
            
        # Obtenir le tenseur de r√©f√©rence
        reference_tensor = self.data_manager.create_tensor_window(
            ticker, interval, window_size=Config.get_window_size(interval)
        )
        
        if reference_tensor is None or reference_tensor.shape[0] == 0:
            logger.warning(f"No valid reference tensor for {ticker}")
            return None
            
        # Appliquer les poids
        weighted_ref = self._apply_weights(reference_tensor[-1:])
        
        # Obtenir TOUS les tickers disponibles
        all_tickers = self.data_manager.get_available_tickers(interval)
        
        # Assurer que le ticker principal est inclus
        if ticker not in all_tickers:
            all_tickers.append(ticker)
        
        # Trier pour rendre d√©terministe
        all_tickers.sort()
        
        # Traiter TOUS les tickers (plus de convergence incr√©mentale)
        all_similarities = []
        
        # Traitement par lot pour m√©moire GPU
        batch_size = min(50, Config.get_batch_size())  # R√©duire la taille des lots
        ticker_batches = [all_tickers[i:i+batch_size] for i in range(0, len(all_tickers), batch_size)]
        
        logger.debug(f"Processing ALL {len(all_tickers)} tickers for {ticker} on {interval}")
        start_time = time.time()
        
        for batch_idx, batch_tickers in enumerate(ticker_batches):
            batch_start = time.time()
            
            # Traiter le lot avec la fonction standard
            batch_similarities = self._process_ticker_batch(batch_tickers, weighted_ref, interval)
            all_similarities.extend(batch_similarities)
            
            # Log pour suivi (seulement en debug)
            if batch_idx % 5 == 0:
                logger.debug(f"Processed batch {batch_idx+1}/{len(ticker_batches)} with {len(batch_tickers)} tickers")
            
            # Lib√©rer m√©moire GPU
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
        
        # Trier par similarit√© d√©croissante
        all_similarities.sort(key=lambda x: x['similarity'], reverse=True)
        
        # Garder les n_similar plus similaires
        top_similarities = all_similarities[:n_similar]
        
        # R√©sultat organis√©
        result = {
            'ticker': ticker,
            'interval': interval,
            'similar_patterns': top_similarities,
            'tickers_analyzed': len(all_tickers),
            'execution_time_s': time.time() - start_time
        }
        
        return result

    def plot_flasc_trajectories(self, similarity_results, ticker_data, cluster_results=None, 
                               n_last=Config.HISTORICAL_BARS, n_forecast=Config.FORECAST_HORIZON, max_patterns=200, highlight_flares=True, analysis_date=None):
        """Plot clustered price trajectories with FLASC analysis."""
        import matplotlib.pyplot as plt
        
        print("Debug: Starting trajectory plot...")
        
        if similarity_results is None or 'similar_patterns' not in similarity_results:
            print("Error: No similar patterns found for visualization")
            return
        
        # Get analysis_date from parameter or dictionary
        if analysis_date is None:
            analysis_date = similarity_results.get('analysis_date')
        
        print(f"Debug: Analysis reference date: {analysis_date}")
        
        if analysis_date is None:
            print("Error: No analysis date found for visualization")
            return
        
        # Extraire l'intervalle des r√©sultats de similarit√©
        interval = similarity_results.get('interval')
        if interval is None:
            print("Error: No interval information found in similarity results")
            return
            
        # Perform clustering if not provided
        if cluster_results is None:
            print("Debug: Performing FLASC clustering...")
            cluster_results = self.flasc_cluster_trajectories(
                similarity_results, 
                ticker_data,
                analysis_date=analysis_date,
                n_forecast=n_forecast,
                historical_bars=n_last,
                min_cluster_size=5
            )
        
        if cluster_results is None:
            print("Error: FLASC clustering failed to produce results")
            return
            
        if not cluster_results.get('clusters'):
            print("Error: No valid clusters found in clustering results")
            return
        
        print(f"Debug: Found {len(cluster_results['clusters'])} valid clusters")
        
        # Extract key information
        ticker = cluster_results['ticker']
        reference_price = cluster_results['reference_price']
        
        # Find analysis point index
        if isinstance(ticker_data.index, pd.DatetimeIndex):
            try:
                idx = ticker_data.index.get_loc(analysis_date)
            except KeyError:
                idx = ticker_data.index.get_indexer([analysis_date], method='nearest')[0]
        else:
            idx = ticker_data.reset_index().index[ticker_data.index == analysis_date][0]
        
        # Determine available data
        available_history = min(n_last, idx + 1)
        available_future = min(n_forecast, len(ticker_data) - idx - 1)
        total_display_bars = n_last + n_forecast
        analysis_x = n_last - 1
        
        print(f"Debug: Index of analysis date: {idx}, Available history: {available_history}, Available future: {available_future}")
        
        # Prepare data sections
        start_idx = max(0, idx - available_history + 1)
        end_idx = min(len(ticker_data), idx + available_future + 1)
        
        print(f"Debug: Data range - start_idx: {start_idx}, idx: {idx}, end_idx: {end_idx}")
        
        # Extract historical and future data
        history_data = ticker_data.iloc[start_idx:idx+1]
        future_data = ticker_data.iloc[idx+1:end_idx] if idx+1 < len(ticker_data) else pd.DataFrame()
        
        print(f"Debug: History data shape: {history_data.shape}, Future data shape: {future_data.shape if not future_data.empty else (0, 0)}")
        
        # Prepare x-axis values representing the entire timeline
        x_values = range(total_display_bars)
        
        # Normalize historical data
        normalized_history = history_data['Close'].values / reference_price
        normalized_future = future_data['Close'].values / reference_price if not future_data.empty else []
        
        # Place historical data correctly
        real_history = np.full(total_display_bars, np.nan)
        real_future = np.full(total_display_bars, np.nan)
        
        history_length = len(normalized_history)
        start_pos = max(0, analysis_x - history_length + 1)
        for i in range(history_length):
            pos = start_pos + i
            if 0 <= pos < total_display_bars:
                real_history[pos] = normalized_history[i]
        
        # Place future data
        for i in range(len(normalized_future)):
            pos = analysis_x + 1 + i
            if pos < total_display_bars:
                real_future[pos] = normalized_future[i]
        
        # Create figure
        plt.figure(figsize=(16, 9), dpi=100)
        
        # Get colormaps
        color_mapping = cluster_results.get('color_mapping', {})
        
        # SOLUTION SIMPLE: Pour chaque trajectoire, √©tirer les donn√©es pour correspondre √† la largeur affich√©e 
        for cluster_id, cluster in cluster_results['clusters'].items():
            color = color_mapping.get(cluster_id, 'blue')
            is_flare = cluster.get('is_flare', False)
            
            for traj, info in zip(cluster['trajectories'], cluster['pattern_info']):
                # Extraire les parties de la trajectoire
                # Point d'analyse = index n_forecast dans le tableau traj
                n_history_in_cluster = cluster_results.get('n_history', n_forecast)
                
                # S√©paration histoire/futur
                history_part = traj[:n_history_in_cluster]
                future_part = traj[n_history_in_cluster:]
                
                # √âliminer les NaN aux extr√©mit√©s
                valid_hist_indices = np.where(~np.isnan(history_part))[0]
                if len(valid_hist_indices) > 0:
                    start_idx_traj = valid_hist_indices[0]
                    end_idx_traj = valid_hist_indices[-1] + 1
                    valid_history = history_part[start_idx_traj:end_idx_traj]
                else:
                    valid_history = []
                
                valid_future = future_part[~np.isnan(future_part)]
                
                # Si on a suffisamment de donn√©es historiques valides, afficher
                if len(valid_history) >= 2:
                    # Cr√©er une version interpol√©e qui couvre toute la largeur visible
                    # de l'axe des x, de 0 √† analysis_x
                    hist_positions = np.linspace(0, analysis_x, len(valid_history))
                    plt.plot(hist_positions, valid_history, '-', color=color, 
                            linewidth=0.6, alpha=0.3)
                
                # Tracer le futur
                if len(valid_future) >= 2:
                    future_positions = np.linspace(analysis_x+1, 
                                                 min(analysis_x+n_forecast, total_display_bars-1), 
                                                 len(valid_future))
                    plt.plot(future_positions, valid_future, '-', color=color, 
                            linewidth=0.6, alpha=0.3)
        
        # Plot main ticker data on top
        plt.plot(x_values, real_history, 'k-', linewidth=3.0, label=f'{ticker} (historical)')
        if len(normalized_future) > 0:
            plt.plot(x_values, real_future, 'k--', linewidth=2.5, label=f'{ticker} (actual)')
        
        # Plot analysis line
        plt.axvline(x=analysis_x, color='blue', linestyle='-', linewidth=1.5, label='Analysis point')
        
        # Similar adjustment for weighted average trajectory
        weighted_future_points = []
        weighted_future_weights = []
        
        # Collecter tous les points futurs pond√©r√©s
        for cluster in cluster_results['clusters'].values():
            for traj, info in zip(cluster['trajectories'], cluster['pattern_info']):
                n_history_in_cluster = cluster_results.get('n_history', n_forecast)
                future_part = traj[n_history_in_cluster:]
                
                for i, value in enumerate(future_part):
                    if not np.isnan(value):
                        weighted_future_points.append((i, value, info['similarity']**2))
        
        # Calculer moyenne pond√©r√©e pour chaque position future
        if weighted_future_points:
            future_x = np.linspace(analysis_x+1, min(analysis_x+n_forecast, total_display_bars-1), n_forecast)
            future_y = np.full(n_forecast, np.nan)
            
            for pos, val, weight in weighted_future_points:
                if pos < n_forecast:
                    if np.isnan(future_y[pos]):
                        future_y[pos] = val * weight
                        weighted_future_weights.append(weight)
                    else:
                        future_y[pos] += val * weight
                        weighted_future_weights[pos] += weight
            
            # Calculer moyennes finales
            for i in range(len(future_y)):
                if not np.isnan(future_y[i]) and weighted_future_weights[i] > 0:
                    future_y[i] /= weighted_future_weights[i]
            
            # Tracer la ligne moyenne pond√©r√©e
            valid_indices = ~np.isnan(future_y)
            if np.sum(valid_indices) >= 2:
                plt.plot(future_x[valid_indices], future_y[valid_indices], 
                        'k:', linewidth=2.5, label='Weighted avg trajectory')
                
                # Ajouter l'annotation du changement terminal
                if np.sum(valid_indices) > 0:
                    last_idx = np.where(valid_indices)[0][-1]
                    terminal_change = (future_y[last_idx] - 1.0) * 100
                    plt.annotate(
                        f"{terminal_change:+.1f}%", 
                        xy=(future_x[last_idx], future_y[last_idx]),
                        xytext=(future_x[last_idx]+1, future_y[last_idx]),
                        fontsize=10, fontweight='bold', color='black'
                    )
        
        # Plot mean trajectories for each cluster with similar adjustment
        for cluster_id, cluster in cluster_results['clusters'].items():
            color = color_mapping.get(cluster_id, 'blue')
            is_flare = cluster.get('is_flare', False)
            size = cluster['size']
            mean_trajectory = cluster['mean_trajectory'] 
            terminal_change = cluster.get('terminal_change', 0)
            
            # Diviser en histoire/futur
            n_history_in_cluster = cluster_results.get('n_history', n_forecast)
            mean_hist = mean_trajectory[:n_history_in_cluster]
            mean_future = mean_trajectory[n_history_in_cluster:]
            
            # Cr√©er label
            direction = "‚Üë" if terminal_change > 0 else "‚Üì"
            cluster_label = f'{"Flare" if is_flare else "Cluster"} {cluster_id+1}: {direction} {size} paths ({terminal_change:.1f}%)'
            
            # Tracer la moyenne historique interpol√©e
            valid_hist_indices = np.where(~np.isnan(mean_hist))[0]
            if len(valid_hist_indices) >= 2:
                start_idx_mean = valid_hist_indices[0]
                end_idx_mean = valid_hist_indices[-1] + 1
                valid_mean_hist = mean_hist[start_idx_mean:end_idx_mean]
                hist_x = np.linspace(0, analysis_x, len(valid_mean_hist))
                plt.plot(hist_x, valid_mean_hist, '-', color=color, linewidth=2.0)
            
            # Tracer la moyenne future
            valid_future_indices = np.where(~np.isnan(mean_future))[0]
            if len(valid_future_indices) >= 2:
                valid_mean_future = mean_future[valid_future_indices]
                future_x = np.linspace(analysis_x+1, 
                                       min(analysis_x+n_forecast, total_display_bars-1),
                                       len(valid_mean_future))
                plt.plot(future_x, valid_mean_future, '-', color=color, 
                         linewidth=2.0, label=cluster_label)
                
                # Ajouter annotation de terminal change
                if len(valid_mean_future) > 0:
                    plt.annotate(
                        f"{terminal_change:+.1f}%",
                        xy=(future_x[-1], valid_mean_future[-1]),
                        xytext=(future_x[-1]+0.5, valid_mean_future[-1]),
                        fontsize=9, color=color, fontweight='bold',
                        ha='left', va='center'
                    )
        
        # Configure axes and labels
        plt.ylabel('Normalized price (1.0 = price at analysis date)', fontsize=10)
        # Correction: calculer une √©chelle adaptative
        all_values = []
        for cluster_id, cluster in cluster_results['clusters'].items():
            for traj in cluster['trajectories']:
                valid_values = traj[~np.isnan(traj)]
                if len(valid_values) > 0:
                    all_values.extend(valid_values)
        
        # D√©finir des limites d'√©chelle raisonnables bas√©es sur les donn√©es r√©elles
        if all_values:
            y_min = max(0.7, np.percentile(all_values, 5) * 0.95)
            y_max = min(1.3, np.percentile(all_values, 95) * 1.05)
            plt.ylim(y_min, y_max)
        
        # Afficher clairement la ligne de r√©f√©rence √† 1.0
        plt.axhline(y=1.0, color='gray', linestyle='--', alpha=0.5)
        
        # Ajouter une grille plus visible pour am√©liorer la lisibilit√©
        plt.grid(True, alpha=0.4)
        plt.title(f'FLASC Trajectory Analysis for {ticker} (reference: {analysis_date})', fontsize=12)
        plt.legend(loc='upper right', framealpha=0.9, fontsize=9)
        
        # Configure x-axis dates
        if isinstance(ticker_data.index, pd.DatetimeIndex):
            # Create date labels for start, analysis point and end
            analysis_date_obj = analysis_date if isinstance(analysis_date, datetime) else pd.to_datetime(analysis_date)
            start_date = analysis_date_obj - pd.Timedelta(days=n_last)
            dates = [start_date + pd.Timedelta(days=i) for i in range(total_display_bars)]
            date_labels = [d.strftime("%Y-%m-%d") for d in [dates[0], dates[analysis_x], dates[-1]]]
        else:
            date_labels = ["Start", str(analysis_date), "End"]
        
        plt.xticks([0, analysis_x, total_display_bars-1], date_labels)
        plt.xlabel('Date', fontsize=10)
        
        # Add price annotation
        plt.annotate(f"{ticker}: ${reference_price:.2f}",
                    xy=(analysis_x, 1.0),
                    xytext=(analysis_x-5, 1.1),
                    arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
                    fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
        
        # Add notes
        plt.figtext(0.5, 0.01, 
                   f"Note: {available_history}/{n_last} historical bars and {available_future}/{n_forecast} future bars available", 
                   ha='center', fontsize=9, color='red')
        
        # Add cluster summary
        total_main = sum(1 for c in cluster_results['clusters'].values() if not c.get('is_flare', False))
        total_flares = sum(1 for c in cluster_results['clusters'].values() if c.get('is_flare', False))
        
        summary_text = f"Total trajectories: {cluster_results['n_trajectories']}\n"
        summary_text += f"Main clusters: {total_main}, Flares: {total_flares}\n"
        
        plt.figtext(0.02, 0.02, summary_text, fontsize=9, color='#555555',
                   bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
        
        plt.tight_layout()
        plots_dir = os.path.join(os.getcwd(), "plots")
        os.makedirs(plots_dir, exist_ok=True)
        plot_path = os.path.join(plots_dir, f"flasc_trajectories_{ticker}_{interval}.png")
        plt.savefig(plot_path, dpi=150, bbox_inches='tight')
        print(f"üìä FLASC trajectories saved: {plot_path}")

        plt.show()
        print("Debug: Trajectory plot completed")
        
        return cluster_results

    def plot_weighted_outcome_distribution(self, similarity_results, ticker_data, n_forecast=Config.FORECAST_HORIZON, 
                                          analysis_date=None, n_bins=15, bandwidth=None):
        """Plot weighted histogram of expected outcomes with KDE."""
        import matplotlib.pyplot as plt
        from scipy.stats import gaussian_kde
        
        if similarity_results is None or 'similar_patterns' not in similarity_results:
            print("No similar patterns found for visualization")
            return
            
        # Extract key information
        ticker = similarity_results['ticker']
        # Use provided analysis_date or try to get from similarity_results
        analysis_date = analysis_date or similarity_results.get('analysis_date')
        if analysis_date is None:
            print("No analysis date provided for visualization")
            return
        patterns = similarity_results['similar_patterns']
        interval = similarity_results['interval']
        
        # Find index of analysis date
        if isinstance(ticker_data.index, pd.DatetimeIndex):
            try:
                idx = ticker_data.index.get_loc(analysis_date)
            except KeyError:
                # If exact match fails, find closest date
                idx = ticker_data.index.get_indexer([analysis_date], method='nearest')[0]
        else:
            idx = ticker_data.reset_index().index[ticker_data.index == analysis_date][0]
        
        # Reference price
        reference_price = ticker_data['Close'].iloc[idx]
        
        # Prepare terminal values and weights
        terminal_values = []
        weights = []
        
        for pattern in patterns:
            try:
                pattern_ticker = pattern['ticker']
                window_idx = pattern['window_idx']
                similarity = pattern['similarity']
                
                # Load pattern data
                pattern_data = self.data_manager._load_ticker_data(pattern_ticker, interval)
                if pattern_data is None or window_idx >= len(pattern_data):
                    continue
                
                # Check future data availability
                future_available = min(n_forecast, len(pattern_data) - window_idx - 1)
                
                if future_available > 0:
                    # Reference price
                    ref_price = pattern_data['Close'].iloc[window_idx]
                    
                    # Final price
                    final_idx = min(window_idx + n_forecast, len(pattern_data) - 1)
                    final_price = pattern_data['Close'].iloc[final_idx]
                    
                    # Normalized change
                    if ref_price > 0:
                        normalized_change = final_price / ref_price
                        
                        # Store with squared similarity weight
                        terminal_values.append(normalized_change)
                        weights.append(similarity**2)
            except Exception:
                continue
        
        if not terminal_values:
            print("Not enough data for distribution visualization")
            return
        
        # Convert to arrays
        terminal_values = np.array(terminal_values)
        weights = np.array(weights)
        
        # Normalize weights
        if weights.sum() > 0:
            weights = weights / weights.sum()
        
        # Calculate statistics
        weighted_avg = np.average(terminal_values, weights=weights)
        weighted_std = np.sqrt(np.average((terminal_values - weighted_avg)**2, weights=weights))
        avg_change = (weighted_avg - 1.0) * 100
        
        # Calculate probabilities
        bullish_prob = np.sum(weights[terminal_values > 1.0]) * 100
        bearish_prob = np.sum(weights[terminal_values < 1.0]) * 100
        
        # Create figure
        plt.figure(figsize=(12, 7), dpi=100)
        
        # Set display limits
        std_range = 3*weighted_std
        min_val = max(0.7, 1.0 - std_range)
        max_val = min(1.3, 1.0 + std_range)
        
        # Ensure symmetry around 1.0
        range_size = max(max_val - 1.0, 1.0 - min_val)
        min_val = 1.0 - range_size
        max_val = 1.0 + range_size
        
        # Create weighted histogram
        hist, bin_edges, _ = plt.hist(
            terminal_values, 
            bins=n_bins,
            weights=weights,
            alpha=0.7, 
            color='skyblue',
            range=(min_val, max_val),
            density=True,
            edgecolor='black',
            linewidth=1
        )
        
        # Create KDE for smoothed distribution
        kde_sample = []
        for val, w in zip(terminal_values, weights):
            n_samples = max(1, int(w * 100000))
            kde_sample.extend([val] * n_samples)
            
        # Only create KDE if we have sufficient distinct samples
        if len(set(kde_sample)) > 5:
            # Determine bandwidth
            if bandwidth is None:
                bandwidth = 0.9 * weighted_std * len(kde_sample)**(-0.2) * 1.5
                
            # Create KDE
            kde = gaussian_kde(kde_sample, bw_method=bandwidth)
            
            # Plot KDE curve
            x = np.linspace(min_val, max_val, 1000)
            kde_values = kde(x)
            plt.plot(x, kde_values, 'b-', linewidth=2, label='Probability Distribution')
            
            # Mark weighted average
            plt.axvline(x=weighted_avg, color='red', linestyle='--', 
                       label=f'Expected change: {avg_change:+.2f}%')
            
            # Add percentage annotation
            y_pos = plt.gca().get_ylim()[1] * 0.9
            plt.text(weighted_avg + 0.005, y_pos, f"{avg_change:+.2f}%", 
                     color='red', fontweight='bold', ha='left', va='center',
                     bbox=dict(facecolor='white', alpha=0.7, edgecolor='red', pad=3))
        
        # Configure axes
        plt.xlabel(f'Final price after {n_forecast} periods ({interval})', fontsize=12)
        plt.ylabel('Probability density', fontsize=12)
        plt.title(f'Outcome distribution for {ticker} (from {analysis_date})', fontsize=14)
        
        # Add percentage ticks
        tick_step = max(0.01, (max_val - min_val) / 10)
        xticks = np.arange(min_val, max_val + tick_step/2, tick_step)
        if 1.0 not in xticks:
            xticks = np.sort(np.append(xticks, 1.0))
        xtick_labels = [f"{(x-1)*100:+.0f}%" for x in xticks]
        plt.xticks(xticks, xtick_labels)
        
        # Highlight zero change
        plt.axvline(x=1.0, color='gray', linestyle='--', alpha=0.5)
        
        # Add summary statistics
        info_text = (
            f"Results after {n_forecast} periods:\n"
            f"Expected change: {avg_change:+.2f}%\n"
            f"Bullish probability: {bullish_prob:.1f}%\n"
            f"Bearish probability: {bearish_prob:.1f}%\n"
            f"Sample: {len(terminal_values)} trajectories"
        )
        
        plt.annotate(info_text, xy=(0.02, 0.02), xycoords='figure fraction',
                    bbox=dict(boxstyle="round,pad=0.5", facecolor='white', alpha=0.8),
                    fontsize=10)
        
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        # AJOUT: Sauvegarde du plot
        plots_dir = os.path.join(os.getcwd(), "plots")
        os.makedirs(plots_dir, exist_ok=True)
        plot_path = os.path.join(plots_dir, f"outcome_distribution_{ticker}_{interval}.png")
        plt.savefig(plot_path, dpi=150, bbox_inches='tight')
        print(f"üìä Outcome distribution saved: {plot_path}")

        plt.show()
        
        return {
            'weighted_avg': weighted_avg,
            'avg_change_pct': avg_change,
            'weighted_std': weighted_std,
            'bullish_prob': bullish_prob,
            'bearish_prob': bearish_prob,
            'n_trajectories': len(terminal_values)
        }
        


# Configuration du logging
def setup_logging(log_level=Config.LOG_LEVEL):
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    log_dir = os.path.join(Config.BASE_PATH, "logs")
    os.makedirs(log_dir, exist_ok=True)
    log_file = os.path.join(log_dir, f"vector_market_analysis_{timestamp}.log")
    
    logger = logging.getLogger()
    logger.setLevel(log_level)
    
    # Supprimer les handlers existants
    for handler in logger.handlers[:]:
        logger.removeHandler(handler)
    
    # Console handler
    console_handler = logging.StreamHandler()
    console_handler.setLevel(log_level)
    console_format = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    console_handler.setFormatter(console_format)
    
    # File handler
    file_handler = logging.FileHandler(log_file)
    file_handler.setLevel(log_level)
    file_format = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
    file_handler.setFormatter(file_format)
    
    logger.addHandler(console_handler)
    logger.addHandler(file_handler)
    
    return logger

logger = setup_logging()

# Optimisation GPU
def optimize_gpu_memory():
    if not torch.cuda.is_available():
        logger.info("GPU non disponible")
        return
    
    # Information sur la GPU
    device_name = torch.cuda.get_device_name(0)
    total_memory = torch.cuda.get_device_properties(0).total_memory / (1024 ** 3)
    logger.info(f"GPU: {device_name} ({total_memory:.2f} GB)")
    
    # Optimisations
    torch.backends.cudnn.benchmark = True
    if torch.cuda.get_device_capability()[0] >= 7:
        # TF32 pour les GPU Ampere et plus r√©cents
        torch.backends.cuda.matmul.allow_tf32 = True
        torch.backends.cudnn.allow_tf32 = True
    
    # R√©server un tenseur pour √©viter la fragmentation
    dummy = torch.ones((1024, 1024), device='cuda')
    del dummy
    torch.cuda.empty_cache()
    
    logger.info("Optimisations GPU appliqu√©es")

optimize_gpu_memory()



# Cr√©er les r√©pertoires n√©cessaires
os.makedirs(Config.EMBEDDINGS_PATH, exist_ok=True)
os.makedirs(Config.MODELS_PATH, exist_ok=True)
os.makedirs(Config.INDEX_PATH, exist_ok=True)
        
class FinancialTimeSeriesEncoder(nn.Module):
    def __init__(self, input_dim, seq_len, embedding_dim=Config.EMBEDDING_DIM):
        super(FinancialTimeSeriesEncoder, self).__init__()
        self.input_dim = input_dim
        self.seq_len = seq_len
        self.embedding_dim = embedding_dim
        
        # 1. Convolutions simplifi√©es mais toujours multi-√©chelles
        self.conv_layers = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(input_dim, 24, kernel_size=k, padding=k//2),
                nn.BatchNorm1d(24),
                nn.LeakyReLU(0.1)
            ) for k in [3, 7, 15]  # Moins de noyaux et plus petits
        ])
        
        # Nouvelle taille d'entr√©e: 24 * 3 = 72
        conv_output_size = 24 * 3
        
        # 2. Conservation de l'attention mais simplifi√©e avec moins de t√™tes
        self.attention_norm = nn.LayerNorm(conv_output_size)
        self.attention = nn.MultiheadAttention(
            embed_dim=conv_output_size,
            num_heads=2,  # Moins de t√™tes d'attention
            dropout=0.1,
            batch_first=True
        )
        
        # 3. Couche GRU simplifi√©e mais pr√©serv√©e
        self.gru = nn.GRU(
            input_size=conv_output_size,
            hidden_size=96,  # R√©duit mais pas trop
            num_layers=1,    # Une seule couche suffit
            batch_first=True,
            dropout=0.1,
            bidirectional=False
        )
        
        # 4. Projection d'embeddings avec moins de r√©gularisation
        self.embedding_projection = nn.Sequential(
            nn.Linear(96, embedding_dim),
            nn.Dropout(0.1)  # Moins de dropout
        )
        
        # 5. T√™te de pr√©diction simplifi√©e mais garde le m√™me nom
        self.prediction_head = nn.Sequential(
            nn.Linear(embedding_dim, 32),
            nn.LeakyReLU(0.1),
            nn.Linear(32, 1)
        )
        
        # 6. D√©codeur simplifi√©
        self.decoder_projection = nn.Linear(embedding_dim, 96)
        
        # D√©codeur plus simple et direct
        self.decoder = nn.Sequential(
            nn.ConvTranspose1d(96, 48, kernel_size=3, padding=1),
            nn.BatchNorm1d(48),
            nn.LeakyReLU(0.1),
            nn.ConvTranspose1d(48, input_dim, kernel_size=3, padding=1)
        )
        
        self._init_weights()
    
    def _init_weights(self):
        """Initialisation contr√¥l√©e pour la stabilit√©"""
        for m in self.modules():
            if isinstance(m, (nn.Conv1d, nn.ConvTranspose1d, nn.Linear)):
                nn.init.kaiming_normal_(m.weight, a=0.1, nonlinearity='leaky_relu')
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.GRU):
                for name, param in m.named_parameters():
                    if 'weight' in name:
                        nn.init.orthogonal_(param)
                    elif 'bias' in name:
                        nn.init.zeros_(param)
    
    def encode(self, x):
        # Force sync CUDA pour √©viter race conditions
        if x.is_cuda:
            torch.cuda.current_stream().synchronize()
            
        # Traiter les valeurs invalides
        x = torch.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0)
        
        # Transformation pour Conv1D: [batch, seq, features] -> [batch, features, seq]
        x_conv = x.permute(0, 2, 1)
        
        # 1. Convolutions multi-√©chelles
        conv_outputs = []
        for conv in self.conv_layers:
            conv_outputs.append(conv(x_conv))
        
        # Concat√©nation des sorties de convolution
        x = torch.cat(conv_outputs, dim=1)
        
        # Transformation pour attention: [batch, channels, seq] -> [batch, seq, channels]
        x = x.permute(0, 2, 1)
        
        # 2. Auto-attention simplifi√©e avec r√©sidus
        x_norm = self.attention_norm(x)
        attn_output, _ = self.attention(x_norm, x_norm, x_norm)
        x = x + attn_output * 0.5  # Connexion r√©siduelle avec r√©duction de la contribution
        
        # 3. GRU pour l'asym√©trie temporelle
        gru_out, _ = self.gru(x)
        
        # S√©lection du dernier √©tat cach√©
        x = gru_out[:, -1, :]
        
        # 4. Projection d'embeddings
        embedding = self.embedding_projection(x)
        
        return embedding
    
    def predict_returns(self, embedding):
        """Pr√©dit les rendements futurs √† partir de l'embedding"""
        return self.prediction_head(embedding).squeeze(-1)
    
    def decode(self, embedding):
        """D√©code l'embedding en s√©rie temporelle reconstruite"""
        # Projection vers l'espace de d√©codage
        features = self.decoder_projection(embedding)
        
        # Redimensionnement pour la d√©convolution
        features = features.unsqueeze(2)
        features = features.expand(-1, -1, self.seq_len)
        
        # D√©codage
        reconstruction = self.decoder(features)
        
        # Retour au format d'origine [batch, seq, features]
        reconstruction = reconstruction.permute(0, 2, 1)
        
        return reconstruction
    
    def forward(self, x):
        embedding = self.encode(x)
        predicted_returns = self.predict_returns(embedding)  # Corrig√©: utilise predict_returns au lieu de predict_head
        reconstructions = self.decode(embedding)
        return embedding, predicted_returns, reconstructions




# Dataset et gestion de donn√©es
class TimeSeriesDataset(Dataset):
    def __init__(self, data_manager, interval, window_size=None, forecast_horizon=Config.FORECAST_HORIZON):
        self.data_manager = data_manager
        self.interval = interval
        self.window_size = window_size or Config.get_window_size(interval)
        self.forecast_horizon = forecast_horizon
        self.tickers = self.data_manager.get_available_tickers(interval)
        self.windows = []
        
        logger.info(f"Initialisation du dataset pour {interval} avec {len(self.tickers)} tickers")
        
        # Pr√©charger les fen√™tres disponibles avec leurs rendements futurs
        self._preload_valid_windows()

    def __len__(self):
        return len(self.windows)
    
    def _preload_valid_windows(self):
        """Pr√©chargement optimis√© avec normalisation par ticker"""
        self.windows = []
        
        # D√©finir des r√©gimes de march√© significatifs
        market_regimes = {
            'strong_uptrend': lambda df: (df['Close'].pct_change(20) > 0.1),
            'strong_downtrend': lambda df: (df['Close'].pct_change(20) < -0.1),
            'weak_uptrend': lambda df: (df['Close'].pct_change(20).between(0.02, 0.1)),
            'weak_downtrend': lambda df: (df['Close'].pct_change(20).between(-0.1, -0.02)),
            'sideways': lambda df: (df['Close'].pct_change(20).abs() < 0.02),
            'volatility_expansion': lambda df: (df['Close'].pct_change().rolling(20).std().pct_change(10) > 0.2),
            'volatility_contraction': lambda df: (df['Close'].pct_change().rolling(20).std().pct_change(10) < -0.2),
        }
        
        regime_counts = {regime: 0 for regime in market_regimes}
        target_samples_per_regime = 2500
        
        for ticker in tqdm(self.tickers, desc=f"Pr√©chargement des fen√™tres ({self.interval})"):
            ticker_data = self.data_manager._load_ticker_data(ticker, self.interval)
            if ticker_data is None or len(ticker_data) < self.window_size + self.forecast_horizon:
                continue
            
            # V√©rifier les donn√©es de base
            essential_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
            missing_ratio = ticker_data[essential_cols].isna().mean().mean()
            if missing_ratio > 0.1 or ticker_data['Close'].pct_change().std() < 0.001:
                continue
            
            # üî• OPTIMISATION: Normaliser tout le ticker UNE SEULE FOIS
            available_metrics = [m for m in Config.METRICS if m in ticker_data.columns]
            normalized_data = enhanced_financial_preprocessing(ticker_data[available_metrics], normalize=True)
            
            if normalized_data is None:
                continue
                
            # Cr√©er tableau de caract√©ristiques pour tout le ticker
            ticker_features = np.zeros((len(ticker_data), len(Config.METRICS)))
            for i, metric in enumerate(Config.METRICS):
                norm_col = f"{metric}_norm" if metric != 'log_return' else 'log_return_norm'
                if norm_col in normalized_data.columns:
                    values = normalized_data[norm_col].values
                    ticker_features[:len(values), i] = values
            
            # Traiter les NaN
            ticker_features = np.nan_to_num(ticker_features, nan=0.0)
            
            # Calculer m√©triques suppl√©mentaires pour classification des r√©gimes
            ticker_data['Volatility'] = ticker_data['Close'].pct_change().rolling(20).std()
            
            # Identifier les fen√™tres pour chaque r√©gime
            for regime_name, condition in market_regimes.items():
                if regime_counts[regime_name] >= target_samples_per_regime:
                    continue
                
                try:
                    regime_mask = condition(ticker_data)
                    valid_indices = [i for i, valid in enumerate(regime_mask) 
                                    if valid and i >= self.window_size and 
                                    i + self.forecast_horizon < len(ticker_data)]
                    
                    max_per_ticker = min(len(valid_indices), 50)
                    
                    if valid_indices:
                        if len(valid_indices) > max_per_ticker:
                            valid_indices = np.array_split(valid_indices, max_per_ticker)
                            valid_indices = [indices[len(indices)//2] for indices in valid_indices]
                        
                        for idx in valid_indices:
                            # Calculer rendement futur
                            reference_price = ticker_data['Close'].iloc[idx]
                            future_prices = ticker_data['Close'].iloc[idx+1:idx+self.forecast_horizon+1]
                            
                            if len(future_prices) < self.forecast_horizon // 2:
                                continue
                                
                            target_idx = min(self.forecast_horizon, len(future_prices)) - 1
                            future_return = float((future_prices.iloc[target_idx] / reference_price) - 1)
                            
                            # üî• EXTRACTION RAPIDE: Prendre la fen√™tre depuis les features pr√©calcul√©es
                            window_start = idx - self.window_size
                            if window_start >= 0:
                                window_features = ticker_features[window_start:window_start+self.window_size]
                                input_tensor = torch.tensor(window_features, dtype=torch.float32)
                                
                                # Stocker le tensor pr√©calcul√©
                                self.windows.append((input_tensor, future_return, regime_name))
                                regime_counts[regime_name] += 1
                            
                except Exception as e:
                    continue
        
        # Log des statistiques
        total = sum(regime_counts.values())
        logger.info(f"Distribution des r√©gimes de march√©:")
        for regime, count in regime_counts.items():
            logger.info(f"  {regime}: {count} ({count/max(1,total)*100:.1f}%)")
            
        logger.info(f"Dataset pour {self.interval}: {len(self.windows)} fen√™tres valides apr√®s filtrage")


    def __getitem__(self, idx):
        input_tensor, future_return, regime = self.windows[idx]
        
        # Appliquer augmentation de donn√©es (avec 30% de probabilit√©)
        if random.random() < 0.3:
            input_tensor = self._augment_window(input_tensor.numpy())
            input_tensor = torch.tensor(input_tensor, dtype=torch.float32)
        
        return input_tensor, float(future_return)
        
    def _prepare_financial_window(self, window_data):
        """
        Pr√©pare une fen√™tre de donn√©es financi√®res pour le mod√®le :
        - Normalisation avanc√©e via enhanced_financial_preprocessing
        - Gestion stricte des NaN et colonnes manquantes
        - Retourne un tensor shape (window_size, n_features)
        """
        # V√©rification d'entr√©e
        if window_data is None or len(window_data) == 0:
            # Retourne un tensor vide si la fen√™tre est invalide
            return torch.zeros((self.window_size, len(Config.METRICS)), dtype=torch.float32)

        # Pr√©traitement avanc√©
        processed = enhanced_financial_preprocessing(window_data, normalize=True)
        feature_array = np.zeros((self.window_size, len(Config.METRICS)), dtype=np.float32)

        for i, metric in enumerate(Config.METRICS):
            norm_col = f"{metric}_norm"
            # Cas particulier pour log_return (souvent d√©j√† normalis√©)
            if metric == 'log_return':
                norm_col = 'log_return_norm'
            if processed is not None and norm_col in processed.columns:
                values = processed[norm_col].values
                # Tronque ou compl√®te √† la bonne taille
                n = min(len(values), self.window_size)
                feature_array[:n, i] = values[:n]
            else:
                # Si la colonne n'existe pas, laisse √† z√©ro (feature neutre)
                continue

        # Nettoyage final : NaN ‚Üí 0
        feature_array = np.nan_to_num(feature_array, nan=0.0, posinf=0.0, neginf=0.0)
        return torch.tensor(feature_array, dtype=torch.float32)
    
    def _augment_window(self, window, strength=0.3):
        """Augmentation de donn√©es sp√©cifique √† la finance"""
        # Implementation de la fonction d'augmentation
        augmented = window.copy()
        
        # S√©lection al√©atoire d'une transformation
        aug_type = random.choice([
            'time_warp', 'add_noise', 'scale_amplitude', 
            'change_trend', 'jitter_volume'
        ])
        
        if aug_type == 'time_warp':
            # √âtirer/comprimer l√©g√®rement l'√©chelle temporelle
            from scipy import ndimage
            time_scale = np.random.uniform(1-strength/2, 1+strength/2)
            augmented = ndimage.zoom(augmented, (1, time_scale), order=1, mode='nearest')
            
            # Recadrage pour maintenir la taille originale
            if augmented.shape[1] > window.shape[1]:
                augmented = augmented[:, :window.shape[1]]
            elif augmented.shape[1] < window.shape[1]:
                padding = np.zeros((augmented.shape[0], window.shape[1] - augmented.shape[1]))
                augmented = np.concatenate([augmented, padding], axis=1)
                
        elif aug_type == 'add_noise':
            # Bruit sp√©cifique par m√©trique
            for i in range(augmented.shape[0]):
                noise_scale = 0.001 if i < 4 else 0.01  # Diff√©rent pour prix vs autres m√©triques
                noise = np.random.normal(0, noise_scale * strength, size=augmented.shape[1])
                augmented[i, :] += noise
                
        elif aug_type == 'scale_amplitude':
            # √âchelle d'amplitude (diff√©rentes volatilit√©s)
            scale_factor = np.random.uniform(1-strength, 1+strength)
            mean_vals = np.mean(augmented, axis=1, keepdims=True)
            augmented = mean_vals + (augmented - mean_vals) * scale_factor
            
        elif aug_type == 'change_trend':
            # Modification l√©g√®re de tendance
            trend_strength = np.random.uniform(-strength, strength) * 0.01
            trend = np.linspace(0, trend_strength * augmented.shape[1], augmented.shape[1])
            
            # Appliquer aux prix
            for i in range(min(4, augmented.shape[0])):
                augmented[i, :] += trend
                
        elif aug_type == 'jitter_volume':
            # Modifier les volumes 
            if augmented.shape[0] > 4:  # S'assurer que l'indice existe
                vol_idx = 4  # Indice du volume
                vol_scale = np.random.uniform(1-strength*2, 1+strength*2)
                augmented[vol_idx, :] *= vol_scale
        
        return augmented


            
# Fonction d'entra√Ænement du mod√®le
def train_financial_model(data_manager, interval, epochs=Config.NUM_EPOCHS, batch_size=None, val_split=Config.VALIDATION_SPLIT):
    """Entra√Æne le mod√®le financier avec utilisation correcte de __getitem__"""
    logger.info(f"Pr√©paration de l'entra√Ænement pour {interval}")
    
    # V√âRIFICATION CRITIQUE: Tickers disponibles
    available_tickers = data_manager.get_available_tickers(interval)
    if len(available_tickers) == 0:
        logger.error(f"‚ùå Aucun ticker disponible pour {interval}")
        print(f"‚ùå ERREUR: Aucune donn√©e trouv√©e pour l'intervalle {interval}")
        print("üí° V√©rifiez que les fichiers parquet existent dans:")
        print(f"   {Config.BASE_PATH}\\{interval}\\")
        return None, None
    
    print(f"‚úÖ {len(available_tickers)} tickers trouv√©s pour {interval}")
    
    # Cr√©er le dataset avec window_size correct
    window_size = Config.get_window_size(interval)
    dataset = TimeSeriesDataset(data_manager, interval, window_size=window_size)
    
    # V√âRIFICATION CRITIQUE: Dataset non vide
    if len(dataset) == 0:
        logger.error(f"‚ùå Dataset vide apr√®s filtrage pour {interval}")
        print(f"‚ùå ERREUR: Aucune fen√™tre valide apr√®s filtrage pour {interval}")
        print("üí° Possible causes:")
        print("   - Donn√©es insuffisantes par ticker")
        print("   - M√©triques manquantes (RSI, MACD, etc.)")
        print("   - Fen√™tres de temps trop grandes")
        return None, None
    
    logger.info(f"‚úÖ Dataset cr√©√©: {len(dataset)} fen√™tres valides")
    
    # ‚úÖ CORRECTION MAJEURE: Utiliser le dataset directement (pas de pr√©chargement)
    # Split train/validation sur le dataset original
    train_size = int(len(dataset) * (1 - val_split))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
    
    logger.info(f"Split: {train_size} entra√Ænement, {val_size} validation")
    
    # GPU warm-up
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        warm_up = torch.randn(1024, 1024, device=Config.DEVICE)
        del warm_up
        torch.cuda.empty_cache()
        logger.info(f"Warm-up GPU termin√©")
    
    # ‚úÖ DataLoaders qui utilisent __getitem__ correctement
    try:
        train_loader = DataLoader(
            train_dataset, 
            batch_size=batch_size or Config.get_batch_size(),
            shuffle=True,
            num_workers=0,
            pin_memory=False,
            drop_last=True
        )
        
        val_loader = DataLoader(
            val_dataset, 
            batch_size=batch_size or Config.get_batch_size(),
            shuffle=False,
            num_workers=0,
            pin_memory=False,
            drop_last=False
        )
        
        # Test du premier batch pour d√©tecter les probl√®mes
        test_batch = next(iter(train_loader))
        logger.info(f"‚úÖ Premier batch cr√©√© avec succ√®s: {test_batch[0].shape}")
        
    except Exception as e:
        logger.error(f"‚ùå ERREUR DataLoader: {str(e)}")
        import traceback
        traceback.print_exc()
        return None, None
    
    # Correction: D√©finition du mod√®le
    # Correction: D√©finition du mod√®le
    model = FinancialTimeSeriesEncoder(
        input_dim=len(Config.METRICS),
        seq_len=Config.get_window_size(interval),
        embedding_dim=Config.EMBEDDING_DIM
    )
    model.to(Config.DEVICE)

    # üîç DEBUG: Test du forward pass
    try:
        test_batch = next(iter(train_loader))
        test_data = test_batch[0][:1].to(Config.DEVICE)  # Un seul √©chantillon
        logger.info(f"üîç Test input shape: {test_data.shape}")
        
        with torch.no_grad():
            embeddings = model.encode(test_data)
            logger.info(f"‚úÖ Encode OK: {embeddings.shape}")
            
            reconstructions = model.decode(embeddings)
            logger.info(f"‚úÖ Decode OK: {reconstructions.shape}")
            
            predicted_returns = model.predict_returns(embeddings)
            logger.info(f"‚úÖ Predict OK: {predicted_returns.shape}")
            
    except Exception as e:
        logger.error(f"‚ùå FORWARD PASS ERROR: {str(e)}")
        import traceback
        traceback.print_exc()
        return None, None
    
    # Optimizer
    optimizer = optim.AdamW(
        model.parameters(), 
        lr=Config.LEARNING_RATE,
        weight_decay=Config.WEIGHT_DECAY,
        eps=1e-8
    )
    
    # Scheduler
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=3, verbose=True, min_lr=1e-6
    )
    
    # Loss functions
    criterion_reconstruction = nn.MSELoss(reduction='mean')
    criterion_prediction = FinancialPredictionLoss(
        direction_weight=Config.OPT_METRIC_WEIGHTS['direction_weight'], 
        magnitude_weight=Config.OPT_METRIC_WEIGHTS['magnitude_weight'], 
        consistency_weight=Config.OPT_METRIC_WEIGHTS['calibration_weight']
    )
    
    # Early stopping
    patience = Config.EARLY_STOPPING_PATIENCE
    best_val_loss = float('inf')
    epochs_no_improve = 0
    history = {'train_loss': [], 'val_loss': [], 'lr': []}
    
    # Mixed precision scaler
    scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None
    
    logger.info("D√©marrage de l'entra√Ænement avec __getitem__ fonctionnel...")
    
    for epoch in range(epochs):
        model.train()
        epoch_train_loss = 0.0
        num_batches = 0
        
        epoch_start = time.time()
        
        # ‚úÖ MAINTENANT __getitem__ est appel√©e √† chaque batch avec augmentation !
        with tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}") as pbar:
            for batch_idx, (data, target) in enumerate(pbar):
                # ‚Üë data et target viennent de __getitem__ avec augmentation al√©atoire !
                
                data = data.to(Config.DEVICE, dtype=torch.float32)
                target = target.to(Config.DEVICE, dtype=torch.float32)
                
                optimizer.zero_grad(set_to_none=True)
                
                if scaler:
                    with torch.cuda.amp.autocast():
                        embeddings = model.encode(data)
                        reconstructions = model.decode(embeddings)
                        predicted_returns = model.predict_returns(embeddings)
                        loss_recon = criterion_reconstruction(reconstructions, data)
                        loss_pred = criterion_prediction(predicted_returns, target)
                        loss = loss_recon + loss_pred
                    
                    scaler.scale(loss).backward()
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    embeddings = model.encode(data)
                    reconstructions = model.decode(embeddings)
                    predicted_returns = model.predict_returns(embeddings)
                    loss_recon = criterion_reconstruction(reconstructions, data)
                    loss_pred = criterion_prediction(predicted_returns, target)
                    loss = loss_recon + loss_pred
                    
                    loss.backward()
                    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                    optimizer.step()
                
                batch_loss = loss.item()
                epoch_train_loss += batch_loss
                num_batches += 1
                
                if torch.cuda.is_available():
                    torch.cuda.synchronize()
                
                speed = (time.time() - epoch_start) / (batch_idx + 1)
                pbar.set_postfix({
                    'loss': batch_loss, 
                    'lr': optimizer.param_groups[0]['lr'],
                    'sec/it': f"{speed:.2f}",
                    'GPU%': gpu_utilization()
                })
        
        if num_batches > 0:
            epoch_train_loss /= num_batches
            history['train_loss'].append(epoch_train_loss)
            history['lr'].append(optimizer.param_groups[0]['lr'])
        else:
            logger.warning("Aucun batch valide dans cette √©poque")
            continue
        
        # Phase de validation
        model.eval()
        epoch_val_loss = 0.0
        val_batches = 0
        
        with torch.no_grad():
            for data, target in val_loader:  # ‚Üê __getitem__ appel√©e ici aussi (mais sans augmentation pour la validation)
                data = data.to(Config.DEVICE, dtype=torch.float32, non_blocking=True)
                target = target.to(Config.DEVICE, dtype=torch.float32, non_blocking=True)
                data.nan_to_num_(nan=0.0, posinf=0.0, neginf=0.0)
                try:
                    embeddings = model.encode(data)
                    reconstructions = model.decode(embeddings)
                    predicted_returns = model.predict_returns(embeddings)
                    loss_recon = criterion_reconstruction(reconstructions, data)
                    loss_pred = criterion_prediction(predicted_returns, target)
                    loss = loss_recon + loss_pred
                    epoch_val_loss += loss.item()
                    val_batches += 1
                except Exception as e:
                    logger.error(f"Erreur validation: {str(e)}")
                    continue
        
        if val_batches > 0:
            epoch_val_loss /= val_batches
            history['val_loss'].append(epoch_val_loss)
            
            scheduler.step(epoch_val_loss)
            
            logger.info(f"Epoch {epoch+1}/{epochs} - "
                      f"Train Loss: {epoch_train_loss:.6f}, "
                      f"Val Loss: {epoch_val_loss:.6f}, "
                      f"LR: {optimizer.param_groups[0]['lr']:.6f}")
            
            # Early stopping et sauvegarde du meilleur mod√®le
            if epoch_val_loss < best_val_loss:
                best_val_loss = epoch_val_loss
                epochs_no_improve = 0
                
                model_path = os.path.join(Config.MODELS_PATH, f"encoder_{interval}.pth")
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': model.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'train_loss': epoch_train_loss,
                    'val_loss': epoch_val_loss,
                    'config': {
                        'input_dim': len(Config.METRICS),
                        'seq_len': Config.get_window_size(interval),
                        'embedding_dim': Config.EMBEDDING_DIM
                    }
                }, model_path)
                logger.info(f"Meilleur mod√®le sauvegard√©: {model_path}")
            else:
                epochs_no_improve += 1
                if epochs_no_improve >= patience:
                    logger.info(f"Early stopping apr√®s {epoch+1} √©poques")
                    break
        else:
            logger.warning("Aucun batch de validation valide")
    
    # Nettoyage final
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
    
    return model, history



class FinancialPredictionLoss(nn.Module):
    """Fonction de perte personnalis√©e pour la pr√©diction financi√®re"""
    def __init__(self, direction_weight=0.6, magnitude_weight=0.35, consistency_weight=0.05):
        super(FinancialPredictionLoss, self).__init__()
        self.direction_weight = direction_weight
        self.magnitude_weight = magnitude_weight
        self.consistency_weight = consistency_weight
        
    def forward(self, predicted_returns, actual_returns):
        # 1. Erreur de direction (binaire)
        direction_actual = (actual_returns > 0).float().to(predicted_returns.device)
        direction_pred = (predicted_returns > 0).float()
        direction_loss = F.binary_cross_entropy_with_logits(
            predicted_returns, 
            direction_actual
        )
        
        # 2. Erreur de magnitude - Huber Loss
        magnitude_loss = F.huber_loss(
            predicted_returns, 
            actual_returns.to(predicted_returns.device), 
            delta=0.05
        )
        
        # 3. R√©gularisation pour la consistance des pr√©dictions
        if len(predicted_returns) > 1:
            consistency_loss = torch.var(predicted_returns) * 0.1
        else:
            consistency_loss = torch.tensor(0.0, device=predicted_returns.device)
        
        total_loss = (
            self.direction_weight * direction_loss + 
            self.magnitude_weight * magnitude_loss +
            self.consistency_weight * consistency_loss
        )
        return total_loss


def enhanced_financial_preprocessing(data, normalize=True):
    """
    Pr√©traitement avanc√© sp√©cifique aux donn√©es financi√®res
    """
    if data is None or len(data) == 0:
        return None
    
    # S√©lectionner les m√©triques disponibles
    available_metrics = [m for m in Config.METRICS if m in data.columns]
    
    if not normalize:
        return data[available_metrics].copy()
    
    # Dataframe pour les r√©sultats normalis√©s
    processed_data = pd.DataFrame(index=data.index)
    
    # 1. Log-transformations pour distributions √† queue lourde
    if 'Volume' in available_metrics:
        # Volume suit g√©n√©ralement une distribution log-normale
        log_vol = np.log(data['Volume'].replace(0, 1))
        processed_data['Volume_norm'] = robust_zscore(log_vol)
    
    # 2. Normalisation robuste pour les prix (relative √† ATR plut√¥t que std)
    price_cols = ['Open', 'High', 'Low', 'Close']
    for col in [c for c in price_cols if c in available_metrics]:
        # Utiliser des rendements relatifs au lieu de diff√©rences absolues
        processed_data[f"{col}_norm"] = calculate_returns(data[col], method='log')
        
        # Option: normaliser par ATR (Average True Range) - mesure de volatilit√©
        if 'High' in available_metrics and 'Low' in available_metrics:
            atr = calculate_atr(data, period=14)
            if atr is not None and atr.mean() > 0:
                processed_data[f"{col}_atr_norm"] = processed_data[f"{col}_norm"] / atr
    
    # 3. Traitement sp√©cial pour les indicateurs techniques qui ont des bornes naturelles
    if 'RSI14' in available_metrics:
        # RSI est d√©j√† entre 0-100, donc normaliser vers [-1,1] avec point neutre √† 50
        rsi_norm = (data['RSI14'] / 50) - 1
        processed_data['RSI14_norm'] = np.clip(rsi_norm, -1.0, 1.0)
        
        # Ajouter une transformation non-lin√©aire pour accentuer les extr√™mes
        # Plus sensible pr√®s de 0 et 100 qu'au milieu
        processed_data['RSI14_extreme'] = np.tanh((data['RSI14'] - 50) / 10)
    
    # 4. D√©tection et traitement des valeurs aberrantes
    for col in processed_data.columns:
        # Appliquer winsorization pour limiter l'impact des outliers
        processed_data[col] = winsorize(processed_data[col], limits=[0.01, 0.01])
    
    # 5. Ajout de caract√©ristiques d√©riv√©es (optionnel, selon utilisation)
    if 'Close_norm' in processed_data.columns:
        # Volatilit√© locale
        processed_data['volatility'] = data['Close'].pct_change().rolling(10).std()
        processed_data['volatility_norm'] = robust_zscore(processed_data['volatility'])
        
        # Momentum (diff√©rentes √©chelles temporelles)
        for period in [5, 10, 20]:
            momentum = data['Close'].pct_change(period)
            processed_data[f'momentum_{period}'] = np.clip(momentum, -0.2, 0.2)
    
    return processed_data

def calculate_returns(series, method='log'):
    """
    Calcule les rendements d'une s√©rie de prix.
    method: 'log' (par d√©faut) ou 'pct'
    """
    series = pd.Series(series)
    if method == 'log':
        returns = np.log(series).diff().fillna(0)
    elif method == 'pct':
        returns = series.pct_change().fillna(0)
    else:
        raise ValueError("M√©thode non support√©e pour calculate_returns")
    return returns

def calculate_atr(df, period=14):
    """
    Calcule l'Average True Range (ATR) sur un DataFrame de prix.
    """
    if not all(col in df.columns for col in ['High', 'Low', 'Close']):
        return pd.Series(np.nan, index=df.index)
    high = df['High']
    low = df['Low']
    close = df['Close']
    prev_close = close.shift(1)
    tr = pd.concat([
        (high - low).abs(),
        (high - prev_close).abs(),
        (low - prev_close).abs()
    ], axis=1).max(axis=1)
    atr = tr.rolling(window=period, min_periods=1).mean()
    return atr

def winsorize(series, limits=[0.01, 0.01]):
    """
    Applique une winsorisation √† une s√©rie (remplace les extr√™mes par les quantiles).
    """
    s = pd.Series(series).copy()
    lower = s.quantile(limits[0])
    upper = s.quantile(1 - limits[1])
    s = s.clip(lower, upper)
    return s

def robust_zscore(series, winsorize_limits=[0.01, 0.01]):
    """Z-score robuste utilisant m√©diane et MAD au lieu de moyenne/√©cart-type"""
    from scipy.stats import mstats
    
    # Winsorisation pour limiter l'impact des outliers
    win_series = mstats.winsorize(series.dropna(), limits=winsorize_limits)
    
    # Z-score robuste
    median = np.median(win_series)
    mad = np.median(np.abs(win_series - median))  # Median Absolute Deviation
    
    # √âviter division par z√©ro
    if mad == 0:
        mad = np.std(win_series) or 0.0001
    
    # Facteur de correction pour MAD
    k = 1.4826  # Pour distribution normale
    
    # Normalisation
    z_scores = k * (series - median) / mad
    
    # Limiter les valeurs extr√™mes
    return np.clip(z_scores, -5, 5)



def gpu_utilization():
    """Retourne l'utilisation GPU en pourcentage si disponible"""
    if not torch.cuda.is_available():
        return "N/A"
    try:
        # Essayer d'obtenir l'utilisation via pynvml
        import pynvml
        pynvml.nvmlInit()
        handle = pynvml.nvmlDeviceGetHandleByIndex(0)
        util = pynvml.nvmlDeviceGetUtilizationRates(handle)
        return f"{util.gpu}%"
    except:
        return "?"

# G√©n√©ration des embeddings
def generate_embeddings(data_manager, interval, model, batch_size=None):
    """G√©n√®re les embeddings avec traitement parall√®le optimis√©."""
    logger.info(f"G√©n√©ration optimis√©e des embeddings pour {interval}")
    
    model.eval()
    model.to(Config.DEVICE)
    
    # R√©cup√©rer tous les tickers disponibles
    tickers = data_manager.get_available_tickers(interval)
    logger.info(f"Traitement de {len(tickers)} tickers pour {interval}")
    
    all_windows = []
    all_metadata = []
    
    # üî• TRAITEMENT PARALL√àLE OPTIMIS√â
    logger.info("Utilisation du mode parall√®le optimis√©")
    
    # Gros batches pour maximiser l'utilisation CPU/GPU
    batch_size_tickers = 200
    
    # Utiliser ProcessPoolExecutor pour le preprocessing
    with ThreadPoolExecutor (max_workers=Config.NUM_WORKERS) as executor:
        batch_futures = []
        
        for i in range(0, len(tickers), batch_size_tickers):
            batch_tickers = tickers[i:i+batch_size_tickers]
            
            # Soumettre le batch au pool de processus
            future = executor.submit(
                preprocess_ticker_chunk, 
                batch_tickers, interval, data_manager.base_path, 
                Config.METRICS, Config.get_window_size(interval)
            )
            batch_futures.append(future)
        
        # Collecter les r√©sultats
        for i, future in enumerate(batch_futures):
            try:
                batch_windows, batch_metadata = future.result(timeout=120)
                all_windows.extend(batch_windows)
                all_metadata.extend(batch_metadata)
                
                logger.info(f"Lot {i+1}/{len(batch_futures)} termin√©: {len(all_windows)} fen√™tres totales")
                
                # Lib√©rer la m√©moire apr√®s chaque lot
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
                    
            except Exception as e:
                logger.error(f"Erreur lot {i+1}: {str(e)}")
                continue
    
    # V√©rifier qu'on a des r√©sultats
    if not all_windows:
        logger.warning(f"Aucun embedding g√©n√©r√© pour {interval}")
        return None, None
    
    # Convertir en array pour le traitement GPU
    logger.info(f"G√©n√©ration d'embeddings GPU pour {len(all_windows)} fen√™tres")
    
    # Utiliser des lots pour l'encodage GPU
    gen_batch_size = batch_size or Config.get_batch_size()
    
    # Forcer l'activation du GPU
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        warm_up = torch.randn(1024, Config.get_window_size(interval), len(Config.METRICS), device=Config.DEVICE)
        del warm_up
        torch.cuda.empty_cache()
        logger.info("GPU activ√© pour la g√©n√©ration d'embeddings")
    
    # Traiter par lots pour maximiser l'utilisation GPU
    embeddings_list = []
    
    with torch.cuda.amp.autocast() if torch.cuda.is_available() else nullcontext():
        for j in range(0, len(all_windows), gen_batch_size):
            batch_end = min(j + gen_batch_size, len(all_windows))
            batch_windows = all_windows[j:batch_end]
            
            # Progression
            if j % (gen_batch_size * 5) == 0:
                logger.info(f"  G√©n√©ration d'embeddings: {j}/{len(all_windows)} fen√™tres ({j/len(all_windows)*100:.1f}%)")
            
            # Convertir en tensors
            batch_tensors = torch.tensor(np.array(batch_windows), dtype=torch.float32, device=Config.DEVICE)
            
            # G√©n√©ration
            with torch.no_grad():
                if torch.cuda.is_available():
                    torch.cuda.synchronize()
                
                batch_embeddings = model.encode(batch_tensors).cpu().numpy()
                embeddings_list.append(batch_embeddings)
    
    # Concat√©ner les embeddings
    embeddings_array = np.vstack(embeddings_list)
    
    # Sauvegarder
    embeddings_path = os.path.join(Config.EMBEDDINGS_PATH, f"embeddings_{interval}.h5")
    metadata_path = os.path.join(Config.EMBEDDINGS_PATH, f"metadata_{interval}.pkl")
    
    with h5py.File(embeddings_path, 'w') as f:
        f.create_dataset('embeddings', data=embeddings_array)
    
    with open(metadata_path, 'wb') as f:
        pickle.dump(all_metadata, f)
    
    logger.info(f"Sauvegarde de {len(all_metadata)} embeddings pour {interval}: {embeddings_path}")
    
    # Lib√©rer la m√©moire
    del all_windows, embeddings_list
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
    
    return embeddings_array, all_metadata

def preprocess_ticker_chunk(tickers, interval, data_manager_path, metrics, window_size):
    """Traite plusieurs tickers √† la fois pour r√©duire les frais g√©n√©raux."""
    
    # Cr√©er un data manager minimal dans le processus enfant
    class SimpleDataManager:
        def __init__(self, base_path):
            self.base_path = base_path
        
        def _load_ticker_data(self, ticker, interval):
            file_path = os.path.join(self.base_path, interval, f"{ticker}.parquet")
            if not os.path.exists(file_path):
                return None
            try:
                return pd.read_parquet(file_path)
            except:
                return None
    
    local_data_manager = SimpleDataManager(data_manager_path)
    all_windows = []
    all_metadata = []
    
    # Traiter chaque ticker du lot
    for ticker in tickers:
        try:
            ticker_data = local_data_manager._load_ticker_data(ticker, interval)
            if ticker_data is None or len(ticker_data) < window_size:
                continue
            
            # Calculer le nombre max de fen√™tres selon la longueur des donn√©es
            data_length = len(ticker_data) - window_size + 1
            if data_length <= 0:
                continue
                
            # Strat√©gie d'√©chantillonnage adaptatif
            max_windows = min(500, max(100, int(data_length * 0.1)))
            
            # √âchantillonnage distribu√© uniform√©ment sur la timeline
            if data_length <= max_windows:
                window_indices = list(range(0, data_length))
            else:
                window_indices = [int(i * data_length / max_windows) for i in range(max_windows)]
            
            # Traiter chaque fen√™tre s√©lectionn√©e
            for window_start in window_indices:
                window_data = ticker_data.iloc[window_start:window_start+window_size]
                
                # V√©rifier les m√©triques essentielles
                required_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
                if not all(col in window_data.columns for col in required_cols):
                    continue
                
                # Normaliser les donn√©es
                try:
                    # S√©lectionner les m√©triques disponibles
                    available_metrics = [m for m in metrics if m in window_data.columns]
                    
                    # Tableau pour stocker les caract√©ristiques normalis√©es
                    feature_array = np.zeros((window_size, len(metrics)))
                    
                    # Normaliser chaque m√©trique selon son type
                    for i, metric in enumerate(metrics):
                        if metric in available_metrics:
                            values = window_data[metric].values
                            
                            # Normalisation sp√©cifique par type de m√©trique
                            if metric in ['Open', 'High', 'Low', 'Close']:
                                if len(values) > 1:
                                    values_norm = np.diff(np.log(np.maximum(values, 0.001)), prepend=np.log(values[0]))
                                    values_norm = np.clip(values_norm, -0.1, 0.1)
                                else:
                                    values_norm = np.zeros_like(values)
                                    
                            elif metric == 'Volume':
                                log_vol = np.log(np.maximum(values, 1))
                                mean_vol = np.nanmean(log_vol)
                                std_vol = np.nanstd(log_vol) or 1.0
                                values_norm = (log_vol - mean_vol) / std_vol
                                values_norm = np.clip(values_norm, -3.0, 3.0)
                                
                            elif metric == 'RSI14':
                                values_norm = ((values / 50) - 1)
                                values_norm = np.sign(values_norm) * np.power(np.abs(values_norm), 0.8)
                                
                            elif metric in ['SMA50', 'EMA20']:
                                if 'Close' in window_data.columns:
                                    values_norm = (values / window_data['Close'].values) - 1
                                    values_norm = np.clip(values_norm, -0.15, 0.15)
                                else:
                                    values_norm = np.zeros_like(values)
                                    
                            elif metric == 'MACD':
                                try:
                                    if 'High' in window_data.columns and 'Low' in window_data.columns:
                                        atr = np.nanmean(window_data['High'] - window_data['Low'])
                                        if atr > 0:
                                            values_norm = values / (atr * 5)
                                            values_norm = np.clip(values_norm, -1.5, 1.5)
                                        else:
                                            values_norm = values / np.nanstd(values) if np.nanstd(values) > 0 else np.zeros_like(values)
                                    else:
                                        values_norm = values / np.nanstd(values) if np.nanstd(values) > 0 else np.zeros_like(values)
                                except:
                                    values_norm = np.zeros_like(values)
                                    
                            elif metric == 'log_return':
                                values_norm = np.clip(values, -0.03, 0.03)
                                
                            else:
                                mean_val = np.nanmean(values)
                                std_val = np.nanstd(values) or 1.0
                                values_norm = (values - mean_val) / std_val
                                values_norm = np.clip(values_norm, -3.0, 3.0)
                                
                            # Stocker dans le tableau de caract√©ristiques
                            feature_array[:len(values_norm), i] = values_norm
                    
                    # Traitement des NaN par interpolation
                    for col in range(feature_array.shape[1]):
                        nan_mask = np.isnan(feature_array[:, col])
                        if np.any(nan_mask) and not np.all(nan_mask):
                            valid_indices = np.where(~nan_mask)[0]
                            nan_indices = np.where(nan_mask)[0]
                            
                            for idx in nan_indices:
                                left_indices = valid_indices[valid_indices < idx]
                                right_indices = valid_indices[valid_indices > idx]
                                
                                if len(left_indices) > 0 and len(right_indices) > 0:
                                    left_idx = left_indices[-1]
                                    right_idx = right_indices[0]
                                    ratio = (idx - left_idx) / (right_idx - left_idx)
                                    feature_array[idx, col] = feature_array[left_idx, col] + ratio * (feature_array[right_idx, col] - feature_array[left_idx, col])
                                elif len(left_indices) > 0:
                                    feature_array[idx, col] = feature_array[left_indices[-1], col]
                                elif len(right_indices) > 0:
                                    feature_array[idx, col] = feature_array[right_indices[0], col]
                        
                        # Remplacer les NaN restants par 0
                        feature_array[np.isnan(feature_array[:, col]), col] = 0.0
                    
                    # V√©rification de qualit√© des donn√©es
                    if np.mean(feature_array == 0) > 0.7:
                        continue
                        
                    # V√©rifier la variance dans les s√©ries
                    col_stds = np.std(feature_array, axis=0)
                    if np.mean(col_stds == 0) > 0.5:
                        continue
                    
                    # Ajouter la fen√™tre et les m√©tadonn√©es
                    all_windows.append(feature_array)
                    all_metadata.append({
                        'ticker': ticker,
                        'interval': interval,
                        'window_idx': window_start,
                        'timestamp': window_data.index[0] if len(window_data) > 0 else None,
                        'last_price': float(window_data['Close'].iloc[-1]) if 'Close' in window_data.columns and len(window_data) > 0 else None
                    })
                    
                except Exception as e:
                    continue
                    
        except Exception as e:
            continue
    
    return all_windows, all_metadata

# Construction et sauvegarde de l'index
def build_faiss_index(embeddings, interval, nlist=100):
    """Construit et sauvegarde un index FAISS pour les embeddings d'un intervalle"""
    if embeddings is None or len(embeddings) == 0:
        logger.warning(f"Pas d'embeddings √† indexer pour {interval}")
        return None
    
    logger.info(f"Construction de l'index FAISS pour {interval} avec {len(embeddings)} vecteurs")
    embedding_dim = embeddings.shape[1]
    
    # D√©terminer le nombre optimal de clusters (nlist) en fonction de la taille des donn√©es
    n_vectors = len(embeddings)
    nlist = min(nlist, max(1, n_vectors // 1000))  # 1 cluster pour 1000 vecteurs environ
    
    # Validation
    if n_vectors < nlist:
        nlist = max(1, n_vectors // 10)  # Ajustement si pas assez de donn√©es
    
    logger.info(f"Configuration de l'index: dim={embedding_dim}, nlist={nlist}")
    
    # Pour des indices GPU, utilisez GpuIndexIVFFlat, mais nous restons sur CPU pour la portabilit√©
    quantizer = faiss.IndexFlatL2(embedding_dim)
    index = faiss.IndexIVFFlat(quantizer, embedding_dim, nlist, faiss.METRIC_L2)
    
    # N√©cessite un entra√Ænement sur les donn√©es
    logger.info("Entra√Ænement de l'index...")
    index.train(embeddings)
    
    # Ajouter les vecteurs √† l'index
    logger.info("Ajout des vecteurs √† l'index...")
    index.add(embeddings)
    
    # Nombre de cellules √† explorer lors de la recherche
    index.nprobe = min(10, nlist)
    
    # Sauvegarder l'index
    index_path = os.path.join(Config.INDEX_PATH, f"faiss_index_{interval}.bin")
    faiss.write_index(index, index_path)
    logger.info(f"Index sauvegard√©: {index_path}")
    
    return index

# Chargement des ressources
def load_resources(interval):
    """Charge le mod√®le, les embeddings et l'index pour un intervalle"""
    resources = {}
    
    # Charger le mod√®le
    model_path = os.path.join(Config.MODELS_PATH, f"encoder_{interval}.pth")
    if os.path.exists(model_path):
        logger.info(f"Chargement du mod√®le: {model_path}")
        checkpoint = torch.load(model_path)
        
        # IMPORTANT: Utiliser FinancialTimeSeriesEncoder au lieu d'FinancialTimeSeriesEncoder
        model = FinancialTimeSeriesEncoder(
            input_dim=checkpoint['config']['input_dim'],
            seq_len=checkpoint['config']['seq_len'],
            embedding_dim=checkpoint['config']['embedding_dim']
        )
        model.load_state_dict(checkpoint['model_state_dict'])
        model.eval()
        model.to(Config.DEVICE)
        resources['model'] = model
    else:
        logger.warning(f"Mod√®le non trouv√©: {model_path}")
        resources['model'] = None
    
    # Charger les m√©tadonn√©es
    metadata_path = os.path.join(Config.EMBEDDINGS_PATH, f"metadata_{interval}.pkl")
    if os.path.exists(metadata_path):
        logger.info(f"Chargement des m√©tadonn√©es: {metadata_path}")
        with open(metadata_path, 'rb') as f:
            metadata = pickle.load(f)
        resources['metadata'] = metadata
    else:
        logger.warning(f"M√©tadonn√©es non trouv√©es: {metadata_path}")
        resources['metadata'] = None
    
    # Charger l'index FAISS
    index_path = os.path.join(Config.INDEX_PATH, f"faiss_index_{interval}.bin")
    if os.path.exists(index_path):
        logger.info(f"Chargement de l'index: {index_path}")
        index = faiss.read_index(index_path)
        resources['index'] = index
    else:
        logger.warning(f"Index non trouv√©: {index_path}")
        resources['index'] = None
    
    # Charger les embeddings si n√©cessaire
    if resources['index'] is None:
        embeddings_path = os.path.join(Config.EMBEDDINGS_PATH, f"embeddings_{interval}.h5")
        if os.path.exists(embeddings_path):
            logger.info(f"Chargement des embeddings: {embeddings_path}")
            with h5py.File(embeddings_path, 'r') as f:
                embeddings = f['embeddings'][:]
            resources['embeddings'] = embeddings
        else:
            logger.warning(f"Embeddings non trouv√©s: {embeddings_path}")
            resources['embeddings'] = None
    
    return resources

# Classe principale d'analyse vectorielle
class VectorSimilarityAnalyzer:
    def __init__(self, data_manager, use_gpu=Config.USE_GPU, load_all=True):
        self.data_manager = data_manager
        self.use_gpu = use_gpu
        self.device = Config.DEVICE
        
        # Structures de donn√©es pour stocker les resources par intervalle
        self.models = {}
        self.indexes = {}
        self.metadata = {}
        
        # Chargement des ressources
        if load_all:
            self.load_resources()

        logger.info(f"VectorSimilarityAnalyzer initialis√©: GPU={self.use_gpu}")
    
    def load_resources(self, intervals=None):
        """Charge les mod√®les, index et m√©tadonn√©es pour les intervalles sp√©cifi√©s"""
        intervals = intervals or Config.INTERVALS
        
        for interval in intervals:
            resources = load_resources(interval)
            
            if resources['model'] is not None:
                self.models[interval] = resources['model']
            
            if resources['index'] is not None:
                self.indexes[interval] = resources['index']
            
            if resources['metadata'] is not None:
                self.metadata[interval] = resources['metadata']
    
    def ensure_resources(self, interval):
        """V√©rifie que les ressources pour un intervalle sont charg√©es"""
        if interval not in Config.INTERVALS:
            logger.warning(f"Intervalle non support√©: {interval}")
            return False
        
        # Si les ressources ne sont pas charg√©es, essayer de les charger
        if (interval not in self.models or 
            interval not in self.indexes or 
            interval not in self.metadata):
            logger.info(f"Chargement des ressources pour {interval}")
            self.load_resources([interval])
        
        # V√©rifier que tout est charg√©
        if (interval not in self.models or 
            interval not in self.indexes or 
            interval not in self.metadata):
            logger.warning(f"Ressources manquantes pour {interval}")
            return False
        
        return True
        
    def load_optimized_weights(self, weights_dir=None):
        """
        Charge les poids optimis√©s depuis un fichier JSON dans le dossier sp√©cifi√©.
        """
        import json
        import os

        if weights_dir is None:
            # Par d√©faut, chercher dans le dossier 'optimization_results/custom_weights'
            base_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
            weights_dir = os.path.join(base_dir, "optimization_results", "custom_weights")

        if not os.path.exists(weights_dir):
            print(f"Aucun dossier de poids optimis√©s trouv√©: {weights_dir}")
            return False

        # Chercher le fichier le plus r√©cent pour chaque intervalle
        for interval in self.models.keys():
            pattern = os.path.join(weights_dir, f"{interval}_weights_latest.json")
            files = glob.glob(pattern)
            if not files:
                # Essayer de trouver un fichier par date
                files = glob.glob(os.path.join(weights_dir, f"{interval}_weights_*.json"))
            if not files:
                print(f"Aucun poids optimis√© trouv√© pour {interval} dans {weights_dir}")
                continue
            # Prendre le plus r√©cent
            weights_file = max(files, key=os.path.getmtime)
            try:
                with open(weights_file, "r") as f:
                    weights = json.load(f)
                # Stocker ou appliquer les poids (√† adapter selon votre logique)
                if not hasattr(self, "optimized_weights"):
                    self.optimized_weights = {}
                self.optimized_weights[interval] = weights
                print(f"Poids optimis√©s charg√©s pour {interval} depuis {weights_file}")
            except Exception as e:
                print(f"Erreur lors du chargement des poids pour {interval}: {e}")
                continue
        return True
    
    def encode_window(self, ticker, interval, window_idx=None):
        """Encode une fen√™tre de prix en vecteur d'embedding"""
        # V√©rifier que le mod√®le est charg√©
        if interval not in self.models:
            logger.warning(f"Mod√®le non charg√© pour {interval}")
            return None
        
        # AJOUTER - Cache de fen√™tres en m√©moire
        cache_key = f"{ticker}_{interval}_{window_idx}"
        if hasattr(self, '_window_cache') and cache_key in self._window_cache:
            return self._window_cache[cache_key]
        
        # Charger les donn√©es du ticker
        ticker_data = self.data_manager._load_ticker_data(ticker, interval)
        if ticker_data is None or len(ticker_data) < Config.get_window_size(interval):
            logger.warning(f"Donn√©es insuffisantes pour {ticker} ({interval})")
            return None
        
        # Si window_idx n'est pas sp√©cifi√©, utiliser la derni√®re fen√™tre
        if window_idx is None:
            window_idx = len(ticker_data) - Config.get_window_size(interval)
        
        if window_idx < 0 or window_idx + Config.get_window_size(interval) > len(ticker_data):
            logger.warning(f"Index de fen√™tre invalide: {window_idx}")
            return None
        
        # Extraire la fen√™tre
        window_data = ticker_data.iloc[window_idx:window_idx+Config.get_window_size(interval)]
        
        # Normaliser les donn√©es
        available_metrics = [m for m in Config.METRICS if m in window_data.columns]
        normalized_data = enhanced_financial_preprocessing(window_data[available_metrics], normalize=True)
        
        # Cr√©er un tableau de caract√©ristiques
        feature_array = np.zeros((Config.get_window_size(interval), len(Config.METRICS)))
        
        for i, metric in enumerate(Config.METRICS):
            norm_col = f"{metric}_norm" if metric != 'log_return' else 'log_return_norm'
            if norm_col in normalized_data.columns:
                values = normalized_data[norm_col].values
                feature_array[:len(values), i] = values
        
        # Remplir les NaN par 0
        feature_array = np.nan_to_num(feature_array, nan=0.0)
        
        # MODIFIER - Traiter par batchs pour efficacit√©
        with torch.no_grad():
            # Convertir en tensor et encoder
            tensor = torch.tensor(feature_array, dtype=torch.float32).unsqueeze(0).to(self.device)
            embedding = self.models[interval].encode(tensor).cpu().numpy()[0]
        
        # AJOUTER - Stocker dans le cache (max 1000 fen√™tres)
        if not hasattr(self, '_window_cache'):
            self._window_cache = {}
        if len(self._window_cache) > 1000:
            self._window_cache.clear()  # Simple cache LRU
        self._window_cache[cache_key] = embedding
        
        return embedding
    
    def find_similar_patterns(self, ticker, interval, window_idx=None, n_similar=None, analysis_date=None, use_dtw=False, window_size=None):
        """Finds similar patterns using vector-based similarity search."""
        if n_similar is None:
            n_similar = Config.N_SIMILAR
        if window_size is None:
            window_size = Config.get_window_size(interval)
        start_time = time.time()
        
        # V√©rifier que les ressources sont charg√©es
        if not self.ensure_resources(interval):
            return None
        
        # Si analysis_date est sp√©cifi√©, trouver l'index correspondant
        if analysis_date is not None:
            ticker_data = self.data_manager._load_ticker_data(ticker, interval)
            if ticker_data is None:
                logger.warning(f"Donn√©es non trouv√©es pour {ticker}")
                return None
            
            # Trouver l'index correspondant √† la date
            try:
                if isinstance(ticker_data.index, pd.DatetimeIndex):
                    window_idx = ticker_data.index.get_loc(analysis_date) - Config.get_window_size(interval) + 1
                else:
                    window_idx = np.where(ticker_data.index == analysis_date)[0][0] - Config.get_window_size(interval) + 1
                
                window_idx = max(0, window_idx)
            except:
                logger.warning(f"Date non trouv√©e: {analysis_date}")
                window_idx = None
        
        # Encoder la fen√™tre cible
        query_embedding = self.encode_window(ticker, interval, window_idx)
        if query_embedding is None:
            logger.warning(f"Impossible d'encoder la fen√™tre pour {ticker}")
            return None
        
        # Convertir en format pour FAISS
        query_vector = np.array([query_embedding], dtype=np.float32)
        
        # Rechercher les vecteurs similaires
        #logger.info(f"Recherche de {n_similar} patterns similaires pour {ticker}")
        distances, indices = self.indexes[interval].search(query_vector, n_similar*2)  # R√©cup√©rer plus pour filtrage
        
        # Construire le r√©sultat
        similar_patterns = []
        
        # Filtrage des r√©sultats (√©viter les duplicatas proches)
        used_tickers = set()
        
        for i, idx in enumerate(indices[0]):
            if idx >= len(self.metadata[interval]):
                continue
                
            pattern_info = self.metadata[interval][idx]
            similarity = 1.0 / (1.0 + distances[0][i])  # Convertir distance en similarit√©
            
            # Filtrer pour √©viter trop de patterns du m√™me ticker
            ticker_count = sum(1 for p in similar_patterns if p['ticker'] == pattern_info['ticker'])
            if ticker_count >= 3 and pattern_info['ticker'] != ticker:
                continue
            
            similar_patterns.append({
                'ticker': pattern_info['ticker'],
                'window_idx': pattern_info['window_idx'],
                'similarity': float(similarity),
                'timestamp': pattern_info['timestamp'] if 'timestamp' in pattern_info else None
            })
            
            used_tickers.add(pattern_info['ticker'])
            
            # Stopper quand on a suffisamment de patterns
            if len(similar_patterns) >= n_similar:
                break
        
        execution_time = time.time() - start_time
        #logger.info(f"Recherche compl√©t√©e en {execution_time:.3f}s - {len(similar_patterns)} patterns trouv√©s")
        
        # Si demand√© et possible, affiner avec DTW
        if use_dtw and hasattr(self.data_manager, 'create_tensor_window'):
            logger.info("Affinement des r√©sultats avec DTW...")
            try:
                dtw_analyzer = OptimizedSimilarityAnalyzer(self.data_manager, use_gpu=self.use_gpu)
                reference_tensor = self.data_manager.create_tensor_window(ticker, interval)
                
                if reference_tensor is not None and window_idx is not None:
                    # Utiliser un sous-ensemble pour DTW (pour des raisons de performance)
                    dtw_patterns = similar_patterns[:min(len(similar_patterns), 10)]
                    refined_patterns = []
                    
                    for pattern in dtw_patterns:
                        pattern_ticker = pattern['ticker']
                        pattern_window_idx = pattern['window_idx']
                        
                        compare_tensor = self.data_manager.create_tensor_window(pattern_ticker, interval)
                        if compare_tensor is not None and pattern_window_idx < compare_tensor.shape[0]:
                            # Calculer similarit√© DTW
                            dtw_similarity = dtw_analyzer.calculate_similarity(
                                reference_tensor[window_idx], compare_tensor[pattern_window_idx]
                            )
                            pattern['similarity'] = dtw_similarity  # Mettre √† jour la similarit√©
                        
                        refined_patterns.append(pattern)
                    
                    # Mettre √† jour les patterns DTW et r√©ordonner
                    similar_patterns[:len(refined_patterns)] = refined_patterns
                    similar_patterns.sort(key=lambda x: x['similarity'], reverse=True)
            except Exception as e:
                logger.error(f"Erreur lors de l'affinement DTW: {str(e)}")
        
        return {
            'ticker': ticker,
            'interval': interval,
            'window_idx': window_idx,
            'analysis_date': analysis_date,
            'similar_patterns': similar_patterns,
            'execution_time_s': execution_time
        }
    
    def calculate_expected_return(self, similarity_results, forecast_horizon=None):
        """Calculates expected return with adaptive outlier management."""
        if forecast_horizon is None:
            forecast_horizon = Config.FORECAST_HORIZON
        if similarity_results is None or 'similar_patterns' not in similarity_results:
            return None
        
        # Extraire les informations
        ticker = similarity_results['ticker']
        interval = similarity_results['interval']
        patterns = similarity_results['similar_patterns']
        
        # Accumulateurs
        all_returns = []
        weighted_returns = []
        
        # Analyser chaque pattern similaire
        for pattern in patterns:
            pattern_ticker = pattern['ticker']
            window_idx = pattern['window_idx']
            similarity = pattern['similarity']
            
            try:
                # Charger les donn√©es futures
                pattern_data = self.data_manager.load_ticker_chunk(
                    pattern_ticker, interval, 
                    start_idx=window_idx, 
                    chunk_size=forecast_horizon+1
                )
                
                if pattern_data is None or len(pattern_data) < 2:
                    continue
                
                # Calculer le rendement futur
                start_price = pattern_data['Close'].iloc[0]
                if start_price <= 0:
                    continue
                
                # Dernier prix disponible
                last_idx = min(forecast_horizon, len(pattern_data)-1)
                end_price = pattern_data['Close'].iloc[last_idx]
                
                # Rendement relatif
                pattern_return = (end_price / start_price) - 1
                
                # Stocker le rendement et la similarit√©
                all_returns.append(pattern_return)
                
                # Pond√©rer par similarit√© au carr√©
                weight = similarity ** 2
                weighted_returns.append((pattern_return, weight))
                
            except Exception as e:
                logger.warning(f"Erreur lors du calcul du rendement pour {pattern_ticker}: {str(e)}")
                continue
        
        # V√©rifier qu'il y a suffisamment de donn√©es
        if not weighted_returns:
            return None
        
        # D√©tection d'outliers adaptative
        all_returns = np.array([r for r, _ in weighted_returns])
        
        # Calculer IQR
        q1 = np.percentile(all_returns, 25)
        q3 = np.percentile(all_returns, 75)
        iqr = q3 - q1
        
        # D√©finir les seuils d'outliers
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Filtrer les outliers
        filtered_returns = []
        for ret, weight in weighted_returns:
            if lower_bound <= ret <= upper_bound:
                filtered_returns.append((ret, weight))
        
        # Si le filtrage a supprim√© trop de points, utiliser les donn√©es originales avec des limites plus souples
        if len(filtered_returns) < max(3, len(weighted_returns) * 0.3):
            # Utiliser des limites plus conservatrices
            lower_bound = q1 - 2.5 * iqr
            upper_bound = q3 + 2.5 * iqr
            
            filtered_returns = []
            for ret, weight in weighted_returns:
                if lower_bound <= ret <= upper_bound:
                    filtered_returns.append((ret, weight))
        
        # Si toujours trop peu de points, utiliser toutes les donn√©es
        if len(filtered_returns) < 3:
            filtered_returns = weighted_returns
        
        # Calculer la moyenne pond√©r√©e
        total_weight = sum(weight for _, weight in filtered_returns)
        if total_weight <= 0:
            return None
            
        expected_return = sum(ret * weight for ret, weight in filtered_returns) / total_weight
        
        return expected_return
    
    def visualize_embeddings(self, interval, n_samples=1000, perplexity=30):
        """Visualise les embeddings en 2D √† l'aide de t-SNE"""
        if interval not in self.metadata:
            logger.warning(f"M√©tadonn√©es non charg√©es pour {interval}")
            return
        
        # Charger les embeddings
        embeddings_path = os.path.join(Config.EMBEDDINGS_PATH, f"embeddings_{interval}.h5")
        if not os.path.exists(embeddings_path):
            logger.warning(f"Embeddings non trouv√©s: {embeddings_path}")
            return
        
        with h5py.File(embeddings_path, 'r') as f:
            all_embeddings = f['embeddings'][:]
        
        # √âchantillonner al√©atoirement si trop de donn√©es
        if len(all_embeddings) > n_samples:
            indices = np.random.choice(len(all_embeddings), n_samples, replace=False)
            embeddings_sample = all_embeddings[indices]
            metadata_sample = [self.metadata[interval][i] for i in indices]
        else:
            embeddings_sample = all_embeddings
            metadata_sample = self.metadata[interval]
        
        # Appliquer t-SNE
        logger.info(f"Application de t-SNE sur {len(embeddings_sample)} embeddings...")
        tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=1000, random_state=42)
        embeddings_2d = tsne.fit_transform(embeddings_sample)
        
        # Visualisation
        plt.figure(figsize=(14, 10))
        
        # Grouper par ticker pour la coloration
        ticker_groups = {}
        for i, meta in enumerate(metadata_sample):
            ticker = meta['ticker']
            if ticker not in ticker_groups:
                ticker_groups[ticker] = []
            ticker_groups[ticker].append(i)
        
        # Limiter le nombre de tickers affich√©s pour la clart√©
        top_tickers = sorted(ticker_groups.keys(), key=lambda t: len(ticker_groups[t]), reverse=True)[:10]
        
        # Tracer les points par ticker
        for i, ticker in enumerate(top_tickers):
            indices = ticker_groups[ticker]
            plt.scatter(
                embeddings_2d[indices, 0], 
                embeddings_2d[indices, 1],
                label=ticker,
                s=30,
                alpha=0.7
            )
        
        # Autres tickers
        other_indices = [i for i, meta in enumerate(metadata_sample) if meta['ticker'] not in top_tickers]
        if other_indices:
            plt.scatter(
                embeddings_2d[other_indices, 0],
                embeddings_2d[other_indices, 1],
                label='Others',
                s=10,
                alpha=0.2,
                color='gray'
            )
        
        plt.title(f'Visualisation des embeddings t-SNE - {interval}')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Sauvegarder et afficher
        plot_path = os.path.join(Config.EMBEDDINGS_PATH, f"embeddings_viz_{interval}.png")
        plt.savefig(plot_path)
        plt.tight_layout()
        plt.show()
        
        logger.info(f"Visualisation sauvegard√©e: {plot_path}")

# Classe VectorMarketAnalyzer pour remplacer MarketAnalyzer
class VectorMarketAnalyzer:
    def __init__(self, base_path=Config.BASE_PATH, config=None, use_gpu=Config.USE_GPU):
        # Configuration
        self.config = {
            'intervals': Config.INTERVALS,
            'forecast_horizon': Config.FORECAST_HORIZON,
            'historical_bars': Config.HISTORICAL_BARS,
            'n_similar': Config.N_SIMILAR,
            'min_similarity': Config.MIN_SIMILARITY,
            'use_gpu': use_gpu,
            'cache_size': Config.CACHE_SIZE
        }
        
        # Mettre √† jour la configuration si fournie
        if config is not None:
            self.config.update(config)
        
        # Initialiser les composants
        self.data_manager = StreamingDataManager(
            base_path, 
            chunk_size=Config.CHUNK_SIZE,
            cache_capacity_mb=Config.CACHE_CAPACITY_MB,
            use_gpu=self.config['use_gpu']
        )
        
        # Initialiser l'analyseur vectoriel
        self.vector_analyzer = VectorSimilarityAnalyzer(
            self.data_manager,
            use_gpu=self.config['use_gpu'],
            load_all=self.config.get('load_all_resources', False)  # Contr√¥l√© par config
        )
        
        # Initialiser l'analyseur DTW classique pour les fonctionnalit√©s qui le n√©cessitent
        self.similarity_analyzer = OptimizedSimilarityAnalyzer(
            self.data_manager,
            use_gpu=self.config['use_gpu']
        )
        
        logger.info(f"VectorMarketAnalyzer initialis√© avec {len(self.config['intervals'])} intervalles")
    
    def analyze_ticker(self, ticker, interval=None, analysis_date=None, use_dtw=False):
        """Analyse un ticker et trouve des patterns similaires"""
        if interval is None:
            interval = self.config['intervals'][0]
        
        logger.info(f"Analyse de {ticker} sur {interval}")
        
        # Chercher les patterns similaires avec l'approche vectorielle
        similarity_results = self.vector_analyzer.find_similar_patterns(
            ticker, 
            interval, 
            n_similar=self.config['n_similar'],
            analysis_date=analysis_date,
            use_dtw=use_dtw
        )
        
        if similarity_results is None:
            logger.warning(f"Aucun pattern similaire trouv√© pour {ticker}")
            return None
        
        # Calculer le rendement attendu
        expected_return = self.vector_analyzer.calculate_expected_return(
            similarity_results, 
            forecast_horizon=self.config['forecast_horizon']
        )
        
        # Ajouter le rendement attendu au r√©sultat
        result = {
            'ticker': ticker,
            'interval': interval,
            'analysis_date': similarity_results.get('analysis_date'),
            'expected_return': expected_return,
            'similarity_results': similarity_results
        }
        
        # Utiliser les fonctionnalit√©s avanc√©es existantes si disponibles
        if hasattr(self.similarity_analyzer, 'flasc_cluster_trajectories'):
            # Charger les donn√©es compl√®tes du ticker
            ticker_data = self.data_manager._load_ticker_data(ticker, interval)
            
            if ticker_data is not None and similarity_results.get('analysis_date') in ticker_data.index:
                try:
                    # Effectuer le clustering FLASC
                    cluster_results = self.similarity_analyzer.flasc_cluster_trajectories(
                        similarity_results, 
                        ticker_data,
                        analysis_date=similarity_results.get('analysis_date'),
                        n_forecast=self.config['forecast_horizon'],
                        historical_bars=self.config['historical_bars'],
                        use_full_history=True
                    )
                    
                    # Ajouter les r√©sultats de clustering
                    result['cluster_results'] = cluster_results
                    
                    # Calculer le rendement FLASC
                    if cluster_results and cluster_results['clusters']:
                        largest_cluster = max(cluster_results['clusters'].values(), key=lambda c: c['size'])
                        if 'terminal_change' in largest_cluster:
                            result['flasc_return'] = largest_cluster['terminal_change'] / 100
                except Exception as e:
                    logger.error(f"Erreur lors du clustering FLASC: {str(e)}")
        
        return result
    
    def plot_analysis_results(self, result, show_clusters=True):
        """Visualise les r√©sultats d'analyse"""
        if result is None:
            logger.warning("Aucun r√©sultat √† visualiser")
            return
        
        ticker = result['ticker']
        interval = result['interval']
        expected_return = result.get('expected_return')
        flasc_return = result.get('flasc_return')
        
        # Afficher les r√©sultats
        logger.info(f"R√©sultats pour {ticker} ({interval}):")
        if expected_return is not None:
            logger.info(f"Rendement attendu: {expected_return:.2%}")
        if flasc_return is not None:
            logger.info(f"Rendement FLASC: {flasc_return:.2%}")
        
        # Si des r√©sultats de clustering sont disponibles et que la fonction de plot existe
        if 'cluster_results' in result and show_clusters and hasattr(self.similarity_analyzer, 'plot_flasc_trajectories'):
            # Charger les donn√©es du ticker
            ticker_data = self.data_manager._load_ticker_data(ticker, interval)
            
            if ticker_data is not None:
                try:
                    # Tracer les trajectoires FLASC
                    self.similarity_analyzer.plot_flasc_trajectories(
                        result['similarity_results'],
                        ticker_data,
                        cluster_results=result['cluster_results'],
                        n_last=self.config['historical_bars'],
                        n_forecast=self.config['forecast_horizon'],
                        analysis_date=result['analysis_date']
                    )
                    
                    # Tracer la distribution des r√©sultats
                    self.similarity_analyzer.plot_weighted_outcome_distribution(
                        result['similarity_results'],
                        ticker_data,
                        n_forecast=self.config['forecast_horizon'],
                        analysis_date=result['analysis_date']
                    )
                except Exception as e:
                    logger.error(f"Erreur lors de la visualisation: {str(e)}")                
    
    def scan_market(self, interval=None, min_return=0.05, max_tickers=None):
        """Scanne le march√© pour trouver des opportunit√©s"""
        if interval is None:
            interval = self.config['intervals'][0]
        
        # R√©cup√©rer les tickers disponibles
        tickers = self.data_manager.get_available_tickers(interval)
        if max_tickers is not None:
            tickers = tickers[:max_tickers]
        
        logger.info(f"Scan de {len(tickers)} tickers sur {interval}")
        
        # R√©sultats
        opportunities = []
        
        # Traitement par lots
        with ThreadPoolExecutor(max_workers=Config.NUM_WORKERS) as executor:  # Utilisez ThreadPoolExecutor au lieu de ProcessPoolExecutor
            batch_size = 100
            for i in range(0, len(tickers), batch_size):
                batch_tickers = tickers[i:i+batch_size]
                logger.info(f"Traitement du lot {i//batch_size + 1}/{(len(tickers) + batch_size - 1)//batch_size}")
                
                # Cr√©er des tuples avec self et les param√®tres pour chaque ticker
                ticker_params = [(self, ticker, interval, min_return) for ticker in batch_tickers]
                
                # Ex√©cuter le traitement en parall√®le
                batch_results = list(executor.map(_process_ticker_for_scan, ticker_params))
                
                # Filtrer les r√©sultats valides
                valid_results = [r for r in batch_results if r is not None]
                opportunities.extend(valid_results)
                
                # Mise √† jour interm√©diaire
                logger.info(f"Opportunit√©s trouv√©es jusqu'√† pr√©sent: {len(opportunities)}")
        
        # Trier par rendement absolu attendu
        opportunities.sort(key=lambda x: abs(x['expected_return']), reverse=True)
        
        # Statistiques
        n_bullish = sum(1 for o in opportunities if o['direction'] == 'bullish')
        n_bearish = sum(1 for o in opportunities if o['direction'] == 'bearish')
        
        logger.info(f"Scan termin√©. {len(opportunities)} opportunit√©s trouv√©es sur {len(tickers)} tickers.")
        logger.info(f"Haussier: {n_bullish}, Baissier: {n_bearish}")
        
        return opportunities

    def update_index(self, interval=None, incremental=True):
        """Met √† jour l'index pour un intervalle sp√©cifique"""
        interval = interval or self.config['intervals'][0]
        
        # V√©rifier si le mod√®le existe
        model_path = os.path.join(Config.MODELS_PATH, f"encoder_{interval}.pth")
        if not os.path.exists(model_path):
            logger.warning(f"Mod√®le non trouv√© pour {interval}. Entra√Ænement n√©cessaire.")
            return False
        
        # Charger le mod√®le
        model = None
        try:
            checkpoint = torch.load(model_path)
            model = FinancialTimeSeriesEncoder(
                input_dim=checkpoint['config']['input_dim'],
                seq_len=checkpoint['config']['seq_len'],
                embedding_dim=checkpoint['config']['embedding_dim']
            )
            model.load_state_dict(checkpoint['model_state_dict'])
            model.eval()
            model.to(Config.DEVICE)
        except Exception as e:
            logger.error(f"Erreur lors du chargement du mod√®le: {str(e)}")
            return False
        
        # Mise √† jour incr√©mentale ou compl√®te?
        if incremental:
            # Charger les m√©tadonn√©es existantes
            metadata_path = os.path.join(Config.EMBEDDINGS_PATH, f"metadata_{interval}.pkl")
            if os.path.exists(metadata_path):
                try:
                    with open(metadata_path, 'rb') as f:
                        existing_metadata = pickle.load(f)
                        
                    # Identifier les tickers d√©j√† index√©s
                    existing_tickers = set(meta['ticker'] for meta in existing_metadata)
                    logger.info(f"{len(existing_tickers)} tickers d√©j√† index√©s pour {interval}")
                except Exception as e:
                    logger.error(f"Erreur lors du chargement des m√©tadonn√©es: {str(e)}")
                    existing_tickers = set()
            else:
                existing_tickers = set()
            
            # R√©cup√©rer tous les tickers disponibles
            all_tickers = set(self.data_manager.get_available_tickers(interval))
            
            # Identifier les nouveaux tickers
            new_tickers = all_tickers - existing_tickers
            logger.info(f"{len(new_tickers)} nouveaux tickers √† indexer pour {interval}")
            
            if not new_tickers:
                logger.info(f"Aucun nouveau ticker √† indexer pour {interval}")
                return True
            
            # G√©n√©rer les embeddings seulement pour les nouveaux tickers
            # (Ceci est une simplification - une impl√©mentation compl√®te g√©rerait √©galement les mises √† jour)
            tickers_to_process = list(new_tickers)
        else:
            # Traiter tous les tickers
            tickers_to_process = self.data_manager.get_available_tickers(interval)
            logger.info(f"Reindexation compl√®te de {len(tickers_to_process)} tickers pour {interval}")
        
        # G√©n√©rer les embeddings
        embedded_count = 0
        all_embeddings = []
        all_metadata = []
        
        # Traitement par lots
        for i in range(0, len(tickers_to_process), 100):
            batch_tickers = tickers_to_process[i:i+100]
            logger.info(f"Traitement du lot {i//100 + 1}/{(len(tickers_to_process) + 99)//100}")
            
            for ticker in tqdm(batch_tickers, desc=f"Embeddings pour {interval}"):
                # Charger les donn√©es du ticker
                ticker_data = self.data_manager._load_ticker_data(ticker, interval)
                if ticker_data is None or len(ticker_data) < Config.get_window_size(interval):
                    continue
                
                # G√©n√©rer les embeddings pour toutes les fen√™tres
                ticker_embeddings = []
                ticker_metadata = []
                
                for window_start in range(0, len(ticker_data) - Config.get_window_size(interval) + 1, 5):  # Pas de 5 pour r√©duire le volume
                    embedding = self.vector_analyzer.encode_window(ticker, interval, window_start)
                    if embedding is not None:
                        ticker_embeddings.append(embedding)
                        ticker_metadata.append({
                            'ticker': ticker,
                            'interval': interval,
                            'window_idx': window_start,
                            'timestamp': ticker_data.index[window_start],
                            'last_price': float(ticker_data['Close'].iloc[-1]) if 'Close' in ticker_data.columns else None
                        })
                
                if ticker_embeddings:
                    all_embeddings.extend(ticker_embeddings)
                    all_metadata.extend(ticker_metadata)
                    embedded_count += 1
                
                # Lib√©rer la m√©moire
                if Config.USE_GPU:
                    torch.cuda.empty_cache()
        
        logger.info(f"{embedded_count}/{len(tickers_to_process)} tickers incorpor√©s dans l'index")
        
        if not all_embeddings:
            logger.warning(f"Aucun embedding g√©n√©r√© pour {interval}")
            return False
        
        # Si mode incr√©mental et des m√©tadonn√©es existantes, fusionner
        if incremental and 'existing_metadata' in locals():
            # Charger les embeddings existants
            embeddings_path = os.path.join(Config.EMBEDDINGS_PATH, f"embeddings_{interval}.h5")
            if os.path.exists(embeddings_path):
                try:
                    with h5py.File(embeddings_path, 'r') as f:
                        existing_embeddings = f['embeddings'][:]
                    
                    # Fusionner
                    all_embeddings = np.vstack([existing_embeddings, np.array(all_embeddings)])
                    all_metadata = existing_metadata + all_metadata
                    
                    logger.info(f"Fusion de {len(existing_embeddings)} embeddings existants et {len(all_embeddings) - len(existing_embeddings)} nouveaux embeddings")
                except Exception as e:
                    logger.error(f"Erreur lors de la fusion des embeddings: {str(e)}")
                    # Continuer avec seulement les nouveaux embeddings
        else:
            # Convertir en array pour FAISS
            all_embeddings = np.array(all_embeddings)
        
        # Sauvegarder les embeddings et m√©tadonn√©es
        embeddings_path = os.path.join(Config.EMBEDDINGS_PATH, f"embeddings_{interval}.h5")
        metadata_path = os.path.join(Config.EMBEDDINGS_PATH, f"metadata_{interval}.pkl")
        
        # Utiliser HDF5 pour les embeddings
        with h5py.File(embeddings_path, 'w') as f:
            f.create_dataset('embeddings', data=embeddings_array, 
                            compression='gzip', compression_opts=1) 
        
        # Pickle pour les m√©tadonn√©es
        with open(metadata_path, 'wb') as f:
            pickle.dump(all_metadata, f)
        
        logger.info(f"Sauvegarde de {len(all_metadata)} embeddings pour {interval}")
        
        # Construire l'index FAISS
        logger.info(f"Construction de l'index FAISS pour {interval}")
        nlist = min(100, max(1, len(all_embeddings) // 1000))  # 1 cluster pour 1000 vecteurs environ
        index = build_faiss_index(all_embeddings, interval, nlist=nlist)
        
        # Recharger les ressources
        self.vector_analyzer.load_resources([interval])
        
        return True


def _process_ticker_for_scan(ticker_info_tuple):
    """Fonction d√©tach√©e pour traiter un ticker durant le scan."""
    self_obj, ticker, interval, min_return = ticker_info_tuple
    try:
        # Patterns similaires
        similarity_results = self_obj.vector_analyzer.find_similar_patterns(
            ticker, interval, n_similar=Config.N_SIMILAR
        )
        
        if similarity_results is None:
            return None
        
        # Rendement attendu
        expected_return = self_obj.vector_analyzer.calculate_expected_return(similarity_results)
        
        if expected_return is None or abs(expected_return) < min_return:
            return None
        
        # Donn√©es r√©centes
        recent_data = self_obj.data_manager.load_ticker_chunk(ticker, interval, chunk_size=10)
        
        if recent_data is None:
            return None
        
        # Opportunit√© d√©tect√©e
        return {
            'ticker': ticker,
            'interval': interval,
            'expected_return': float(expected_return),
            'direction': 'bullish' if expected_return > 0 else 'bearish',
            'last_price': float(recent_data['Close'].iloc[-1]),
            'n_patterns': len(similarity_results['similar_patterns']),
            'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
    except Exception as e:
        logger.error(f"Erreur lors du traitement de {ticker}: {str(e)}")
        return None
        

# REMPLACEMENT COMPLET de VectorWeightOptimizer.__init__ et optimize_interval

class VectorWeightOptimizer:
    def __init__(self, vector_analyzer, intervals=None, output_dir=None, verbose_level=logging.INFO):
        self.vector_analyzer = vector_analyzer
        self.data_manager = vector_analyzer.data_manager
        self.intervals = intervals or Config.INTERVALS
        
        if output_dir is None:
            output_dir = r"C:\Users\Alex\Python\Finance\Strat002. SP_DTW\optimization_results"
        self.output_dir = output_dir
        
        self.verbose_level = verbose_level
        self.n_trials = 0
        self.optimization_start_time = 0
        self.best_weights = {}
        self.optimization_history = {}
        
        # CORRECTION CRITIQUE : Optimiser les m√©triques financi√®res SEULEMENT
        self.param_space = {}
        
        # Les vraies m√©triques financi√®res de Config.METRICS
        financial_metrics = ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI14', 'MACD', 'SMA50', 'EMA20', 'log_return']
        
        for metric in financial_metrics:
            self.param_space[f'{metric}_weight'] = (0.1, 2.0)
        
        # Param√®tres de d√©croissance temporelle
        self.param_space.update({
            'decay_factor': (0.7, 0.99), 
            'decay_type_idx': (0, 3)
        })
        
        print(f"üéØ Optimisation configur√©e pour {len(financial_metrics)} m√©triques financi√®res")
        print(f"üìä M√©triques: {financial_metrics}")
        
        # Poids d'√©valuation (DIFF√âRENTS des poids de m√©triques)
        self.evaluation_weights = {
            'mape': 0.2,        # Pr√©cision la plus importante
            'direction': 0.4,   # Direction importante
            'calibration': 0.25, # Coh√©rence
            'pattern_var': 0.15  # Variance patterns
        }
        
        os.makedirs(output_dir, exist_ok=True)
        os.makedirs(os.path.join(output_dir, "history"), exist_ok=True)
        os.makedirs(os.path.join(output_dir, "plots"), exist_ok=True)

        
    def _evaluate_weights(self, weights, interval, tickers, n_samples=Config.OPT_SAMPLE_SIZE):
        """√âvalue les poids des m√©triques financi√®res en utilisant FAISS"""
        eval_start = time.time()
        
        # Extraire et normaliser les poids des m√©triques
        financial_metrics = ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI14', 'MACD', 'SMA50', 'EMA20', 'log_return']
        raw_weights = {}
        
        for metric in financial_metrics:
            weight_key = f'{metric}_weight'
            raw_weights[metric] = weights.get(weight_key, 1.0)
        
        # NORMALISATION des poids (comme log returns)
        weight_values = np.array(list(raw_weights.values()))
        mean_weight = np.mean(weight_values)
        std_weight = np.std(weight_values)
        
        normalized_weights = {}
        for metric in financial_metrics:
            if std_weight > 0:
                # Normalisation z-score puis transformation vers [0.1, 2.0]
                z_score = (raw_weights[metric] - mean_weight) / std_weight
                normalized_weight = 1.0 + 0.5 * z_score  # Centr√© sur 1.0 avec variation ¬±0.5
                normalized_weight = max(0.1, min(2.0, normalized_weight))  # Born√©
            else:
                normalized_weight = 1.0
            normalized_weights[metric] = normalized_weight
        
        # Param√®tres temporels
        decay_factor = weights.get('decay_factor', 0.9)
        decay_type = weights.get('decay_type', 'exponential')
        
        print(f"üîß Poids normalis√©s: Close={normalized_weights['Close']:.2f}, Volume={normalized_weights['Volume']:.2f}, RSI={normalized_weights['RSI14']:.2f}")
        
        # √âchantillonner les tickers
        sample_tickers = random.sample(tickers, min(n_samples, len(tickers)))
        print(f"üîç Testing {len(sample_tickers)} tickers avec FAISS+poids normalis√©s")
        
        # M√©triques d'√©valuation
        metrics = {'mape': [], 'direction': [], 'calibration': [], 'pattern_var': []}
        processed_count = 0
        
        for i, ticker in enumerate(sample_tickers):
            try:
                if i % 25 == 0:
                    print(f"   Processing {i}/{len(sample_tickers)} tickers...")
                
                # √âTAPE 1: Utiliser VectorSimilarityAnalyzer avec FAISS
                similarity_results = self.vector_analyzer.find_similar_patterns(
                    ticker, interval, 
                    n_similar=Config.N_SIMILAR  # Utiliser la vraie valeur, pas tronqu√©e
                )
                
                if not similarity_results or not similarity_results.get('similar_patterns'):
                    continue
                
                # √âTAPE 2: Cr√©er un OptimizedSimilarityAnalyzer temporaire avec les nouveaux poids
                temp_analyzer = OptimizedSimilarityAnalyzer(self.data_manager, use_gpu=Config.USE_GPU)
                
                # Appliquer les poids normalis√©s
                temp_analyzer.metric_weights = {
                    f'{metric}_norm': weight for metric, weight in normalized_weights.items()
                }
                temp_analyzer.decay_factor = decay_factor
                temp_analyzer.decay_type = decay_type
                
                # √âTAPE 3: Calculer le rendement attendu avec les poids appliqu√©s
                expected_return = temp_analyzer.calculate_expected_return(
                    similarity_results, 
                    forecast_horizon=Config.FORECAST_HORIZON
                )
                
                if expected_return is None:
                    continue
                
                # √âTAPE 4: Validation sur donn√©es historiques
                ticker_data = self.data_manager._load_ticker_data(ticker, interval)
                if ticker_data is None or len(ticker_data) < Config.FORECAST_HORIZON + 50:
                    continue
                
                # Point de test al√©atoire
                max_idx = len(ticker_data) - Config.FORECAST_HORIZON - 1
                min_idx = Config.get_window_size(interval) + 10
                if max_idx <= min_idx:
                    continue
                    
                test_idx = random.randint(min_idx, max_idx)
                
                # Rendement r√©el
                start_price = ticker_data['Close'].iloc[test_idx]
                end_price = ticker_data['Close'].iloc[test_idx + Config.FORECAST_HORIZON]
                if start_price <= 0 or end_price <= 0:
                    continue
                    
                actual_return = (end_price / start_price) - 1
                
                # √âTAPE 5: Calculer m√©triques
                # 1. MAPE
                mape_error = abs((actual_return - expected_return) / max(0.001, abs(actual_return)))
                mape_score = max(0.0, 1.0 - min(1.0, mape_error))
                
                # 2. Direction
                direction_score = 1.0 if (expected_return > 0) == (actual_return > 0) else 0.0
                
                # 3. Calibration
                patterns = similarity_results['similar_patterns']
                avg_similarity = np.mean([p['similarity'] for p in patterns])
                calibration_score = max(0.0, 1.0 - abs(avg_similarity - mape_score))
                
                # 4. Variance patterns (TOUS les patterns, pas tronqu√©)
                pattern_returns = []
                for pattern in patterns:  # TOUS les patterns
                    p_ticker = pattern['ticker']
                    p_idx = pattern['window_idx']
                    try:
                        p_data = self.data_manager.load_ticker_chunk(
                            p_ticker, interval, 
                            start_idx=p_idx, 
                            chunk_size=Config.FORECAST_HORIZON+2
                        )
                        if p_data is not None and len(p_data) > Config.FORECAST_HORIZON:
                            p_start = p_data['Close'].iloc[0]
                            p_end = p_data['Close'].iloc[Config.FORECAST_HORIZON]
                            if p_start > 0 and p_end > 0:
                                p_return = (p_end / p_start) - 1
                                pattern_returns.append(p_return)
                    except:
                        continue
                
                if len(pattern_returns) > 1:
                    variance = np.var(pattern_returns)
                    ideal_variance = 0.01
                    variance_score = max(0.0, 1.0 / (1.0 + abs(variance - ideal_variance)))
                else:
                    variance_score = 0.0
                
                # Stocker m√©triques
                metrics['mape'].append(mape_score)
                metrics['direction'].append(direction_score)
                metrics['calibration'].append(calibration_score)
                metrics['pattern_var'].append(variance_score)
                processed_count += 1
                
            except Exception as e:
                logger.debug(f"Ticker {ticker} failed: {str(e)}")
                continue
        
        # Calcul final
        eval_duration = time.time() - eval_start
        
        if processed_count >= 5:
            avg_metrics = {k: np.mean(v) for k, v in metrics.items()}
            
            # Score composite
            composite_score = sum(avg_metrics[k] * self.evaluation_weights[k] for k in self.evaluation_weights)
            
            # Ajustement bas√© sur le succ√®s
            success_ratio = processed_count / len(sample_tickers)
            if success_ratio > 0.7:
                composite_score *= 1.05
            elif success_ratio < 0.3:
                composite_score *= 0.9
            
            print(f"‚úÖ Score={composite_score:.4f}, MAPE={avg_metrics['mape']:.3f}, Dir={avg_metrics['direction']:.3f}")
            print(f"   Success: {processed_count}/{len(sample_tickers)} tickers ({success_ratio:.1%}) in {eval_duration:.1f}s")
            
            return {
                'loss': 1.0 - composite_score,
                'metrics': avg_metrics,
                'processed': processed_count,
                'duration': eval_duration,
                'status': 'ok'
            }
        
        print(f"‚ùå √âchec: {processed_count}/{len(sample_tickers)} tickers en {eval_duration:.1f}s")
        return {
            'loss': 1.0,
            'metrics': {k: 0.0 for k in metrics},
            'processed': processed_count,
            'duration': eval_duration,
            'status': 'fail'
        }


    def optimize_interval(self, interval, n_trials=None, optimization_sample_size=None, resume=True):
        """Optimise avec debug complet"""
        if n_trials is None:
            n_trials = Config.OPT_N_TRIALS
        if optimization_sample_size is None:
            optimization_sample_size = Config.OPT_SAMPLE_SIZE
            
        print(f"{'='*60}")
        print(f"OPTIMIZATION FOR INTERVAL {interval} - {n_trials} TRIALS")
        print(f"{'='*60}")
        
        self.n_trials = n_trials
        self.optimization_start_time = time.time()
        
        # Create interval subdirectory
        interval_dir = os.path.join(self.output_dir, interval)
        os.makedirs(interval_dir, exist_ok=True)
        
        # DEBUG: Afficher l'espace des param√®tres
        print(f"üîç Parameter space:")
        for param, (low, high) in self.param_space.items():
            print(f"   {param}: [{low}, {high}]")
        
        # Create history directory
        history_dir = os.path.join(self.output_dir, "history", interval)
        os.makedirs(history_dir, exist_ok=True)
        
        # Initialize or load optimization history
        history_path = os.path.join(history_dir, f"history_{interval}.pkl")
        print(f"üîç Recherche de l'historique: {history_path}")
        
        if resume and os.path.exists(history_path):
            try:
                print(f"üìÅ Historique trouv√©, chargement...")
                with open(history_path, 'rb') as f:
                    self.optimization_history[interval] = pickle.load(f)
                print(f"üìÇ Historique charg√©: {len(self.optimization_history[interval]['trials'])} essais")
            except Exception as e:
                print(f"‚ùå Erreur de chargement: {str(e)}")
                self.optimization_history[interval] = {
                    'trials': [],
                    'best_trial': None,
                    'metrics_history': {
                        'mape': [], 'direction': [], 'calibration': [], 'pattern_var': []
                    },
                    'params_history': []
                }
        else:
            print(f"‚ö†Ô∏è Historique non trouv√©: {history_path}")
            self.optimization_history[interval] = {
                'trials': [],
                'best_trial': None,
                'metrics_history': {
                    'mape': [], 'direction': [], 'calibration': [], 'pattern_var': []
                },
                'params_history': []
            }
        
        # Get tickers
        tickers = self.vector_analyzer.data_manager.get_available_tickers(interval)
        
        if len(tickers) < 10:
            print(f"‚ùå Not enough tickers for {interval}, using default weights")
            
            financial_metrics = ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI14', 'MACD', 'SMA50', 'EMA20', 'log_return']
            default_weights = {f'{metric}_weight': 1.0 for metric in financial_metrics}
            default_weights.update({'decay_type': 'exponential', 'decay_factor': 0.85})
            
            self.best_weights[interval] = {
                **default_weights,
                'score': 0.5,
                'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            }
            
            save_json(self.best_weights[interval], os.path.join(interval_dir, f"default_weights_{interval}.json"))
            return self.best_weights[interval]
        
        print(f"\nüöÄ Starting optimization with {n_trials} trials on {optimization_sample_size} tickers per trial")
        
        objective = PicklableObjective(
            self, interval, tickers, optimization_sample_size, 
            self.optimization_history[interval]
        )
        
        progress_callback = PicklableProgressCallback(self, interval)
        
        # CORRECTION CRITIQUE: Cr√©er l'espace pour skopt avec ordre correct
        space = []
        param_names = list(self.param_space.keys())
        
        for param_name in param_names:
            low, high = self.param_space[param_name]
            if param_name == 'decay_type_idx':
                space.append(Integer(low, high, name=param_name))
            else:
                space.append(Real(low, high, name=param_name))
        
        print(f"üîç Skopt space order: {[dim.name for dim in space]}")
        
        try:
            # Setup checkpoint saver
            checkpoint_path = os.path.join(interval_dir, f"checkpoint_{interval}.pkl")
            checkpoint_saver = CheckpointSaver(
                checkpoint_path, compress=9, 
                store_objective=False
            )
            
            # Run optimization
            result = gp_minimize(
                objective,
                space,
                n_calls=n_trials,
                n_random_starts=min(5, n_trials),
                random_state=42,
                callback=[checkpoint_saver, progress_callback]
            )
            
            # Extract best trial from history
            best_trial = self.optimization_history[interval]['best_trial']
            
            if best_trial:
                # Save best weights
                self.best_weights[interval] = {
                    **best_trial['params'],
                    'score': best_trial['composite_score'],
                    'mape': best_trial['metrics']['mape'],
                    'direction': best_trial['metrics']['direction'],
                    'calibration': best_trial['metrics']['calibration'],
                    'pattern_var': best_trial['metrics']['pattern_var'],
                    'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                    'trial_id': best_trial['trial_id']
                }
                
                # Save as JSON
                save_json(self.best_weights[interval], os.path.join(interval_dir, f"optimized_weights_{interval}.json"))
                
                # Save centralized version
                custom_weights_dir = os.path.join(self.output_dir, "custom_weights")
                os.makedirs(custom_weights_dir, exist_ok=True)
                
                timestamp = datetime.now().strftime("%Y%m%d")
                save_json(self.best_weights[interval], os.path.join(custom_weights_dir, f"{interval}_weights_{timestamp}.json"))
                save_json(self.best_weights[interval], os.path.join(custom_weights_dir, f"{interval}_weights_latest.json"))
                
                # Final plots - TOUJOURS en fin d'optimisation
                print("üìä G√©n√©ration des plots finaux...")
                self.plot_optimization_history(interval)
                
                total_time = time.time() - self.optimization_start_time
                print(f"\n‚úÖ Optimization completed for {interval} in {total_time:.1f}s")
                print(f"üèÜ Best composite score: {best_trial['composite_score']:.4f}")
                print(f"üìè MAPE: {best_trial['metrics']['mape']:.4f}")
                print(f"üìä Direction: {best_trial['metrics']['direction']:.4f}")
                print(f"üéØ Calibration: {best_trial['metrics']['calibration']:.4f}")
                print(f"üìà Pattern Variance: {best_trial['metrics']['pattern_var']:.4f}")
                
                return self.best_weights[interval]
            else:
                print(f"‚ùå No successful trials for {interval}")
                return None
                
        except KeyboardInterrupt:
            print("\n‚õî Optimization interrupted by user.")
            print("üìä G√©n√©ration des plots finaux apr√®s interruption...")
            self.plot_optimization_history(interval)
            
            if self.optimization_history[interval]['best_trial']:
                best_trial = self.optimization_history[interval]['best_trial']
                print(f"üèÜ Best score so far: {best_trial['composite_score']:.4f}")
                
                self.best_weights[interval] = {
                    **best_trial['params'],
                    'score': best_trial['composite_score'],
                    'interrupted': True
                }
                
                return self.best_weights[interval]
            
            return None

    def plot_optimization_history(self, interval):
        """Plot avec dur√©es correctes et actualisation syst√©matique"""
        if interval not in self.optimization_history:
            return
            
        import matplotlib.pyplot as plt
        history = self.optimization_history[interval]
        plots_dir = os.path.join(self.output_dir, "plots", interval)
        os.makedirs(plots_dir, exist_ok=True)
        
        if not history['trials']:
            return
        
        # CORRECTION: Filtrer explicitement les trials pour cet intervalle
        all_trials = history['trials']
        interval_trials = []
        
        for trial in all_trials:
            # V√©rifier si le trial appartient √† cet intervalle
            trial_params = trial.get('params', {})
            # Si le trial contient des m√©triques financi√®res, c'est probablement le bon intervalle
            has_financial_metrics = any(key.endswith('_weight') for key in trial_params.keys())
            if has_financial_metrics:
                interval_trials.append(trial)
        
        # Si aucun trial filtr√©, utiliser tous les trials (r√©trocompatibilit√©)
        if not interval_trials:
            interval_trials = all_trials
        
        if not interval_trials:
            return
        
        # 1. M√©triques d'optimisation
        plt.figure(figsize=(15, 10))
        trials = range(1, len(interval_trials) + 1)
        
        # Extraire les vraies dur√©es des trials
        real_durations = [trial.get('duration', 0.0) for trial in interval_trials]
        scores = [trial.get('composite_score', 0.0) for trial in interval_trials]
        mape_scores = [trial.get('metrics', {}).get('mape', 0.0) for trial in interval_trials]
        direction_scores = [trial.get('metrics', {}).get('direction', 0.0) for trial in interval_trials]
        
        # Subplot 1: Scores
        plt.subplot(2, 2, 1)
        plt.plot(trials, scores, 'b-', linewidth=2, label='Composite Score')
        plt.plot(trials, mape_scores, 'g-', label='MAPE Score')
        plt.plot(trials, direction_scores, 'r-', label='Direction Score')
        
        if history['best_trial']:
            # Trouver l'index du best trial dans les trials filtr√©s
            best_trial_id = None
            best_trial = history['best_trial']
            for i, trial in enumerate(interval_trials):
                if trial.get('trial_id') == best_trial.get('trial_id'):
                    best_trial_id = i
                    break
            
            if best_trial_id is not None:
                plt.axvline(x=best_trial_id + 1, color='purple', linestyle='--', 
                        label=f"Best Trial: #{best_trial_id + 1}")
        
        plt.xlabel('Trial Number')
        plt.ylabel('Score')
        plt.title(f'Optimization Progress - {interval}')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.ylim(0, 1.05)
        
        # Subplot 2: Dur√©es VRAIES
        plt.subplot(2, 2, 2)
        plt.bar(trials, real_durations, alpha=0.7, color='steelblue')
        plt.xlabel('Trial Number')
        plt.ylabel('Duration (seconds)')
        plt.title(f'Real Trial Durations - {interval}')
        plt.grid(True, alpha=0.3, axis='y')
        
        # Moyenne mobile des dur√©es
        if len(real_durations) >= 3:
            window_size = min(3, len(real_durations))
            moving_avg = np.convolve(real_durations, np.ones(window_size)/window_size, mode='valid')
            ma_trials = range(window_size, len(real_durations) + 1)
            plt.plot(ma_trials, moving_avg, 'r-', linewidth=2, label=f'{window_size}-Trial MA')
            plt.legend()
        
        # Subplot 3: Evolution des poids des m√©triques principales
        plt.subplot(2, 2, 3)
        main_metrics = ['Close_weight', 'Volume_weight', 'RSI14_weight', 'MACD_weight']
        
        for metric in main_metrics:
            metric_values = []
            for trial in interval_trials:
                params = trial.get('params', {})
                value = params.get(metric, 1.0)
                metric_values.append(value)
            
            if metric_values:
                plt.plot(trials, metric_values, label=metric.replace('_weight', ''), linewidth=2)
        
        plt.xlabel('Trial Number')
        plt.ylabel('Weight Value')
        plt.title(f'Financial Metrics Weights Evolution - {interval}')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Subplot 4: Tickers trait√©s par essai
        plt.subplot(2, 2, 4)
        processed_counts = [trial.get('processed_tickers', 0) for trial in interval_trials]
        plt.plot(trials, processed_counts, 'o-', color='orange', linewidth=2)
        plt.xlabel('Trial Number')
        plt.ylabel('Tickers Processed')
        plt.title(f'Processing Success Rate - {interval}')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        
        # Sauvegarde avec timestamp pour tra√ßabilit√©
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        plt.savefig(os.path.join(plots_dir, f"optimization_progress_{interval}.png"), 
                    dpi=150, bbox_inches='tight')
        plt.show()
        plt.close()
        print(f"üìä Plots sauvegard√©s dans {plots_dir}")

# Correction de PicklableObjective pour dur√©es correctes et parall√©lisation
class PicklableObjective:
    def __init__(self, optimizer, interval, tickers, optimization_sample_size, history):
        self.optimizer = optimizer
        self.interval = interval
        self.tickers = tickers
        self.optimization_sample_size = optimization_sample_size
        self.history = history
        self.initial_trials = len(self.history['trials'])
        
    def __call__(self, params):
        trial_index = len(self.history['trials'])
        current_session_trial = trial_index - self.initial_trials + 1
        trial_start = time.time()  # CORRECTION: Mesure r√©elle du temps
        
        print(f"\n{'='*30} TRIAL {current_session_trial}/{self.optimizer.n_trials} {'='*30}")
        
        # Convertir params en dict avec les VRAIES m√©triques
        decay_type_idx = int(params[-1])
        decay_types = ['linear', 'exponential', 'quadratic', 'none']
        weights = {}
        
        param_names = list(self.optimizer.param_space.keys())
        for i, param_name in enumerate(param_names):
            if i < len(params):
                weights[param_name] = params[i]
        
        weights['decay_type'] = decay_types[decay_type_idx % len(decay_types)]
        
        # Afficher les vraies m√©triques financi√®res
        financial_weights = {k: v for k, v in weights.items() if k.endswith('_weight')}
        print(f"üîß Testing financial weights: {financial_weights}")
        
        try:
            # CORRECTION: Parall√©lisation possible ici
            result = self.optimizer._evaluate_weights(
                weights, self.interval, self.tickers, 
                n_samples=self.optimization_sample_size
            )
            
            eval_duration = time.time() - trial_start  # VRAIE dur√©e
            loss = result['loss']
            metrics = result['metrics']
            composite_score = 1.0 - loss
            
            trial_result = {
                'trial_id': trial_index,
                'params': weights.copy(),
                'loss': loss,
                'metrics': metrics.copy(),
                'composite_score': composite_score,
                'duration': eval_duration,  # CORRECTION: Vraie dur√©e
                'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                'processed_tickers': result['processed']
            }
            
            self.history['trials'].append(trial_result)
            
            if (self.history['best_trial'] is None or 
                composite_score > self.history['best_trial']['composite_score']):
                self.history['best_trial'] = trial_result
                print(f"üíé NEW BEST SCORE: {composite_score:.4f}")
            
            # Mise √† jour historique
            self.history['metrics_history']['mape'].append(metrics['mape'])
            self.history['metrics_history']['direction'].append(metrics['direction'])
            self.history['metrics_history']['calibration'].append(metrics['calibration'])
            self.history['params_history'].append(list(params))
            
            # CORRECTION: Plots actualis√©s √† chaque essai
            if trial_index % 5 == 0:  # Tous les 5 essais
                try:
                    self.optimizer.plot_optimization_history(self.interval)
                    print(f"üìä Plots mis √† jour")
                except Exception as e:
                    print(f"‚ö†Ô∏è Erreur plot: {str(e)}")
            
            print(f"\nüìã Trial {trial_index + 1} summary:")
            print(f"‚è±Ô∏è  Real Duration: {eval_duration:.2f}s")  # VRAIE dur√©e
            print(f"üíØ Score: {composite_score:.4f}")
            
            # ETA correct
            elapsed = time.time() - self.optimizer.optimization_start_time
            if current_session_trial > 0:
                avg_trial_time = elapsed / current_session_trial
                remaining_trials = self.optimizer.n_trials - current_session_trial
                eta = max(0, avg_trial_time * remaining_trials)
                print(f"üïí Elapsed: {format_time(elapsed)}, ETA: {format_time(eta)}")
            
            # Sauvegarde
            try:
                history_path = os.path.join(
                    self.optimizer.output_dir, "history", self.interval, 
                    f"history_{self.interval}.pkl"
                )
                os.makedirs(os.path.dirname(history_path), exist_ok=True)
                with open(history_path, 'wb') as f:
                    pickle.dump(self.history, f)
            except Exception as e:
                print(f"‚ùå Erreur sauvegarde: {str(e)}")
            
            return loss
            
        except Exception as e:
            eval_duration = time.time() - trial_start
            print(f"‚ùå ERROR: Trial {trial_index + 1} failed after {eval_duration:.2f}s: {str(e)}")
            
            failed_trial = {
                'trial_id': trial_index,
                'params': weights.copy(),
                'loss': 1.0,
                'metrics': {k: 0.0 for k in ['mape', 'direction', 'calibration', 'pattern_var']},
                'composite_score': 0.0,
                'duration': eval_duration,  # VRAIE dur√©e d'√©chec
                'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                'processed_tickers': 0,
                'failed': True,
                'error': str(e)
            }
            
            self.history['trials'].append(failed_trial)
            return 1.0

# OPTION: Parall√©lisation de l'√©valuation
def parallel_evaluate_weights(optimizer, weights, interval, tickers, n_samples):
    """Version parall√©lis√©e de l'√©valuation"""
    from concurrent.futures import ProcessPoolExecutor
    import multiprocessing as mp
    
    # Diviser les tickers en chunks
    n_workers = min(4, mp.cpu_count())  # Max 4 processus
    chunk_size = len(tickers) // n_workers
    ticker_chunks = [tickers[i:i+chunk_size] for i in range(0, len(tickers), chunk_size)]
    
    # Traitement parall√®le
    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        futures = []
        for chunk in ticker_chunks:
            future = executor.submit(
                optimizer._evaluate_weights_chunk, 
                weights, interval, chunk, len(chunk)
            )
            futures.append(future)
        
        # Collecter r√©sultats
        all_results = []
        for future in futures:
            try:
                result = future.result(timeout=60)
                all_results.append(result)
            except Exception as e:
                print(f"‚ö†Ô∏è Chunk failed: {str(e)}")
                continue
    
    # Agr√©ger les r√©sultats
    if all_results:
        # Moyenner les m√©triques
        final_metrics = {}
        for key in ['mape', 'direction', 'calibration', 'pattern_var']:
            values = [r['metrics'][key] for r in all_results if r['status'] == 'ok']
            final_metrics[key] = np.mean(values) if values else 0.0
        
        total_processed = sum(r['processed'] for r in all_results)
        
        return {
            'metrics': final_metrics,
            'processed': total_processed,
            'status': 'ok' if total_processed > 0 else 'fail'
        }
    
    return {'metrics': {k: 0.0 for k in ['mape', 'direction', 'calibration', 'pattern_var']}, 
            'processed': 0, 'status': 'fail'}


class PicklableProgressCallback:
    def __init__(self, optimizer, interval):
        self.optimizer = optimizer
        self.interval = interval
        # Compter les essais initiaux de l'historique
        self.initial_trials = len(optimizer.optimization_history[interval]['trials'])
        
    def __call__(self, result):
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            
        history = self.optimizer.optimization_history[self.interval]
        trials_completed = len(history['trials'])
        
        # Calculer les essais de la session actuelle
        current_session_trials = trials_completed - self.initial_trials
        
        elapsed = time.time() - self.optimizer.optimization_start_time
        if current_session_trials > 0:
            avg_time = elapsed / current_session_trials
            remaining = max(0, self.optimizer.n_trials - current_session_trials)
            eta = avg_time * remaining
            
            # Utiliser la fonction format_time d√©finie au niveau du module
            print(f"‚è≥ Overall Progress: {current_session_trials}/{self.optimizer.n_trials} trials "
                 f"({min(100, current_session_trials/self.optimizer.n_trials*100):.1f}%) - "
                 f"ETA: {format_time(eta)}")



# Pipeline complet d'apprentissage et d'indexation
def build_vector_index_pipeline(data_manager, intervals=None, train_financial_models=True, compute_embeddings=True, epochs=None, batch_size=None):
    """Ex√©cute le pipeline complet avec param√®tres ajustables"""
    intervals = intervals if intervals is not None else Config.INTERVALS
    if epochs is None:
        epochs = Config.NUM_EPOCHS
    logger.info("D√©marrage du pipeline d'indexation vectorielle")
    
    for interval in intervals:
        logger.info(f"Traitement de l'intervalle: {interval}")
        
        # 1. Entra√Ænement ou chargement du mod√®le
        if train_financial_models:
            logger.info(f"Entra√Ænement du mod√®le pour {interval}")
            model, history = train_financial_model(
                data_manager, 
                interval, 
                epochs=epochs, 
                batch_size=batch_size
            )
            if model is None:
                logger.warning(f"Impossible d'entra√Æner le mod√®le pour {interval}")
                continue
        else:
            # Charger le mod√®le existant
            model_path = os.path.join(Config.MODELS_PATH, f"encoder_{interval}.pth")
            if os.path.exists(model_path):
                logger.info(f"Chargement du mod√®le existant: {model_path}")
                try:
                    checkpoint = torch.load(model_path)
                    model = FinancialTimeSeriesEncoder(
                        input_dim=checkpoint['config']['input_dim'],
                        seq_len=checkpoint['config']['seq_len'],
                        embedding_dim=checkpoint['config']['embedding_dim']
                    )
                    model.load_state_dict(checkpoint['model_state_dict'])
                    model.eval()
                except Exception as e:
                    logger.error(f"Erreur lors du chargement du mod√®le: {str(e)}")
                    continue
            else:
                logger.warning(f"Mod√®le non trouv√© pour {interval}: {model_path}")
                continue
        
        # 2. G√©n√©ration des embeddings (seulement si compute_embeddings est True)
        if compute_embeddings:
            logger.info(f"G√©n√©ration des embeddings pour {interval}")
            
            # üî• OPTIMISATION: R√©utiliser le dataset d√©j√† cr√©√©
            if 'dataset' in locals():
                logger.info(f"R√©utilisation du dataset d'entra√Ænement pour les embeddings")
                embeddings_list = []
                metadata_list = []
                
                # Traiter par gros batches GPU
                batch_size = 2048
                model.eval()
                
                with torch.no_grad():
                    for i in range(0, len(dataset), batch_size):
                        batch_end = min(i + batch_size, len(dataset))
                        
                        # R√©cup√©rer les tenseurs d√©j√† normalis√©s
                        batch_tensors = []
                        for j in range(i, batch_end):
                            tensor, _, regime = dataset.windows[j]
                            batch_tensors.append(tensor)
                        
                        # Stack et encoder sur GPU
                        batch_tensor = torch.stack(batch_tensors).to(Config.DEVICE)
                        batch_embeddings = model.encode(batch_tensor).cpu().numpy()
                        
                        # M√©tadonn√©es simplifi√©es
                        for j, embedding in enumerate(batch_embeddings):
                            embeddings_list.append(embedding)
                            metadata_list.append({
                                'ticker': f'window_{i+j}',
                                'interval': interval,
                                'window_idx': i+j,
                                'timestamp': None
                            })
                        
                        if i % (batch_size * 5) == 0:
                            logger.info(f"Embeddings: {len(embeddings_list)}/{len(dataset)} ({len(embeddings_list)/len(dataset)*100:.1f}%)")
                
                embeddings = np.array(embeddings_list)
                metadata = metadata_list
                
            else:
                # Fallback vers l'ancienne m√©thode
                embeddings, metadata = generate_embeddings(data_manager, interval, model)

            # 3. Construction de l'index FAISS
            logger.info(f"Construction de l'index FAISS pour {interval}")
            if embeddings is not None and len(embeddings) > 0:
                nlist = min(100, max(1, len(embeddings) // 1000))
                index = build_faiss_index(embeddings, interval, nlist=nlist)
                if index is not None:
                    logger.info(f"‚úÖ Index FAISS cr√©√© avec {len(embeddings)} vecteurs")
                else:
                    logger.warning(f"‚ùå √âchec cr√©ation index FAISS pour {interval}")
            else:
                logger.warning(f"‚ùå Pas d'embeddings pour cr√©er l'index {interval}")
        
        logger.info(f"Pipeline termin√© avec succ√®s pour {interval}")
    
    logger.info("Pipeline d'indexation vectorielle termin√©")
    return True


class NumpyJSONEncoder(json.JSONEncoder):
    """Encodeur JSON personnalis√© qui g√®re les types NumPy"""
    def default(self, obj):
        if isinstance(obj, (np.integer, np.int64, np.int32, np.int16, np.int8)):
            return int(obj)
        elif isinstance(obj, (np.floating, np.float64, np.float32, np.float16)):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, np.bool_):
            return bool(obj)
        return super(NumpyJSONEncoder, self).default(obj)

def save_json(data, filepath):
    """Sauvegarde des donn√©es en JSON en g√©rant les types NumPy"""
    with open(filepath, 'w') as f:
        json.dump(data, f, indent=2, cls=NumpyJSONEncoder)

def format_time(seconds):
    """Formate un nombre de secondes en texte lisible (heures, minutes, secondes)"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    if hours > 0:
        return f"{hours}h {minutes:02d}m {secs:02d}s"
    elif minutes > 0:
        return f"{minutes}m {secs:02d}s"
    else:
        return f"{secs}s"
        
def integrate_all_enhancements():
    # Apply all enhancements
    
    def weights_optimization_example(market_analyzer, interval='1d', n_trials=Config.OPT_N_TRIALS):
        # Cr√©er un chemin absolu pour le r√©pertoire d'optimisation
        import os
        # Utiliser le chemin de base de votre projet comme racine
        base_dir = Config.BASE_PATH
        output_dir = os.path.join(base_dir, "optimization_results")
        
        # Cr√©er l'optimiseur avec le chemin absolu
        optimizer = VectorWeightOptimizer(
            vector_analyzer=market_analyzer.vector_analyzer,
            intervals=[interval],
            output_dir=output_dir  # Utiliser le chemin absolu ici
        )
        
        # Ex√©cuter l'optimisation
        weights = optimizer.optimize_interval(interval, n_trials=n_trials, optimization_sample_size=Config.OPT_SAMPLE_SIZE)
        
        # Appliquer les poids optimis√©s
        market_analyzer.vector_analyzer.load_optimized_weights(weights_dir=os.path.join(output_dir, "custom_weights"))
        
        print(f"Optimized weights applied for {interval}")
        return optimizer
    
    def enhanced_analysis_example(market_analyzer, ticker, interval='1d', use_dtw=True):
        if use_dtw:
            result = market_analyzer.analyze_with_dtw(
                ticker, interval=interval, 
                dtw_percent=Config.DTW_PERCENT,
                early_stop_threshold=Config.EARLY_STOPPING_PATIENCE
            )
        else:
            result = market_analyzer.analyze_ticker(ticker, interval=interval)
            
        return result
        
    return weights_optimization_example, enhanced_analysis_example


if torch.cuda.is_available():
    torch.cuda.empty_cache()
    print(f"M√©moire GPU disponible: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
    print(f"M√©moire GPU utilis√©e: {torch.cuda.memory_allocated() / 1e9:.2f} GB")
    
import os
import torch

# Constantes globales pour le traitement par lots
BATCH_SIZE = Config.get_batch_size()
print("R√©pertoire de travail actuel:", os.getcwd())


def main(config=Config, target_interval='1d'):
    """Point d'entr√©e principal avec configuration flexible"""
    print("üöÄ D√©marrage de l'analyse vectorielle des march√©s")
    
    # Fonction helper pour r√©cup√©rer la config
    def get_config_value(config_obj, key_lower, key_upper, default):
        if hasattr(config_obj, key_lower):
            return getattr(config_obj, key_lower)
        elif hasattr(config_obj, key_upper):
            return getattr(config_obj, key_upper)
        else:
            return default

    config_dict = {
        'base_path': get_config_value(config, 'base_path', 'BASE_PATH', 
                                      r"C:\Users\Alex\Python\Finance\@@@ Downloader @@@\parquet\datasets_technicals"),
        'build_indices': get_config_value(config, 'build_indices', 'BUILD_INDICES', True),
        'train_models': get_config_value(config, 'train_models', 'TRAIN_MODELS', True),
        'train_epochs': get_config_value(config, 'train_epochs', 'NUM_EPOCHS', 100),
        'batch_size': get_config_value(config, 'batch_size', 'BATCH_SIZE', 1024),
        'intervals': get_config_value(config, 'intervals', 'INTERVALS', ['1d']),
        'run_optimization': get_config_value(config, 'run_optimization', 'RUN_OPTIMIZATION', False),
        'test_interval': target_interval,
        'test_ticker': get_config_value(config, 'test_ticker', 'TEST_TICKER', 'TSLA'),
        'test_analyzer': get_config_value(config, 'test_analyzer', 'TEST_ANALYZER', True),
        'use_dtw': get_config_value(config, 'use_dtw', 'USE_DTW', False),
        'scan_market': get_config_value(config, 'scan_market', 'SCAN_MARKET', False),
        'market_scan_size': get_config_value(config, 'market_scan_size', 'MARKET_SCAN_SIZE', 100),
        'optimization_trials': get_config_value(config, 'optimization_trials', 'OPT_N_TRIALS', 100),
        'compute_embeddings': get_config_value(config, 'compute_embeddings', 'COMPUTE_EMBEDDINGS', True),
        'forecast_horizon': get_config_value(config, 'forecast_horizon', 'FORECAST_HORIZON', Config.FORECAST_HORIZON),
        'historical_bars': get_config_value(config, 'historical_bars', 'HISTORICAL_BARS', Config.HISTORICAL_BARS),
        'save_plots': get_config_value(config, 'save_plots', 'SAVE_PLOTS', True),
        'plot_dir': get_config_value(config, 'plot_dir', 'PLOT_DIR', 'plots')
    }

    print("üìã Configuration active :")
    for key, value in config_dict.items():
        if key in ['train_models', 'compute_embeddings', 'build_indices', 'run_optimization']:
            print(f"   {key}: {value}")

    base_path = config_dict['base_path']

    # Initialisation du data manager
    data_manager = StreamingDataManager(
        base_path=base_path,
        chunk_size=getattr(config, 'CHUNK_SIZE', 1000),
        cache_capacity_mb=getattr(config, 'CACHE_CAPACITY_MB', 2000),
        use_gpu=getattr(config, 'USE_GPU', torch.cuda.is_available())
    )

    # √âtape 1 : Indexation vectorielle
    if config_dict['build_indices']:
        print(f"\nüèóÔ∏è Ex√©cution du pipeline d'indexation vectorielle")
        print(f"   - Entra√Ænement des mod√®les: {config_dict['train_models']}")
        print(f"   - Calcul des embeddings: {config_dict['compute_embeddings']}")
        try:
            build_vector_index_pipeline(
                data_manager, 
                intervals=config_dict['intervals'], 
                train_financial_models=config_dict['train_models'],
                compute_embeddings=config_dict['compute_embeddings'],
                epochs=config_dict['train_epochs'],
                batch_size=config_dict['batch_size']
            )
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
                print("‚úÖ Lib√©ration de la m√©moire GPU")
                
            # üîç ATTENDRE que les mod√®les soient cr√©√©s
            for interval in config_dict['intervals']:
                model_path = os.path.join(Config.MODELS_PATH, f"encoder_{interval}.pth")
                if not os.path.exists(model_path):
                    print(f"‚è≥ Attente de la cr√©ation du mod√®le {interval}...")
                    return None  # Arr√™ter ici si l'entra√Ænement a √©chou√©
                    
            print("‚úÖ Pipeline d'indexation termin√© avec succ√®s")
            
        except Exception as e:
            print(f"‚ùå Erreur durant l'indexation: {str(e)}")
            import traceback
            print(traceback.format_exc())
            return None

    # √âtape 2 : Int√©gration des am√©liorations
    weights_optimization, enhanced_analysis = integrate_all_enhancements()
    print("\nüîß Am√©liorations int√©gr√©es: optimisation des poids, analyses DTW, visualisations FLASC")

    # √âtape 3 : Cr√©ation de l'analyseur (APR√àS l'indexation)
    market_analyzer = VectorMarketAnalyzer(base_path, use_gpu=torch.cuda.is_available())

    # √âtape 4 : Optimisation
    if config_dict['run_optimization']:
        print(f"\n‚öôÔ∏è Optimisation des poids vectoriels pour {target_interval}")
        optimization_output_dir = r"C:\Users\Alex\Python\Finance\Strat002. SP_DTW\optimization_results"
        optimizer = VectorWeightOptimizer(
            vector_analyzer=market_analyzer.vector_analyzer,
            intervals=[target_interval],
            output_dir=optimization_output_dir
        )
        best_weights = optimizer.optimize_interval(
            target_interval,
            n_trials=config_dict['optimization_trials']
        )
        print("‚úÖ Optimisation termin√©e")
    else:
        weights_loaded = market_analyzer.vector_analyzer.load_optimized_weights(
            weights_dir=r"C:\Users\Alex\Python\Finance\Strat002. SP_DTW\optimization_results\custom_weights"
        )
        print("\nüìà Poids optimis√©s charg√©s" if weights_loaded else "\n‚öôÔ∏è Aucun poids optimis√© trouv√©, utilisation des poids par d√©faut")

    # √âtape 5 : Analyse de test
    if config_dict['test_analyzer']:
        print(f"\nüîç Test d'analyse sur {config_dict['test_ticker']} ({target_interval})")
        result = enhanced_analysis(
            market_analyzer, 
            config_dict['test_ticker'], 
            interval=target_interval,
            use_dtw=config_dict['use_dtw']
        )
        if result:
            print(f"‚úÖ Analyse termin√©e pour {config_dict['test_ticker']}: "
                  f"Rendement attendu: {result.get('expected_return', 0):.2%}")
            print(f"\nüìä G√©n√©ration des visualisations pour {config_dict['test_ticker']}...")
            try:
                market_analyzer.plot_analysis_results(result, show_clusters=True)
                print("‚úÖ Visualisations FLASC g√©n√©r√©es et sauvegard√©es")
            except Exception as e:
                print(f"‚ö†Ô∏è Erreur lors de la g√©n√©ration des visualisations: {str(e)}")
                import traceback
                print(traceback.format_exc())

    # √âtape 6 : Scan de march√©
    if config_dict['scan_market']:
        print(f"\nüîé Lancement du scan de march√© sur {target_interval}...")
        opportunities = market_analyzer.scan_market(
            target_interval, 
            min_return=0.05, 
            max_tickers=config_dict['market_scan_size']
        )
        if opportunities:
            print("\nüèÜ Top opportunit√©s:")
            for i, opp in enumerate(opportunities[:30]):
                print(f"{i+1}. {opp['ticker']} ({opp['interval']}): "
                      f"{opp['expected_return']*100:.2f}% ({opp['direction']})")

    print("\n‚úÖ Ex√©cution termin√©e")
    return market_analyzer


# Ex√©cution principale
if __name__ == "__main__":
    try:
        class CustomConfig:
            def __init__(self):
                for attr in dir(Config):
                    if not attr.startswith('_'):
                        setattr(self, attr, getattr(Config, attr))
                self.train_models = False
                self.compute_embeddings = False
                self.build_indices = False
                self.test_analyzer = True
                self.scan_market = False
                self.run_optimization = True
                self.optimization_trials = 6
                self.save_plots = True
                self.forecast_horizon = 20
                self.historical_bars = 50

        custom_config = CustomConfig()
        print("üîß Configuration personnalis√©e appliqu√©e:")
        print(f"   - Entra√Ænement: {custom_config.train_models}")
        print(f"   - Embeddings: {custom_config.compute_embeddings}")
        print(f"   - Pipeline: {custom_config.build_indices}")
        
        analyzer = main(config=custom_config, target_interval='15m')

    except NameError as e:
        print(f"‚ùå Erreur: {str(e)}")
        raise
```
